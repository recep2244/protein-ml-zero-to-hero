## Interview Tip Bank
> **Itertools trick**: `combinations_with_replacement('ATGC', 2)` generates all 2-mers including repeats — useful for k-mer enumeration in bioinformatics.

> **Counter trick**: `Counter(a) - Counter(b)` removes elements in b from a, but only keeps positive counts. Good for "find characters not in common."

> **deque rotation**: `d.rotate(n)` rotates right by n. Negative n rotates left. Used in circular buffer problems.

> **Regex gotcha**: `re.match()` only matches at the START of the string. Use `re.search()` to find anywhere. Use `re.fullmatch()` to match the ENTIRE string.

# SECTION 12: Closures & Decorators

````python
# HackerRank Section 12 — Closures & Decorators
import time
from functools import wraps

# Closure: inner function retains outer scope
def make_gc_calculator(threshold):
    """Factory: returns a function tuned to a GC threshold."""
    def check_gc(seq):
        gc = (seq.count('G') + seq.count('C')) / len(seq)
        return gc, gc > threshold
    return check_gc

is_gc_rich = make_gc_calculator(0.6)
is_gc_poor = make_gc_calculator(0.3)

seqs = ["GCGCGCGC", "ATATAT", "ATGCGA"]
print("GC analysis:")
for s in seqs:
    gc, rich = is_gc_rich(s)
    print(f"  '{s}': GC={gc:.2f}, GC-rich (>0.6)={rich}")

# Decorator: wraps a function to add behavior
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        elapsed = time.time() - start
        print(f"  {func.__name__} took {elapsed:.6f}s")
        return result
    return wrapper

@timer
def count_kmers(seq, k=3):
    return {seq[i:i+k] for i in range(len(seq)-k+1)}

dna = "ATGCGATCGATCGATCG" * 10
result = count_kmers(dna, k=3)
print(f"\nUnique 3-mers: {len(result)}")
````

**Real-world application**: Substring counting in genome annotation tools

```python
# Stuart wins points for substrings starting with consonants.
# Kevin wins points for substrings starting with vowels.
# Score += number of substrings starting at position i = (n - i)

def minion_game(s: str):
    vowels = set('AEIOU')
    n = len(s)
    kevin, stuart = 0, 0
    for i, c in enumerate(s):
        if c in vowels:
            kevin  += n - i   # all substrings starting at i
        else:
            stuart += n - i
    if kevin > stuart:
        print(f"Kevin {kevin}")
    elif stuart > kevin:
        print(f"Stuart {stuart}")
    else:
        print("Draw")

minion_game('BANANA')   # Stuart 12
minion_game('AEIOU')    # Kevin 15
minion_game('ATGCATG')  # Try this yourself
```

## Problem 2 — Maximize It! (HackerRank Itertools — Hard)
**HackerRank**: [hackerrank.com/challenges/maximize-it](https://www.hackerrank.com/challenges/maximize-it)
**Real-world**: Feature combination search in ML hyperparameter tuning

```python
from itertools import product

def maximize_it(k: int, m: int, lists: list[list[int]]) -> int:
    """
    Pick one element from each of the k lists.
    Maximize (sum of squares) % m.
    Returns the maximum value.
    """
    # Precompute squares mod m (reduces search space)
    squared = [[x**2 % m for x in lst] for lst in lists]
    max_val = 0
    for combo in product(*squared):
        max_val = max(max_val, sum(combo) % m)
    return max_val

# Test
```

```python
from collections import OrderedDict

def word_order(words: list[str]) -> tuple[int, list[int]]:
    """
    Count distinct words, preserving first-occurrence order.
    Return (n_distinct, [count1, count2, ...]) in order of first occurrence.
    """
    d = OrderedDict()
    for w in words:
        d[w] = d.get(w, 0) + 1
    return len(d), list(d.values())
```

## Interview Q&A — Python Fundamentals

**Q: What is the difference between a list and a tuple in Python?**
A: Lists are mutable (can be changed after creation): `[1,2,3]`. Tuples are immutable: `(1,2,3)`. Tuples are faster and hashable (can be dict keys/set elements). Use tuples for fixed data, lists for collections you'll modify.

**Q: How does Python's GIL affect multi-threading?**
A: The Global Interpreter Lock (GIL) prevents true parallel CPU execution in CPython. For CPU-bound tasks: use multiprocessing. For I/O-bound tasks (network, disk): threading still helps. NumPy/SciPy release the GIL for numerical operations.

**Q: What is a generator and why use one for bioinformatics?**
A: A generator produces values lazily (`yield` keyword). For genomics: a genome file can be 3GB — loading it all into a list would crash. A generator reads one sequence at a time: `for seq in parse_fasta(file):` without storing the whole file in memory.

**Q: Explain list comprehension vs for loop — when does it matter?**
A: List comprehension `[f(x) for x in seq]` creates the full list in memory. For large genomic data, use generator expressions `(f(x) for x in seq)` instead — same syntax, lazy evaluation, O(1) memory.

**Q: What is the time complexity of `in` for list vs set?**
A: `x in list` is O(n) — linear scan. `x in set` is O(1) average — hash lookup. For checking if a k-mer appears in a large set of sequences, always use a `set` or `frozenset`, not a list.

**Q: What are *args and **kwargs?**
A: `*args` captures positional arguments as a tuple; `**kwargs` captures keyword arguments as a dict. Used to write flexible functions: `def log(*args, **kwargs): print(args, kwargs)`.

````python
# DEBUG EXERCISE — find and fix 3 bugs

# Bug 1: mutating list while iterating over it
nums = [1, 2, 3, 4, 5]
doubled = []
for x in nums:
    nums.remove(x)          # mutates nums mid-iteration — skips elements
    doubled.append(x * 2)
print("doubled:", doubled)  # expected [2, 4, 6, 8, 10], got fewer items


# Bug 2: mutable default argument
def append_val(val, lst=[]):   # lst persists between calls!
    lst.append(val)
    return lst

print("append_val([1]):", append_val(1))   # expected [1]
print("append_val([2]):", append_val(2))   # expected [2], got [1, 2]


# Bug 3: comprehension filters wrong parity
squares_of_evens = [x**2 for x in range(1, 6) if x % 2 == 1]  # odd, not even
print("squares of evens:", squares_of_evens)  # expected [4, 16], got [1, 9, 25]
````
