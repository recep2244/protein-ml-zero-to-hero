## Statistics Formulas Quick Reference

| Concept | Formula | When to Use |
|---------|---------|-------------|
| Mean | Σxᵢ/n | Center of symmetric distribution |
| Median | Middle value | Center of skewed distribution |
| Variance | Σ(xᵢ-μ)²/n | Spread (population) |
| Std Dev | √Variance | Same units as data |
| IQR | Q3 - Q1 | Robust spread measure |
| Pearson r | Cov(X,Y)/(σₓσᵧ) | Linear correlation |
| Spearman ρ | Pearson r of ranks | Monotonic/nonlinear correlation |
| Least squares b1 | Σ(xᵢ-x̄)(yᵢ-ȳ) / Σ(xᵢ-x̄)² | Regression slope |
| t-statistic | (x̄-μ)/(s/√n) | 1-sample t-test |

**Quick Glossary — Statistics Terms**

| Term | Plain English |
|------|--------------|
| **Mean** | Average: sum / count |
| **Median** | Middle value when sorted |
| **Standard deviation** | How spread out values are around the mean |
| **Normal distribution** | Bell curve: most values near mean, few far away |
| **Correlation** | How much two variables move together (−1 to +1) |
| **p-value** | Probability of seeing this result by random chance |
| **Null hypothesis** | The default assumption ("no effect") we're testing against |

# DAY 1: Quartiles, IQR, Standard Deviation

````python
# Day 1 — Quartiles, IQR, Standard Deviation
import numpy as np

data = [3, 7, 8, 5, 12, 14, 21, 13, 18]
n = len(data)
data_sorted = sorted(data)

# Quartiles (HackerRank method: split at median)
def hr_quartiles(arr):
    """HackerRank uses: Q1=median(lower half), Q3=median(upper half)."""
    arr = sorted(arr)
    n = len(arr)
    mid = n // 2
    lower = arr[:mid]
    upper = arr[mid+1:] if n % 2 == 1 else arr[mid:]
    def med(a): n=len(a); return (a[n//2-1]+a[n//2])/2 if n%2==0 else a[n//2]
    return med(lower), med(arr), med(upper)

q1, q2, q3 = hr_quartiles(data)
iqr = q3 - q1
print(f"Data (sorted): {data_sorted}")
print(f"Q1={q1}, Q2={q2}, Q3={q3}")
print(f"IQR = Q3 - Q1 = {iqr}")

# Standard deviation (population)
mean = sum(data) / n
variance = sum((x - mean)**2 for x in data) / n
std = variance**0.5
print(f"\nMean = {mean:.2f}")
print(f"Population variance = {variance:.2f}")
print(f"Population std = {std:.2f}")
print(f"\nNumPy confirmation: std={np.std(data):.2f}, var={np.var(data):.2f}")

# Outlier detection with IQR fence
lower_fence = q1 - 1.5*iqr
upper_fence = q3 + 1.5*iqr
outliers = [x for x in data if x < lower_fence or x > upper_fence]
print(f"Outlier fences: [{lower_fence:.1f}, {upper_fence:.1f}]")
print(f"Outliers: {outliers}")
````

## Problem 1 — A/B Test for Clinical Trial (Hypothesis Testing)
**Real world**: Pharma/biotech companies run statistical tests on clinical trial data
**Skills**: t-test, p-value interpretation, effect size (Cohen's d)

```python
import numpy as np
from scipy import stats

# Clinical trial: new drug vs placebo for blood pressure reduction
np.random.seed(42)
n_drug, n_placebo = 50, 50

# Simulate: drug lowers BP by ~15 mmHg, placebo by ~5 mmHg
drug_reduction    = np.random.normal(15, 8, n_drug)
placebo_reduction = np.random.normal(5, 8, n_placebo)

# Two-sample t-test (independent groups)
t_stat, p_value = stats.ttest_ind(drug_reduction, placebo_reduction, equal_var=False)

# Effect size (Cohen's d)
pooled_std = np.sqrt((drug_reduction.std()**2 + placebo_reduction.std()**2) / 2)
cohens_d = (drug_reduction.mean() - placebo_reduction.mean()) / pooled_std

print(f"Drug group:    mean={drug_reduction.mean():.2f} ± {drug_reduction.std():.2f} mmHg")
print(f"Placebo group: mean={placebo_reduction.mean():.2f} ± {placebo_reduction.std():.2f} mmHg")
print(f"t-statistic: {t_stat:.4f}")
print(f"p-value: {p_value:.6f}")
print(f"Cohen's d: {cohens_d:.3f}  ({'large' if cohens_d > 0.8 else 'medium' if cohens_d > 0.5 else 'small'})")
print(f"Conclusion: {'Reject H0 — drug is effective' if p_value < 0.05 else 'Fail to reject H0'}")
# Multiple testing note: if running 100 drugs, expect 5 false positives at α=0.05
```

## Problem 2 — Predicting Protein Melting Temperature (Regression)
**Dataset**: [Kaggle: Novozymes Enzyme Stability](https://www.kaggle.com/competitions/novozymes-enzyme-stability-prediction)
**Skills**: linear regression, RMSE, R², feature engineering

```python
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Simulate enzyme stability dataset
# Features: sequence-based (AA composition, length, charge)
np.random.seed(42)
n = 300

# Simple features correlated with Tm
seq_length = np.random.randint(100, 500, n)
gly_content = np.random.uniform(0.03, 0.12, n)    # Gly% (flexibility → lower Tm)
pro_content = np.random.uniform(0.02, 0.10, n)    # Pro% (rigidity → higher Tm)
charged_fraction = np.random.uniform(0.15, 0.35, n)

# Tm model (simplified from real data)
Tm = (60
      + 0.02 * seq_length           # longer → slightly more stable
      - 50 * gly_content            # glycine destabilizes
      + 80 * pro_content            # proline stabilizes (rigid)
      + 20 * charged_fraction       # charge stabilizes
      + np.random.normal(0, 5, n))  # noise

X = np.column_stack([seq_length, gly_content, pro_content, charged_fraction])
feature_names = ['Length', 'Gly%', 'Pro%', 'Charged%']

# Regression
from sklearn.model_selection import train_test_split
X_tr, X_te, y_tr, y_te = train_test_split(X, Tm, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_tr, y_tr)
y_pred = model.predict(X_te)

print("Linear Regression for Tm prediction:")
for name, coef in zip(feature_names, model.coef_):
    print(f"  {name:12s}: {coef:+.3f}")
print(f"  Intercept: {model.intercept_:.3f}")
print(f"R²:   {r2_score(y_te, y_pred):.4f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_te, y_pred)):.3f} °C")

# Pearson correlation of each feature with Tm
print("\nFeature-Tm correlations:")
for i, name in enumerate(feature_names):
    r, p = stats.pearsonr(X[:, i], Tm)
    print(f"  {name}: r={r:.3f}, p={p:.4f}")
```

```python
# Predict house prices from multiple features
# This is the exact format of HackerRank's Day 9 challenge

import numpy as np
from scipy import stats

def solve_day9_regression():
    """
    HackerRank Day 9 format:
    Input: f (features), n (training samples), training data, m (test samples), test data
    Output: predicted values for test samples
    """
    # Sample data (2 features: size, rooms)
    train_data = np.array([
        [0.18, 0.89, 109.85],   # feature1, feature2, price
        [1.0,  0.26, 155.72],
        [0.92, 0.11, 137.66],
        [0.07, 0.37, 76.17],
        [0.85, 0.16, 139.75],
        [0.99, 0.41, 162.6],
        [0.87, 0.47, 151.77],
    ])
    test_data = np.array([
        [0.49, 0.18],
        [0.57, 0.83],
        [0.56, 0.64],
        [0.76, 0.18],
    ])
    
    # Solve using normal equation: beta = (X^T X)^{-1} X^T y
    X = np.column_stack([np.ones(len(train_data)), train_data[:, :-1]])  # add intercept
    y = train_data[:, -1]
    
    beta = np.linalg.lstsq(X, y, rcond=None)[0]
    
    # Predict
    X_test = np.column_stack([np.ones(len(test_data)), test_data])
    predictions = X_test @ beta
    
    print("Day 9 MLR predictions:")
    for pred in predictions:
        print(f"  {pred:.2f}")
    # Expected: ~105.22, ~142.68, ~132.94, ~129.71

solve_day9_regression()
```

## 🐛 Debug Exercise — Statistics Calculation Bugs

Find and fix **3 bugs** in this statistics pipeline.

<details><summary>Show bugs</summary>

**Bug 1:** `np.mean(data)` computes population mean correctly, but `std` uses `ddof=0` (population) instead of `ddof=1` (sample) for small samples — causes underestimation.

**Bug 2:** Pearson correlation: dividing by `n` instead of `n-1` in covariance — same ddof issue.

**Bug 3:** The z-score normalization uses the wrong formula: `(x - min) / (max - min)` is min-max scaling, NOT z-score. Z-score is `(x - mean) / std`.

</details>

````python
import numpy as np
np.random.seed(42)
data = np.array([2, 4, 4, 4, 5, 5, 7, 9], dtype=float)

# BUG 1: wrong ddof for sample std
sample_std = np.std(data)           # should be np.std(data, ddof=1)
print(f"Sample std (buggy):  {sample_std:.4f}")
print(f"Sample std (correct): {np.std(data, ddof=1):.4f}  <- should be ~2.0")

x = np.array([1,2,3,4,5], dtype=float)
y = np.array([2,4,5,4,5], dtype=float)

# BUG 2: Pearson correlation with wrong ddof
cov_buggy = np.sum((x - x.mean()) * (y - y.mean())) / len(x)  # should divide by n-1
cov_correct = np.cov(x, y)[0,1]  # np.cov uses ddof=1 by default
print(f"Cov (buggy):  {cov_buggy:.4f}")
print(f"Cov (correct): {cov_correct:.4f}")

# BUG 3: wrong normalization (min-max instead of z-score)
z_buggy  = (data - data.min()) / (data.max() - data.min())  # min-max, not z-score!
z_correct = (data - data.mean()) / np.std(data, ddof=1)
print(f"Z-score (buggy):  {z_buggy.round(3)}")
print(f"Z-score (correct): {z_correct.round(3)}  <- should have mean≈0, std≈1")
print(f"Mean of correct z: {z_correct.mean():.4f} (should be ~0)")
````
