## 1.2 Lists, Tuples, Sets, Dicts

## ⚠️ Python Syntax Note: List Comprehensions

The next cell uses a **list comprehension** — a compact way to build a list:
```python
# Normal loop:
result = []
for x in my_list:
    if condition(x):
        result.append(transform(x))

# List comprehension (same thing, one line):
result = [transform(x) for x in my_list if condition(x)]
```
Read it as: "give me `transform(x)` for every `x` in `my_list` where `condition(x)` is True."


````python
# HackerRank 1.2 — Lists, Tuples, Sets, Dicts
from collections import defaultdict, Counter

# Lists
lst = [3, 1, 4, 1, 5, 9, 2, 6]
lst.sort()
print(f"Sorted: {lst}")
print(f"Sorted reverse: {sorted(lst, reverse=True)}")
print(f"List comprehension (squares): {[x**2 for x in lst if x % 2 == 0]}")

# Tuples (immutable, faster than lists)
point = (3, 4)
x, y = point  # unpacking
print(f"\nTuple unpacking: x={x}, y={y}")
print(f"Distance from origin: {(x**2 + y**2)**0.5:.2f}")

# Sets — O(1) lookup
genes_A = {'BRCA1', 'TP53', 'MYC', 'EGFR'}
genes_B = {'TP53', 'KRAS', 'MYC', 'PTEN'}
print(f"\nIntersection: {genes_A & genes_B}")
print(f"Union: {genes_A | genes_B}")
print(f"Difference A-B: {genes_A - genes_B}")

# Dicts — Counter and defaultdict
expr = Counter({'BRCA1': 5, 'TP53': 10, 'MYC': 8, 'TP53': 10})
print(f"\nTop 2 expressed: {expr.most_common(2)}")

# defaultdict avoids KeyError
pathway = defaultdict(list)
for gene, pw in [('BRCA1','DNA_repair'), ('TP53','apoptosis'), ('MYC','cell_cycle')]:
    pathway[pw].append(gene)
print(f"Pathways: {dict(pathway)}")
````

````python
# HackerRank 1.3 — Functions, Lambdas, Comprehensions
from functools import reduce

# Functions with *args and **kwargs
def align_sequences(*seqs, gap_char='-', verbose=False):
    """Mock alignment function — shows *args, **kwargs patterns."""
    if verbose:
        print(f"Aligning {len(seqs)} sequences")
    return [s.replace('T', 'U') for s in seqs]  # DNA→RNA

result = align_sequences("ATGC", "TTTT", gap_char='-', verbose=True)
print(f"Transcribed: {result}")

# Lambda functions
gc_content = lambda s: (s.count('G') + s.count('C')) / len(s)
seqs = ["GCGCGC", "ATATAT", "ATGCGA"]
print(f"\nGC contents: {list(map(gc_content, seqs))}")

# Sorted with lambda key
seqs_sorted = sorted(seqs, key=gc_content, reverse=True)
print(f"Sorted by GC: {seqs_sorted}")

# Comprehensions — faster than for loops
nucleotides = 'ACGT'
all_codons = [a+b+c for a in nucleotides for b in nucleotides for c in nucleotides]
print(f"\nTotal codons: {len(all_codons)}")
stop_codons = {'TAA', 'TAG', 'TGA'}
coding_codons = [c for c in all_codons if c not in stop_codons]
print(f"Coding codons: {len(coding_codons)}")

# map/filter/reduce
lengths = list(map(len, seqs))
long_seqs = list(filter(lambda s: len(s) > 5, seqs))
total_len = reduce(lambda a, b: a + len(b), seqs, 0)
print(f"\nlengths={lengths}, long_seqs={long_seqs}, total={total_len}")
````

````python
# HackerRank 1.5 — Error Handling, Generators, Itertools
import itertools

# Exception handling
def safe_gc(seq):
    try:
        if not isinstance(seq, str): raise TypeError(f"Expected str, got {type(seq).__name__}")
        valid = set('ACGTN')
        invalid = set(seq.upper()) - valid
        if invalid: raise ValueError(f"Invalid characters: {invalid}")
        return (seq.count('G') + seq.count('C')) / len(seq)
    except (TypeError, ValueError) as e:
        print(f"  Error: {e}")
        return None

print("Error handling:")
for seq in ["ATGCGC", 123, "ATGXYZ"]:
    result = safe_gc(seq)
    if result is not None: print(f"  '{seq}' GC = {result:.2f}")

# Generators — memory efficient
def kmer_generator(seq, k):
    """Yield k-mers one at a time (O(1) memory)."""
    for i in range(len(seq) - k + 1):
        yield seq[i:i+k]

dna = "ATGATGATGCGATCGATCG"
kmers = list(kmer_generator(dna, k=3))[:5]
print(f"\nFirst 5 3-mers of {dna[:12]}...: {kmers}")

# itertools — powerful combinatorics
from itertools import combinations, permutations, product, chain

bases = 'ACGT'
# All 2-base combinations
combos = list(combinations(bases, 2))
print(f"\nCombinations(ACGT, 2): {combos}")
# All 2-mers (with repetition)
dimers = [''.join(p) for p in product(bases, repeat=2)]
print(f"All 2-mers: {dimers}")
# Chaining sequences
seq1, seq2 = "ATG", "CGA"
combined = ''.join(chain(seq1, seq2))
print(f"chain('{seq1}', '{seq2}') = '{combined}'")
````

````python
# HackerRank Module 2 — Linked List & Binary Tree
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val; self.next = next

class LinkedList:
    def __init__(self): self.head = None

    def append(self, val):
        node = ListNode(val)
        if not self.head: self.head = node; return
        cur = self.head
        while cur.next: cur = cur.next
        cur.next = node

    def to_list(self):
        result, cur = [], self.head
        while cur: result.append(cur.val); cur = cur.next
        return result

    def reverse(self):
        prev, cur = None, self.head
        while cur: nxt = cur.next; cur.next = prev; prev = cur; cur = nxt
        self.head = prev

ll = LinkedList()
for x in [1, 2, 3, 4, 5]: ll.append(x)
print(f"Linked list: {ll.to_list()}")
ll.reverse()
print(f"Reversed: {ll.to_list()}")

# Binary search tree
class TreeNode:
    def __init__(self, val): self.val = val; self.left = self.right = None

def insert(root, val):
    if not root: return TreeNode(val)
    if val < root.val: root.left = insert(root.left, val)
    else: root.right = insert(root.right, val)
    return root

def inorder(root):
    if not root: return []
    return inorder(root.left) + [root.val] + inorder(root.right)

root = None
for v in [5, 3, 7, 1, 4, 6, 8]: root = insert(root, v)
print(f"\nBST inorder (sorted): {inorder(root)}")
````

def dna_health_naive(genes, health, queries):
    """
    O(q * |genes| * |dna|) — passes partial test cases.
    For full score: use Aho-Corasick multi-pattern matching O(q * |dna| + total_gene_len)
    """
    results = []

        for i in range(first, last + 1):

def is_balanced(s: str) -> str:
    pairs = {')': '(', ']': '[', '}': '{'}
    stack = []
    for c in s:
        if c in '([{':
            stack.append(c)
        elif c in ')]}':
            if not stack or stack[-1] != pairs[c]:
                return 'NO'
            stack.pop()
    return 'YES' if not stack else 'NO'

test_cases = [
    ('{[()]}', 'YES'),
    ('{[(])}', 'NO'),
    ('{{[[(())]]}}', 'YES'),
    ('(()', 'NO'),
]
for expr, expected in test_cases:

    status = '✓' if result == expected else '✗'

## Problem 3 — Dijkstra Shortest Reach (HackerRank Hard)
**HackerRank**: [hackerrank.com/challenges/dijkstrashortreach](https://www.hackerrank.com/challenges/dijkstrashortreach)
**Skills**: heapq, Dijkstra — real use in Google Maps, network routing

```python
# Find shortest distances from node s to all other nodes
# If unreachable, print -1
import heapq

def dijkstra_reach(n: int, edges: list, s: int) -> list[int]:
    graph = {i: [] for i in range(1, n+1)}
    for u, v, w in edges:
        graph[u].append((v, w))
        graph[v].append((u, w))
    
    dist = {i: float('inf') for i in range(1, n+1)}
    dist[s] = 0
    pq = [(0, s)]
    
    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v, w in graph[u]:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))
    
    return [-1 if dist[i] == float('inf') else dist[i]
            for i in range(1, n+1) if i != s]

# Test
edges = [(1,2,24),(1,3,3),(2,4,7),(3,4,5),(3,5,2),(4,5,10)]
print(dijkstra_reach(5, edges, 1))  # [24, 3, 8, 5]
```
