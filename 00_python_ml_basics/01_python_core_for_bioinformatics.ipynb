## 1. String Manipulation (DNA Sequences)

````python
# Reverse complement of a DNA sequence
complement = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}

def reverse_complement(seq):
    return ''.join(complement[b] for b in reversed(seq.upper()))

dna = "ATGCGATACGTTAG"
rc = reverse_complement(dna)
print(f"Original:  5'-{dna}-3'")
print(f"Rev comp:  5'-{rc}-3'")
assert reverse_complement(rc) == dna, "RC of RC should equal original"
print("✓ reverse_complement(reverse_complement(dna)) == dna")

# Biological meaning: complementary strand runs antiparallel
pairs = list(zip(dna, rc[::-1]))
print(f"Base pairs (first 5): {pairs[:5]}")
````

````python
# k-mer frequency analysis
from collections import Counter

def kmer_frequency(seq, k=3):
    """Count all k-mers in sequence."""
    kmers = [seq[i:i+k] for i in range(len(seq) - k + 1)]
    return Counter(kmers)

dna = "ATGATGATGATGCCC"
# 3-mer frequencies
freq_3 = kmer_frequency(dna, k=3)
print("3-mer frequencies:")
for kmer, count in sorted(freq_3.items(), key=lambda x: -x[1])[:6]:
    print(f"  {kmer}: {count}")

# 4-mers (unique substrings)
freq_4 = kmer_frequency(dna, k=4)
print(f"\nTotal unique 3-mers: {len(freq_3)}")
print(f"Total unique 4-mers: {len(freq_4)}")

# Biological application: k-mer composition is a sequence signature
total_possible = 4**3
print(f"4^3 = {total_possible} possible 3-mers; observed {len(freq_3)}")
````

## 6. Generators & Iterators for Large Sequence Files

````python
# Generators for memory-efficient sequence processing
def fasta_stream(text):
    """Generator: yield (header, sequence) one at a time.
    Key: O(1) memory — never holds all sequences simultaneously."""
    header, parts = None, []
    for line in text.strip().split('\n'):
        line = line.strip()
        if line.startswith('>'):
            if header is not None:
                yield header, ''.join(parts)
            header = line[1:]
            parts = []
        elif line:
            parts.append(line)
    if header is not None:
        yield header, ''.join(parts)

FASTA = """
>chr1_100k
ATGCGATCGATCGATCGTAGCTAGCTAGCTAGCTATGCATGCATGCATGC
GCGCGCGCGCGCGCGCATCGATCGATCGATCGATCGATCGATCGATCGAT
>chr1_200k
TTTTAAAACCCCGGGGTTTTAAAACCCCGGGGTTTTAAAACCCCGGGGTT
>chr2_500k
ATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCG"""

# Memory-efficient processing
stats = {'total_bases': 0, 'sequences': 0, 'gc_rich': 0}
for header, seq in fasta_stream(FASTA):
    stats['sequences'] += 1
    stats['total_bases'] += len(seq)
    gc = (seq.count('G') + seq.count('C')) / len(seq)
    if gc > 0.5:
        stats['gc_rich'] += 1
    print(f"Processed: {header[:30]} ({len(seq)} bp, GC={gc:.2f})")

print(f"\nSummary: {stats['sequences']} seqs, {stats['total_bases']} bp, {stats['gc_rich']} GC-rich")
print("Generator pattern: no list comprehension — O(1) memory per sequence")
````

## HackerRank-Style Practice Problems

### Problem 1 — DNA Health
Given genes with health values and a DNA string, find the sum of health values of all gene substrings found in the DNA.


````python
# HackerRank-style problem: DNA Health
def dna_health_naive(genes, health, dna):
    """Naive O(|dna| * |genes|) solution — correct but slow."""
    total = 0
    for i, gene in enumerate(genes):
        start = 0
        while True:
            pos = dna.find(gene, start)
            if pos == -1: break
            total += health[i]
            start = pos + 1
    return total

genes  = ["a", "b", "c", "aa", "d", "b"]
health = [1, 2, 3, 4, 5, 6]

test_cases = [
    (1, 5, "caaab", 19),
    (0, 2, "abc", 6),
    (0, 5, "bcd", 8),
]

print("DNA Health problem:")
for first, last, dna, expected in test_cases:
    g = genes[first:last+1]
    h = health[first:last+1]
    result = dna_health_naive(g, h, dna)
    status = "✓" if result == expected else "✗"
    print(f"  {status} genes[{first}:{last+1}], dna='{dna}' → {result} (expected {expected})")
````

### Problem 2 — GC Content Ranking

````python
# GC Content Ranking (Rosalind GC problem)
def gc_content(seq):
    gc = seq.count('G') + seq.count('C')
    return gc / len(seq) if seq else 0.0

sequences = {
    "Rosalind_6404": "CCTGCGGAAGATCGGCACTAGAATAGCCAGAACCGTTTCTCTGAGGCTTCCGGCCTTCCCTCCCACTAATAATTCTGAGG",
    "Rosalind_5959": "CCATCGGTAGCGCATCCTTAGTCCAATTAAGTCCCTATCCAGGCGCTCCGCCGAAGGTCTATATCCATTTGTCAGCAGACACGC",
    "Rosalind_0808": "CCACCCTCGTGGTATGGCTAGGCATTCAGGAACCGGAGAACGCTTCAGACCAGCCCGGACTGGGAACCTGCGGGCAGTAGGTGGAAT",
}

results = {name: gc_content(seq) for name, seq in sequences.items()}
best = max(results, key=results.get)

print("GC content per sequence:")
for name, gc in sorted(results.items(), key=lambda x: -x[1]):
    marker = " ← winner" if name == best else ""
    print(f"  {name}: {gc*100:.4f}%{marker}")

print(f"\nAnswer:")
print(best)
print(f"{results[best]*100:.6f}")
````

### Time Complexity Quick Reference — Python Operations
| Operation | Code | Time | Space |
|-----------|------|------|-------|
| String slice | `s[i:j]` | O(j-i) | O(j-i) |
| String find | `s.find(p)` | O(n·m) naive | O(1) |
| KMP string search | manual | O(n+m) | O(m) |
| List append | `lst.append(x)` | O(1) amortized | — |
| Dict lookup | `d[key]` | O(1) average | — |
| Sorting | `sorted(lst)` | O(n log n) | O(n) |
| Set membership | `x in s` | O(1) | — |
| Counter | `Counter(seq)` | O(n) | O(k) |
| Reverse complement | loop | O(n) | O(n) |
| GC content | sum/len | O(n) | O(1) |

````python
# OOP: Sequence and Structure classes
from dataclasses import dataclass, field
from typing import List, Dict

@dataclass
class DNASequence:
    name: str
    sequence: str

    def __post_init__(self):
        self.sequence = self.sequence.upper()

    def gc_content(self) -> float:
        return (self.sequence.count('G') + self.sequence.count('C')) / len(self.sequence)

    def reverse_complement(self) -> 'DNASequence':
        comp = str.maketrans('ATGCN', 'TACGN')
        rc = self.sequence.translate(comp)[::-1]
        return DNASequence(f"{self.name}_rc", rc)

    def transcribe(self) -> str:
        return self.sequence.replace('T', 'U')

    def translate(self, codon_table: Dict[str, str]) -> str:
        rna = self.transcribe()
        return ''.join(codon_table.get(rna[i:i+3], '?')
                       for i in range(0, len(rna)-2, 3)
                       if codon_table.get(rna[i:i+3], '*') != '*')

    def __len__(self):
        return len(self.sequence)

    def __repr__(self):
        return f"DNASequence(name='{self.name}', len={len(self)}, gc={self.gc_content():.2f})"

seq = DNASequence("EGFR_exon1", "ATGCGATCGATCGTAGCTAGCTAGCTATGCGATCG")
print(repr(seq))
print(f"RC: {seq.reverse_complement().sequence[:15]}...")
print(f"mRNA: {seq.transcribe()[:15]}...")
````

## Problem 4 — k-mer Frequency Fingerprinting for Sequence Classification
**Source**: Metagenomics (gut microbiome classification)
**Skills**: Counter, k-mers, cosine similarity, defaultdict

**Task**: Represent each sequence as a 4-mer frequency vector, then compute cosine similarity between pairs. Cluster 10 bacterial genomes into species groups.

```python
import numpy as np
from itertools import product

def kmer_frequency_vector(seq: str, k: int = 4) -> np.ndarray:
    """
    Return a normalized 4^k frequency vector over all k-mers.
    This is used as a species fingerprint in metagenomics.
    """
    all_kmers = [''.join(p) for p in product('ATGC', repeat=k)]
    kmer_to_idx = {km: i for i, km in enumerate(all_kmers)}
    vec = np.zeros(len(all_kmers))
    for km in kmers(seq, k):
        if km in kmer_to_idx:
            vec[kmer_to_idx[km]] += 1
    total = vec.sum()
    return vec / total if total > 0 else vec

def cosine_sim(v1: np.ndarray, v2: np.ndarray) -> float:
    denom = np.linalg.norm(v1) * np.linalg.norm(v2)
    return float(np.dot(v1, v2) / denom) if denom > 0 else 0.0

# Compare 4-mer profiles of sequences
seqs = {'E.coli_K12': 'ATGCGATACGCTTGCATGC' * 50,
        'E.coli_O157': 'ATGCGATACGCTTGCATGT' * 50,
        'Salmonella': 'TTGCGATACGCTTGCATGC' * 50}

profiles = {name: kmer_frequency_vector(seq) for name, seq in seqs.items()}
names = list(profiles.keys())
for i in range(len(names)):
    for j in range(i+1, len(names)):
        sim = cosine_sim(profiles[names[i]], profiles[names[j]])
        print(f"{names[i]} vs {names[j]}: cosine={sim:.4f}")
```

**Real impact**: Core of Kraken2, Mash — tools that classify billions of reads from microbiome samples.

## ⏱️ Algorithmic Complexity — See It Empirically

Asymptotic complexity is only meaningful when you *see the curves*. The next cell times several operations on increasing input sizes and plots them.

````python
import time
import numpy as np
import matplotlib.pyplot as plt

def time_it(fn, args, repeats=3):
    times = []
    for _ in range(repeats):
        t0 = time.perf_counter()
        fn(*args)
        times.append(time.perf_counter() - t0)
    return min(times)

# O(n) — linear scan
def linear_scan(seq):
    return sum(1 for c in seq if c == 'G')

# O(n^2) — naive all-pairs k-mer comparison (toy example)
def quadratic_pairwise(seq):
    count = 0
    kmers = [seq[i:i+5] for i in range(len(seq)-4)]
    for i in range(len(kmers)):
        for j in range(i+1, len(kmers)):
            if kmers[i] == kmers[j]:
                count += 1
    return count

# O(n log n) — sort-based unique counting
def nlogn_unique(seq):
    kmers = sorted(seq[i:i+5] for i in range(len(seq)-4))
    return len(set(kmers))

import random, string
sizes = [100, 300, 500, 1000, 2000]
results = {'O(n) linear': [], 'O(n²) quadratic': [], 'O(n log n) sort': []}

print("Timing on increasing sequence lengths...")
for n in sizes:
    seq = ''.join(random.choices('ACGT', k=n))
    results['O(n) linear'].append(time_it(linear_scan, [seq]) * 1000)
    results['O(n²) quadratic'].append(time_it(quadratic_pairwise, [seq]) * 1000)
    results['O(n log n) sort'].append(time_it(nlogn_unique, [seq]) * 1000)
    print(f"  n={n:5d}: O(n)={results["O(n) linear"][-1]:.3f}ms  "
          f"O(n²)={results["O(n²) quadratic"][-1]:.3f}ms  "
          f"O(nlogn)={results["O(n log n) sort"][-1]:.3f}ms")

fig, ax = plt.subplots(figsize=(9, 5))
for label, times in results.items():
    ax.plot(sizes, times, 'o-', label=label, linewidth=2)
ax.set_xlabel('Sequence length (n)')
ax.set_ylabel('Time (ms)')
ax.set_title('Empirical Complexity — See the Curves')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('complexity_curves.png', dpi=100)
plt.show()
print("\nNotice: O(n²) bends sharply upward — this is why quadratic algorithms fail at scale.")
````
