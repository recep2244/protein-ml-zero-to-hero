## 🧬 Reading Scientific Papers — Concepts for Beginners

| Term | Plain English |
|------|--------------|
| **Abstract** | A 150-300 word summary of the entire paper: problem, method, key result, conclusion |
| **Introduction** | Sets up the problem, reviews prior work, and states the paper's contribution |
| **Methods** | Detailed description of how the experiments were done (reproducibility section) |
| **Results** | Presents the data, figures, and statistical analyses — the evidence |
| **Discussion** | Interprets the results, acknowledges limitations, and suggests future work |
| **Figure** | A visual representation of data or concepts — often the most information-dense part |
| **Citation** | A reference to another paper; follow citations to understand the lineage of ideas |
| **DOI** | Digital Object Identifier — a permanent, unique URL for any published paper |
| **Preprint** | A paper posted before peer review (bioRxiv, arXiv) — latest results, but not yet validated |
| **Peer review** | The process where other scientists evaluate a paper before journal publication |
| **Supplementary material** | Extra figures, tables, methods, and data that don't fit in the main paper |
| **Related work** | Section comparing the paper's approach to existing methods in the field |

## PART 1 — Why Reading Papers Matters

### The Gap Between Textbooks and Research

Textbooks describe methods that were invented 5–20 years ago and have been refined by thousands of instructors. They are excellent for building foundations. But the methods used at Isomorphic Labs, DeepMind, and EMBL-EBI today are not in any textbook — they are in papers published in the last 12–36 months.

**The structural ML field moves especially fast.** Consider the timeline:
- 2020: AlphaFold2 (Jumper et al., Nature)
- 2022: ESM-2 (Lin et al., Science); RFdiffusion (Watson et al., Nature)
- 2023: ProteinMPNN, Chai-1 preprints
- 2024: AlphaFold3 (Abramson et al., Nature); ESMFold, AlphaProteo

None of this is in a textbook. If you cannot read papers, you are permanently 3–5 years behind the field.

### Why Papers Feel Hard

Papers are hard for two reasons that have nothing to do with your intelligence:

1. **Dense notation**: authors compress months of work into 8 pages. Every symbol carries meaning built over multiple paragraphs.
2. **Missing context**: papers assume you have read 10–20 cited papers. You have not. That is normal.

The solution is **structured reading with clear goals per pass** — not reading harder or longer.

## PART 2 — The Three-Pass Method (Keshav 2007)

This method is used by experienced researchers worldwide. It prevents you from wasting hours on a paper that is irrelevant to your goal.

### Pass 1: The Bird's Eye View (5–10 minutes)

**Goal:** Decide whether this paper is worth reading further.

**Read only:**
1. Title and authors
2. Abstract (every sentence)
3. Introduction (first and last paragraph only)
4. All section headings
5. Conclusions section
6. Skim figures and captions (do not read figure text yet)

**After Pass 1 you should know:**
- What problem does the paper solve?
- What is the main claim?
- Is this paper relevant to you?
- Category: measurement, analysis, description of a new system, or research?

---

### Pass 2: The Careful Read (1 hour)

**Goal:** Understand the main content. You may skip proofs and dense math.

**Read carefully:**
- Introduction in full
- Methods (architecture, training data, evaluation)
- Main results figures and tables (read captions fully)
- Discussion / Conclusion in full

**Tools:** Annotate as you read. Circle undefined terms. Mark key claims with C, key evidence with E.

**Stop point:** If you cannot understand a section after 10 minutes, mark it and move on. You do not need 100% comprehension in Pass 2.

---

### Pass 3: The Deep Dive (4–6 hours, over multiple sessions)

**Goal:** Understand the paper as well as the authors do. Re-implement a key result.

**Do:**
- Work through every equation by hand
- Identify every assumption and ask whether it is valid
- Trace every claim back to a figure or table
- Re-implement the key algorithmic contribution
- Read 2–3 cited papers for context

**You reach Pass 3 mastery when** you could give a 20-minute technical talk on the paper.

## PART 4 — Pass 2 Walkthrough: Methods Section Structure

### What to Look For in Any ML Methods Section

When you read the Methods of a protein ML paper, answer these 8 questions in order:

1. **Architecture:** What is the model topology? (transformer, GNN, diffusion model, hybrid?) Draw a box diagram from the description alone.

2. **Inputs:** What does the model take as input? (sequence, MSA, template, atom coordinates, graph?) What is the tokenization?

3. **Outputs:** What does the model produce? (coordinates, distance matrix, logits?) Is output continuous or discrete?

4. **Training data:** What database? What split? How many samples? How was redundancy handled (sequence identity cutoff)?

5. **Loss function:** What is being optimized? Is it differentiable end-to-end? Are there auxiliary losses?

6. **Evaluation protocol:** What benchmark? Who created it? Is the test set independent from training? What metric?

7. **Baselines:** What are they being compared to? Are comparisons fair (same inputs, same compute budget)?

8. **Compute:** How many GPUs, how long to train? Can you reproduce this without a supercomputer?

---

### Applying These Questions to AlphaFold3

After Pass 2 of Abramson et al. 2024, you should be able to fill this table:

| Question | AF3 Answer |
|---|---|
| Architecture | Pairformer (replaces Evoformer) + diffusion module for coordinate generation |
| Inputs | Tokenised atoms (residues + ligand atoms), MSA (optional), templates |
| Outputs | 3D coordinates for all heavy atoms via diffusion denoising |
| Training data | PDB + Cambridge Structural Database (CSD) for small molecules |
| Loss function | Diffusion loss (denoising) + FAPE auxiliary + confidence head |
| Evaluation | PoseBusters v2 (ligand), CASP15-derived protein set, antibody benchmark |
| Baselines | AF-Multimer v2.3 (protein-protein), AutoDock-Vina (docking), RoseTTAFold2NA |
| Compute | TPU v4 pods, 8 × 128-core pods — not reproducible without Google-scale hardware |

**Critical observation:** The compute requirement (TPU v4 pods) means you cannot fully reproduce AF3 training. This is normal for flagship models. But you CAN reproduce the architecture and run inference using the released server or OpenFold3 when available.

### Specific Questions to Ask of Any Protein ML Paper

These questions expose weaknesses that peer review often misses:

**Data leakage:**
- What is the sequence identity cutoff between train and test? (30% is standard; 70% is suspicious)
- Were structures solved after the training cutoff date excluded from evaluation?
- For AF3: the PDB training cutoff is September 2021; CASP15 targets were solved after this — good.

**Benchmark fairness:**
- Did the authors create the benchmark themselves? If so, were they blinded during benchmark creation?
- Is the benchmark publicly available for others to use?
- For AF3: PoseBusters is a third-party benchmark (Buttenschoen et al. 2024), which strengthens the claim.

**Ablation completeness:**
- Are ablation experiments shown? (remove one component, measure the drop in performance)
- For AF3: Supplementary Figure S12 shows ablations of the diffusion module components.

**Failure mode documentation:**
- Does the paper discuss when the model fails?
- For AF3: Section 4.5 discusses limitations including large flexible complexes and heavily modified residues.

## PART 5 — Pass 3 Walkthrough: Re-implement a Result

The best evidence that you understand a paper is that you can re-implement a key component from the methods description alone, without looking at code.

For AlphaFold3, we will re-implement the **simplified Pairformer attention step** — the triangle attention mechanism that updates the pair representation. This was introduced in AF2's Evoformer and refined in AF3.

The key equation from the paper (Algorithm 11 in the AF2 supplement, carried into AF3):

```
b_ij = Linear(z_ij)                           # bias from pair representation
a_ij = softmax_j(1/sqrt(c) * q_i · k_j + b_ij) * v_j   # attended value
z_ij += Linear(a_ij)                           # update pair representation
```

This is standard self-attention with a pair-representation bias — a crucial difference from vanilla transformers.

````python
import numpy as np

# Simplified Pairformer attention forward pass
# Re-implemented from description in Abramson et al. 2024 Methods + AF2 supplement
# This is NOT production code — it is a learning tool

np.random.seed(42)

def scaled_dot_product_attention(Q, K, V, bias=None):
    """
    Standard scaled dot-product attention with optional pair bias.

    Q, K, V: shape (seq_len, head_dim)
    bias:    shape (seq_len, seq_len) — comes from pair representation z_ij

    Paper reference: AF2 Supplement Algorithm 1 (Attention with pair bias)
    """
    seq_len, head_dim = Q.shape
    scale = np.sqrt(head_dim)

    # Compute attention logits: (seq_len, seq_len)
    logits = Q @ K.T / scale  # shape: (L, L)

    # Add pair representation bias (the key AF2/AF3 innovation over vanilla attention)
    if bias is not None:
        logits = logits + bias

    # Softmax over key dimension (axis=-1)
    logits_shifted = logits - logits.max(axis=-1, keepdims=True)  # numerical stability
    attn_weights = np.exp(logits_shifted)
    attn_weights /= attn_weights.sum(axis=-1, keepdims=True)

    # Weighted sum of values
    output = attn_weights @ V  # shape: (L, head_dim)
    return output, attn_weights


def pairformer_attention_step(s, z, d_s=32, d_z=16, n_heads=4):
    """
    One step of Pairformer row-wise attention with pair bias.

    s: single representation  (L, d_s)   — per-residue features
    z: pair representation    (L, L, d_z) — per-pair features

    Returns updated s after attending with pair bias from z.

    Paper reference: AF3 Methods section; AF2 Supplement Algorithm 6
    """
    L = s.shape[0]
    head_dim = d_s // n_heads
    assert d_s % n_heads == 0, "d_s must be divisible by n_heads"

    # Linear projections (in practice: learned weights; here: random for shape demonstration)
    W_Q = np.random.randn(d_s, d_s) * 0.1
    W_K = np.random.randn(d_s, d_s) * 0.1
    W_V = np.random.randn(d_s, d_s) * 0.1
    W_bias = np.random.randn(d_z, n_heads) * 0.1  # projects pair dim to n_heads
    W_out = np.random.randn(d_s, d_s) * 0.1

    Q = s @ W_Q  # (L, d_s)
    K = s @ W_K  # (L, d_s)
    V = s @ W_V  # (L, d_s)

    # Pair bias: z has shape (L, L, d_z); project to (L, L, n_heads)
    pair_bias = z @ W_bias  # (L, L, n_heads)

    # Multi-head attention
    all_head_outputs = []
    for h in range(n_heads):
        Qh = Q[:, h*head_dim:(h+1)*head_dim]  # (L, head_dim)
        Kh = K[:, h*head_dim:(h+1)*head_dim]
        Vh = V[:, h*head_dim:(h+1)*head_dim]
        bias_h = pair_bias[:, :, h]            # (L, L)

        out_h, attn_h = scaled_dot_product_attention(Qh, Kh, Vh, bias=bias_h)
        all_head_outputs.append(out_h)

    # Concatenate heads and project
    concatenated = np.concatenate(all_head_outputs, axis=-1)  # (L, d_s)
    s_updated = s + concatenated @ W_out  # residual connection
    return s_updated


# Test with a small protein (L=8 residues)
L, d_s, d_z = 8, 32, 16
s = np.random.randn(L, d_s)    # single representation
z = np.random.randn(L, L, d_z) # pair representation

s_out = pairformer_attention_step(s, z, d_s=d_s, d_z=d_z, n_heads=4)

print("Pairformer attention step: shape verification")
print(f"  Input  s: {s.shape}  (L={L} residues, d_s={d_s})")
print(f"  Input  z: {z.shape}  (L x L pairs, d_z={d_z})")
print(f"  Output s: {s_out.shape}")
print()
print("Key AF3 innovation: the pair bias b_ij = Linear(z_ij) is added to attention")
print("logits BEFORE softmax. This lets the model use pairwise distance/orientation")
print("information to modulate which residues attend to each other.")
print()
# Verify the residual connection preserved scale
print(f"Input s mean abs: {np.abs(s).mean():.4f}")
print(f"Output s mean abs: {np.abs(s_out).mean():.4f}  (residual: scale should be similar)")
````

### What Pass 3 Confirms

By re-implementing the Pairformer attention step from the paper description, you have confirmed:

1. The pair representation `z_ij` is projected to per-head biases and added before softmax — this is different from standard transformer cross-attention
2. The residual connection `s = s + attention_output` is standard
3. The output shape matches the input shape — Pairformer is a sequence-to-sequence model over residues

For a full Pass 3, you would also implement the triangle update operations on `z` (triangle multiplicative update, triangle self-attention) — these are in Module 07/02.

## PART 6 — How to Critically Evaluate a Paper

### The 10 Questions

Ask these in order. A paper that fails more than 3 should be read skeptically.

| # | Question | What to look for |
|---|---|---|
| 1 | Is the baseline fair? | Same inputs, same compute, same training data? |
| 2 | Is the test set independent? | No overlap with training (check sequence identity cutoff) |
| 3 | Are error bars reported? | Mean ± std or confidence intervals on all key results |
| 4 | Is the code available? | GitHub link, not just "code available upon request" |
| 5 | Are negative results shown? | Ablations that did not work are as informative as those that did |
| 6 | Is the metric appropriate? | Does it measure what the claim says? (e.g., RMSD vs TM-score) |
| 7 | Is the benchmark third-party? | Self-designed benchmarks can be unconsciously optimised toward |
| 8 | Are failure cases shown? | Any honest paper discusses when the method fails |
| 9 | Is the sample size adequate? | 5 examples is not enough to claim a general improvement |
| 10 | Does compute match the claim? | A result requiring 1000 GPU-hours is not "practical" |

### Applying to AF3

| Question | AF3 Assessment |
|---|---|
| Fair baseline? | Mostly yes: AutoDock-Vina uses crystal structure input, AF3 does not — slight advantage to AF3 |
| Test set independent? | Yes: PDB cutoff Sept 2021; CASP15 structures solved 2022 |
| Error bars? | Partially: some figures show SE, others only point estimates |
| Code available? | Weights released 2024-11 (conditional license); server always free |
| Ablations? | Yes: Supplementary Fig S12, S13 show ablation results |
| Appropriate metric? | PoseBusters checks physical validity, not just RMSD — good choice |
| Third-party benchmark? | Yes: PoseBusters (Buttenschoen et al. 2024), CASP15 (external) |
| Failure cases? | Section 4.5 in supplement: large complexes, heavily modified residues |
| Sample size? | >700 targets in CASP15, >300 in PoseBusters — adequate |
| Compute honest? | TPU v4 pods disclosed; clearly not reproducible for most |

**Verdict:** AF3 is a high-quality paper. The main legitimate criticism is the initial non-open-source weight release and the slight input asymmetry in the docking comparison.

## PART 7 — How to Manage a Reading List

- Set alerts for new papers by author (e.g., alert on David Baker, Demis Hassabis)

## Interview Q&A — Reading Papers

**Q: Walk me through the AlphaFold3 paper at a high level.**

A: AF3 extends AF2 to all biomolecule types (proteins, DNA, RNA, small molecules, ions, modified residues) using a diffusion-based architecture. The key changes from AF2: (1) Evoformer is replaced by Pairformer, which operates on single + pair representations without the MSA stack; (2) the structure module is replaced by a diffusion module that generates all-atom coordinates by denoising; (3) training uses the PDB plus the Cambridge Structural Database for small molecules. The headline result is 50% improvement on the PoseBusters protein-ligand docking benchmark versus AF-Multimer v2.3, and +20pp on antibody-antigen prediction.

---

**Q: What are the limitations of AlphaFold3?**

A: Three main limitations stated in the paper: (1) large flexible complexes with intrinsic disorder are predicted less accurately; (2) heavily modified residues not well-represented in training data perform worse; (3) inference is non-deterministic (diffusion sampling), so multiple runs can give different results. Unstated limitation: training required TPU v4 pod-scale compute, so community replication of training from scratch is not feasible.

---

**Q: How would you critically evaluate a new protein structure prediction paper?**

A: The ten-question framework: (1) Is the baseline fair — same inputs, same compute? (2) Is the test set temporally or sequence-separated from training? (3) Are error bars reported? (4) Is code available? (5) Are ablations shown? (6) Is the metric appropriate — RMSD, TM-score, or GDT-TS? (7) Is the benchmark third-party? (8) Are failure cases documented? (9) Is the sample size sufficient? (10) Is the compute cost disclosed? For AF3 specifically: the PoseBusters benchmark is third-party (strong point), but the comparison to AutoDock-Vina is slightly unfair because AutoDock requires a crystal structure pocket while AF3 predicts de novo.

---

**Q: What is the difference between ESMFold and AlphaFold3?**

A: Both predict protein structure from sequence, but the paradigms differ. ESMFold uses a 15B-parameter protein language model (ESM-2) as an encoder, then adds a folding head — no multiple sequence alignment at inference time, which makes it 60x faster than AF2. The tradeoff is accuracy: for proteins with rich evolutionary information (large MSAs), AF2/AF3 are more accurate because MSA encodes co-evolutionary constraints. AF3 requires MSA (optional) plus templates and predicts all biomolecule types, not just proteins. Choose ESMFold for speed at scale (metagenomics); choose AF3 for highest accuracy on protein-ligand or multi-chain complexes.

## 🎤 Interview Questions

**Q1 (Easy):** What is the three-pass method for reading a scientific paper?
<details><summary>Answer</summary>
Pass 1 (5-10 min): Read title, abstract, introduction, section headings, conclusion — decide if the paper is relevant. Pass 2 (1 hour): Read the whole paper but skip proofs/details — understand figures, key equations, and the main argument. Pass 3 (4-5 hours): Read in detail, re-derive key results, mentally replicate the experiments, identify assumptions, and note weaknesses. Most papers only need Pass 1-2.
</details>

**Q2 (Easy):** Where should you look first to quickly assess if a paper is worth reading in depth?
<details><summary>Answer</summary>
Abstract (for the claim), Figures (for the evidence quality), and Conclusion (for the takeaway). If the figures are convincing and the abstract claims are supported, proceed to a deeper read. Also check: when was it published, how many citations, and what venue (top conference/journal vs. unknown). For ML papers, check if code is available.
</details>

**Q3 (Medium):** How do you critically evaluate the experimental methodology of an ML paper?
<details><summary>Answer</summary>
Check: (1) Is there a proper train/val/test split with no data leakage? (2) Are baselines fair and recent? (3) Are confidence intervals or error bars reported? (4) Is the evaluation metric appropriate for the task? (5) Is the dataset representative? (6) Are hyperparameters tuned on validation, not test? (7) Is there an ablation study showing which components matter? (8) Is the code available for reproduction? Red flags: no error bars, unfair baselines, tuning on test set, cherry-picked examples.
</details>

**Q4 (Medium):** A paper reports "state-of-the-art" results. What questions should you ask before believing this claim?
<details><summary>Answer</summary>
(1) State-of-the-art on which benchmark and which metric? (2) Were all baselines re-run with the same hardware and hyperparameter budget, or are numbers copied from original papers? (3) Is the improvement statistically significant or within noise? (4) Does the method use more data, more compute, or more parameters than baselines (unfair advantage)? (5) Was the benchmark saturated (small absolute improvement)? (6) Were there concurrent works not compared against? (7) Does the improvement hold across multiple datasets or just one?
</details>

**Q5 (Hard):** You're reviewing the AlphaFold 3 paper for a journal club. What are the key technical claims to verify, and what limitations should you highlight?
<details><summary>Answer</summary>
Claims to verify: (1) Joint protein-ligand-nucleic acid prediction accuracy (check DockQ scores, ligand RMSD distributions, not just averages). (2) Improvement over AF2 on protein-only targets (how much comes from diffusion vs. more training data?). (3) Generalization to novel folds/complexes not in training set. Limitations to highlight: (1) Not open-source initially — results can't be independently verified. (2) Performance on disordered regions and conformational ensembles. (3) Computational cost vs. AF2. (4) Training data temporal cutoff — does it generalise to truly novel complexes? (5) Ligand accuracy varies significantly by ligand type. (6) No explicit handling of protonation states or water molecules.
</details>
