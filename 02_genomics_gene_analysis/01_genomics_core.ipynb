## Interview Tip Bank
> **ORF uniqueness**: The Rosalind ORF problem asks for DISTINCT protein sequences. Use a set to deduplicate — nested ORFs can produce the same protein.

> **Frame vs strand**: Frame = starting position offset (0, 1, or 2). Strand = forward or reverse complement. 3 frames × 2 strands = 6 total reading frames.

> **Jukes-Cantor limit**: When p > 0.75 (more than 75% of sites differ), the formula breaks down — you can't distinguish signal from noise. This is the "saturation" problem.

> **Variant calling datastructure**: On HackerRank, variants are usually represented as (position, ref, alt) tuples. Always check if 0-indexed or 1-indexed.

## Genomics — Concepts for Beginners

| **Genome** | The complete DNA instruction set of an organism — all chromosomes combined |

## 2. ORF Finding (Open Reading Frames)

````python
def find_orfs(dna, min_len=100):
    """Find all ORFs in all 6 reading frames."""
    orfs = []
    dna = dna.upper()
    rc = dna.translate(str.maketrans('ACGT','TGCA'))[::-1]

    for seq, strand in [(dna, '+'), (rc, '-')]:
        for frame in range(3):
            i = frame
            while i < len(seq) - 2:
                codon = seq[i:i+3]
                if codon == 'ATG':
                    # found start
                    j = i + 3
                    while j < len(seq) - 2:
                        stop = seq[j:j+3]
                        if stop in ('TAA','TAG','TGA'):
                            orf_len = j + 3 - i
                            if orf_len >= min_len:
                                orfs.append((strand, frame, i, j+3, seq[i:j+3]))
                            break
                        j += 3
                i += 3
    return orfs

# Example
seq = "ATGAAACCCGGGTTTTAA" * 3
orfs = find_orfs(seq, min_len=15)
for strand, frame, start, end, orf in orfs:
    print(f"ORF {strand} frame {frame}: {start}-{end}, len={len(orf)}, prot={orf}")
````

## Interview Q&A — Genomics & Gene Analysis

**Q: What is the central dogma of molecular biology and its exceptions?**
A: DNA → (transcription) → mRNA → (translation) → Protein. Exceptions: retroviruses (RNA → DNA via reverse transcriptase), RNA viruses (RNA → RNA), some non-coding RNAs that never get translated. For ML: each arrow is a prediction problem — splice site prediction, codon optimization, signal peptide detection.

**Q: What is variant calling and what makes it difficult?**
A: Identifying differences between a sequenced genome and a reference. Difficulties: sequencing errors (1 in 1000 bases), PCR amplification bias, mapping ambiguity (repetitive regions), germline vs somatic variants. Tools: GATK (gold standard), DeepVariant (CNN-based, Google). For ML: distinguish true variants from sequencing artifacts.

**Q: Explain FPKM/RPKM/TPM normalization for RNA-seq.**
A: RPKM: Reads Per Kilobase per Million mapped reads. Normalizes for gene length (longer genes get more reads) and sequencing depth. FPKM: same but for paired-end (Fragments). TPM: Transcripts Per Million — divide by gene length FIRST, then normalize by sum. TPM is preferred: samples are truly comparable. RPKM/FPKM are not directly comparable across samples.

**Q: What is a phylogenetic tree and what does bootstrap support mean?**
A: Phylogenetic tree: branching diagram showing evolutionary relationships. Internal nodes = common ancestors. Bootstrap support: resample columns of the alignment 1000 times, rebuild the tree, count what fraction of trees contain each branch. >70% = moderately supported, >95% = well-supported.

### Genomics — Time Complexity
| Operation | Time | Notes |
|-----------|------|-------|
| ORF finding | O(n) | scan all 6 reading frames |
| Codon frequency table | O(n) | n = sequence length |
| Pairwise distance matrix | O(k² × n) | k sequences, n sites |
| UPGMA clustering | O(k²) per merge, O(k³) total | k sequences |
| Neighbor-joining | O(k³) | k sequences |
| Multiple sequence alignment | NP-hard exact, O(k² × n) heuristic | Progressive alignment |

````python
def call_variants(ref, query, min_qual=20):
    """Simple pairwise variant calling."""
    variants = []
    assert len(ref) == len(query), "Sequences must be same length (aligned)"
    for i, (r, q) in enumerate(zip(ref, query)):
        if r != q and r != '-' and q != '-':
            variants.append({
                'pos': i+1,
                'ref': r,
                'alt': q,
                'type': 'SNV'
            })
        elif r == '-' and q != '-':
            variants.append({'pos': i+1, 'ref': '-', 'alt': q, 'type': 'INS'})
        elif r != '-' and q == '-':
            variants.append({'pos': i+1, 'ref': r, 'alt': '-', 'type': 'DEL'})
    return variants

ref   = "ATGCGTACGTTAGC"
query = "ATGCCTACGTTGGC"
variants = call_variants(ref, query)
print(f"Found {len(variants)} variants:")
for v in variants:
    print(f"  pos {v['pos']}: {v['ref']}->{v['alt']} ({v['type']})")
````

def call_variants(reference: str, observed: str) -> list[tuple]:
    """Simple SNP calling: find all positions where sequences differ."""
    assert len(reference) == len(observed), "Must be same length (pre-aligned)"
    variants = []
    for i, (r, o) in enumerate(zip(reference, observed)):
        if r != o:
            variants.append((i+1, r, o))  # 1-indexed position
    return variants

# Generate Omicron sequence
omicron_seq = apply_variants(WUHAN_SPIKE_FRAGMENT,
                              {k: v for k, v in OMICRON_MUTATIONS.items()
                               if k < len(WUHAN_SPIKE_FRAGMENT)})

# Call variants
snps = call_variants(WUHAN_SPIKE_FRAGMENT[:100], omicron_seq[:100])
print(f"Variants detected: {len(snps)}")
for pos, ref, alt in snps:
    print(f"  Position {pos}: {ref} → {alt}")

## 🐛 Debug Exercise — Genomics Core Algorithm Bugs

Find and fix **3 bugs** in this ORF finder and codon usage calculator.

<details><summary>Show bugs</summary>

**Bug 1:** ORF search only checks reading frame +1 (i=0 step 3). Must check all 3 forward frames (i=0,1,2) and 3 reverse complement frames.

**Bug 2:** Stop codon check uses `if codon == "TAA" or "TAG" or "TGA"` — in Python, `"TAG"` is always truthy. Must use `if codon in {"TAA","TAG","TGA"}`.

**Bug 3:** Codon frequency normalized by `len(sequence)/3` (total possible codons including non-coding) instead of by the actual number of codons in coding regions.

</details>

````python
def find_orfs_buggy(seq):
    stop_codons = {"TAA", "TAG", "TGA"}
    orfs = []
    # BUG 1: only checks reading frame 0, misses frames 1, 2, and reverse strand
    for i in range(0, len(seq)-2, 3):  # should try i in [0,1,2] separately
        if seq[i:i+3] == "ATG":
            for j in range(i+3, len(seq)-2, 3):
                codon = seq[j:j+3]
                # BUG 2: wrong stop codon check (always True due to "TAG" being truthy)
                if codon == "TAA" or "TAG" or "TGA":  # should be: if codon in stop_codons
                    orfs.append(seq[i:j+3])
                    break
    return orfs

def codon_frequency_buggy(orfs):
    from collections import Counter
    all_codons = []
    total_seq_len = sum(len(o) for o in orfs)
    for orf in orfs:
        for i in range(0, len(orf)-2, 3):
            all_codons.append(orf[i:i+3])
    counts = Counter(all_codons)
    # BUG 3: wrong denominator
    return {k: v / (total_seq_len / 3) for k,v in counts.items()}  # should be len(all_codons)

seq = "ATGATCTAGATGAAATAA"
orfs = find_orfs_buggy(seq)
print(f"ORFs found (buggy): {orfs}")
print(f"Expected: ['ATGATCTAG', 'ATGAAATAA']")
print(f"Bug 2 makes every codon look like a stop codon -> ORFs are too short")

freq = codon_frequency_buggy(orfs)
print(f"\nCodon freq sum (buggy):  {sum(freq.values()):.3f}  (should be 1.0 for relative freq)")
````
