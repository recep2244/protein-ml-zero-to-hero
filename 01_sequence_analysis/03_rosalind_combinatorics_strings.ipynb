## Algorithm Pattern Map

| Problem | Algorithm | Key Insight |
|---------|-----------|-------------|
| CORR | Frequency counting + Hamming | Rev. complement = same read |
| RNAS | Interval DP O(n³) | Count perfect matchings |
| TRIE | Prefix tree construction | O(total_chars) build |
| LREP | Binary search + hashing | O(n log n) |
| LING | Suffix set counting | distinct / theoretical max |
| REAR | BFS on permutation space | State = permutation |
| SORT | Greedy reversal | Always fix position i |
| MOTZ | DP with sum of products | M[n] = M[n-1] + products |
| SCSP | LCS-based reconstruction | Length = m + n - LCS |
| ITWV | DP on grid | Can two motifs interleave? |

## Interview Tip Bank
> **TRIE vs KMP**: Use KMP for one pattern in one text. Use TRIE (or Aho-Corasick) for many patterns in one text. TRIE build is O(total pattern length), search is O(text + matches).

> **Suffix array trick**: Sort all suffixes lexicographically. The LCP array gives max length shared between adjacent suffixes. Max LCP = length of longest repeated substring.

> **REAR BFS gotcha**: The state space is n! for permutations of length n. For small n (≤ 8) this is feasible. For larger n, you need more sophisticated algorithms (breakpoint graph).

> **Motzkin vs Catalan**: Catalan(n) counts ways where EVERY position is paired. Motzkin(n) allows unpaired positions too. RNA uses Motzkin.

## CORR — Error Correction in Reads

````python
from collections import Counter
def error_correction(reads):
    complement = str.maketrans('ACGT', 'TGCA')
    # Find all k-mers
    all_reads = set()
    for r in reads:
        all_reads.add(r)
        all_reads.add(r.translate(complement)[::-1])
    correct = set()
    for r in all_reads:
        if reads.count(r) > 1 or reads.count(r.translate(complement)[::-1]) > 1:
            correct.add(r)
    corrections = []
    for r in reads:
        rc = r.translate(complement)[::-1]
        if r in correct or rc in correct:
            pass  # correct read
        else:
            # Find correction with Hamming distance 1
            for cand in correct:
                if len(cand) == len(r):
                    diffs = sum(a!=b for a,b in zip(r,cand))
                    if diffs == 1:
                        corrections.append((r, cand)); break
    return corrections
reads = ["TCATC","CCATG","CAATG","TAATG","CAATG"]
for orig, corr in error_correction(reads): print(f"{orig}->{corr}")
````

## REAR — Reversal Distance

````python
from itertools import permutations
def signed_perms(n):
    for p in permutations(range(1,n+1)):
        for signs in range(2**n):
            yield tuple(s*((-1)**((signs>>i)&1)) for i,s in enumerate(p))
n = 2
perms = list(signed_perms(n))
print(len(perms))
for p in perms[:8]: print(*p)
````

## LREP — Finding the Longest Multiple Repeat

````python
def greedy_assemble(reads):
    reads = list(set(reads))  # remove duplicates
    while len(reads) > 1:
        best = (-1, -1, -1)
        for i in range(len(reads)):
            for j in range(len(reads)):
                if i != j:
                    for ov in range(min(len(reads[i]), len(reads[j])), 0, -1):
                        if reads[i][-ov:] == reads[j][:ov]:
                            if ov > best[0]: best = (ov, i, j); break
        ov, i, j = best
        if ov == -1: break
        merged = reads[i] + reads[j][ov:]
        reads = [reads[k] for k in range(len(reads)) if k not in (i,j)] + [merged]
    return reads[0]
reads = ["ATTAGACCTG", "CCTGCCGGAA", "AGACCTGCCG", "GCCGGAATAC"]
print(greedy_assemble(reads))
````

## NWCK — Distances in Trees (Newick format)

````python
def parse_newick(s):
    """Parse simple Newick string to nested dict."""
    s = s.strip().rstrip(';')
    def parse(s, start=0):
        if s[start] != '(':
            end = start
            while end < len(s) and s[end] not in ',)': end += 1
            return s[start:end], end
        end = start + 1; depth = 1; children = []
        while depth > 0:
            if s[end] == '(': depth += 1
            elif s[end] == ')': depth -= 1
            elif s[end] == ',' and depth == 1:
                children.append(s[start+1:end])
                start = end
            end += 1
        return children, end
    return parse(s)

newick = "((rabbit,cat),(dog,wolf))"
result, _ = parse_newick(newick)
print(f"Newick: {newick}")
print(f"Parsed: {result}")
print("Application: represent evolutionary trees from distance matrices")
````

```python
import random
from collections import Counter

def generate_reads(genome, n_reads=200, read_len=50, error_rate=0.001):
    bases = 'ACGT'
    reads = []
    for _ in range(n_reads):
        start = random.randint(0, len(genome) - read_len)
        read = list(genome[start:start + read_len])
        for i in range(len(read)):
            if random.random() < error_rate:
                read[i] = random.choice([b for b in bases if b != read[i]])
        reads.append(''.join(read))

def hamming(s, t):
    return sum(a != b for a, b in zip(s, t))

def error_correct_reads(reads):
    counts = Counter(reads)
    trusted = {r for r, c in counts.items() if c >= 3}
    corrected = []
    for read in reads:
        if read in trusted:
            corrected.append(read)
        else:
            best = next((t for t in trusted if len(t) == len(read) and hamming(read, t) == 1), read)
            corrected.append(best)
    return corrected
```

```python
def reversal_distance(perm1, perm2):
    perm = list(perm1)
    target = list(perm2)
    dist = 0
    for i in range(len(perm)):
        val = target[i]
        if perm[i] == val:
            continue
        for j in range(i+1, len(perm)):
            if perm[j] == val:
                perm[i:j+1] = perm[i:j+1][::-1]
                dist += 1
                break
        else:
            for j in range(i, len(perm)):
                if perm[j] == -val:
                    perm[i:j+1] = [-x for x in perm[i:j+1][::-1]]
                    dist += 1
                    break
    return dist
```

```python
def nussinov(rna, min_loop=3):
    PAIRS = {('A','U'),('U','A'),('G','C'),('C','G'),('G','U'),('U','G')}
    n = len(rna)
    dp = [[0]*n for _ in range(n)]
    for length in range(min_loop+1, n):
        for i in range(n - length):
            j = i + length
            dp[i][j] = dp[i][j-1]
            for k in range(i, j - min_loop):
                if (rna[k], rna[j]) in PAIRS:
                    left = dp[i][k-1] if k > i else 0
                    dp[i][j] = max(dp[i][j], left + dp[k+1][j-1] + 1)
    return dp[0][n-1]
```

```python
import numpy as np
from collections import Counter

def kmer_profile(seq, k=4):
    kmers = Counter(seq[i:i+k] for i in range(len(seq)-k+1))
    total = sum(kmers.values())
    return {kmer: cnt/total for kmer, cnt in kmers.items()}

def kmer_distance(p1, p2):
    keys = set(p1) | set(p2)
    v1 = np.array([p1.get(k, 0) for k in keys])
    v2 = np.array([p2.get(k, 0) for k in keys])
    return 1 - np.dot(v1, v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)+1e-9)

def simulate_genome(length, gc, seed=0):
    rng = np.random.default_rng(seed)
    return ''.join(rng.choice(['G','C'] if rng.random()<gc else ['A','T'])
                   for _ in range(length))

profiles = {
    'E. coli (51% GC)':    kmer_profile(simulate_genome(10000, 0.51, 1)),
    'S. aureus (33% GC)':  kmer_profile(simulate_genome(10000, 0.33, 2)),
    'Mycobacterium (65%)': kmer_profile(simulate_genome(10000, 0.65, 3)),
}
unknown = kmer_profile(simulate_genome(5000, 0.32, 4))  # should match S. aureus

print('k-mer distances to unknown sample:')
for name, profile in profiles.items():
    print(f'  {name}: {kmer_distance(profile, unknown):.4f}')
print('Classified as:', min(profiles, key=lambda n: kmer_distance(profiles[n], unknown)))
```
