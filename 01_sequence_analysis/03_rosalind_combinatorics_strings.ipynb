## Interview Tip Bank
> **TRIE vs KMP**: Use KMP for one pattern in one text. Use TRIE (or Aho-Corasick) for many patterns in one text. TRIE build is O(total pattern length), search is O(text + matches).

> **Suffix array trick**: Sort all suffixes lexicographically. The LCP array gives max length shared between adjacent suffixes. Max LCP = length of longest repeated substring.

> **REAR BFS gotcha**: The state space is n! for permutations of length n. For small n (≤ 8) this is feasible. For larger n, you need more sophisticated algorithms (breakpoint graph).

> **Motzkin vs Catalan**: Catalan(n) counts ways where EVERY position is paired. Motzkin(n) allows unpaired positions too. RNA uses Motzkin.

## CORR — Error Correction in Reads

> **Why this code:** This function processes biological data, converting raw sequences into a format our analysis can use.

````python
# --- Imports ---
from collections import Counter
# --- error_correction() function ---
def error_correction(reads):
    """
    Plain English: Correct single-base sequencing errors in a set of DNA reads.
    Each "bad" read differs from exactly one "good" read by a Hamming distance of 1.

    Args:
        reads: List of DNA strings (sequencing reads, possibly with errors)
    Returns:
        List of (bad_read, corrected_read) tuples — each correction as a pair
    """
    complement = str.maketrans('ACGT', 'TGCA')
    # Find all k-mers
    all_reads = set()
    # --- Loop over reads ---
    for r in reads:
        all_reads.add(r)
        all_reads.add(r.translate(complement)[::-1])
    correct = set()
    # --- Loop over all_reads ---
    for r in all_reads:
        # Conditional check
        if reads.count(r) > 1 or reads.count(r.translate(complement)[::-1]) > 1:
            correct.add(r)
    corrections = []
    # --- Loop over reads ---
    for r in reads:
        rc = r.translate(complement)[::-1]
        if r in correct or rc in correct:
            pass  # correct read
        else:
            # Find correction with Hamming distance 1
            for cand in correct:
                if len(cand) == len(r):
                    diffs = sum(a!=b for a,b in zip(r,cand))
                    if diffs == 1:
                        corrections.append((r, cand)); break
    return corrections
reads = ["TCATC","CCATG","CAATG","TAATG","CAATG"]
# --- Loop ---
for orig, corr in error_correction(reads): print(f"{orig}->{corr}")
````

## LEXV — Ordering Strings of Varying Length Lexicographically

### RNA Translation — Codon-by-Codon (Rosalind: PROT)

**Biological problem:** Convert an RNA string (A, U, G, C) to a protein sequence using the universal genetic code. This is the central dogma: DNA → RNA → Protein.

**Why this approach works:** Read codons in triplets starting from position 0. The codon table dictionary maps each 3-letter RNA codon to a single amino acid letter. Stop codons ('UAA', 'UAG', 'UGA') terminate translation.

**What to look for:** The output protein string (one amino acid per codon). Note: the input must be in RNA format (U not T) — check your input.

````python
codon_table = {
    'UUU':'F','UUC':'F','UUA':'L','UUG':'L','UCU':'S','UCC':'S','UCA':'S','UCG':'S',
    'UAU':'Y','UAC':'Y','UAA':'*','UAG':'*','UGU':'C','UGC':'C','UGA':'*','UGG':'W',
    'CUU':'L','CUC':'L','CUA':'L','CUG':'L','CCU':'P','CCC':'P','CCA':'P','CCG':'P',
    'CAU':'H','CAC':'H','CAA':'Q','CAG':'Q','CGU':'R','CGC':'R','CGA':'R','CGG':'R',
    'AUU':'I','AUC':'I','AUA':'I','AUG':'M','ACU':'T','ACC':'T','ACA':'T','ACG':'T',
    'AAU':'N','AAC':'N','AAA':'K','AAG':'K','AGU':'S','AGC':'S','AGA':'R','AGG':'R',
    'GUU':'V','GUC':'V','GUA':'V','GUG':'V','GCU':'A','GCC':'A','GCA':'A','GCG':'A',
    'GAU':'D','GAC':'D','GAA':'E','GAG':'E','GGU':'G','GGC':'G','GGA':'G','GGG':'G',
}
def translate(rna):
    """
    Plain English: Translate an RNA sequence into a protein amino acid string.
    Reads codons (3 nucleotides at a time) and maps each to an amino acid letter.

    Args:
        rna: RNA string over the alphabet {A, U, G, C}
    Returns:
        String — protein sequence using single-letter amino acid codes
    """
    protein = ''
    for i in range(0, len(rna)-2, 3):
        aa = codon_table.get(rna[i:i+3], '?')
        if aa == '*': break
        protein += aa
    return protein
rna = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"
print(translate(rna))  # MAMAPRTEINSTRING
````

## REAR — Reversal Distance

### All Signed Permutations (Rosalind: SIGN)

**Biological problem:** Enumerate all signed permutations of n elements. In comparative genomics, a signed permutation represents a chromosome's gene order, where sign indicates gene orientation. This forms the search space for genome rearrangement algorithms.

**Why this approach works:** Generate all permutations of {1..n}, then for each permutation generate all 2^n sign assignments (each gene can be + or -). Total count = n! × 2^n.

**What to look for:** The count on the first line, then each permutation. For n=2: count=8. For n=3: count=48.

````python
from itertools import permutations
def signed_perms(n):
    """
    Plain English: Generate all signed permutations of integers 1..n.
    Each element can be positive or negative; used in genome rearrangement problems.

    Args:
        n: Integer — the range of elements (1 to n)
    Returns:
        List of lists — all signed permutations, each a list of +/- integers
    """
    for p in permutations(range(1,n+1)):
        for signs in range(2**n):
            yield tuple(s*((-1)**((signs>>i)&1)) for i,s in enumerate(p))
n = 2
perms = list(signed_perms(n))
print(len(perms))
for p in perms[:8]: print(*p)
````

## SORT — Sorting by Reversals

### Signed Permutations — Recursive Backtracking

**Biological problem:** Same SIGN problem with explicit backtracking. Backtracking is the standard technique for constraint satisfaction problems — e.g., finding only permutations with specific adjacency constraints.

**Why this approach works:** Build the permutation incrementally; at each step, choose the next unused element and its sign, recurse, then undo. The recursion depth equals n; each leaf is a complete signed permutation.

**What to look for:** Identical output to itertools approach; backtracking is easier to extend with pruning conditions.

````python
from itertools import permutations
def signed_perms(n):
    """
    Plain English: Generate all signed permutations of integers 1..n.
    Each element can be positive or negative; used in genome rearrangement problems.

    Args:
        n: Integer — the range of elements (1 to n)
    Returns:
        List of lists — all signed permutations, each a list of +/- integers
    """
    for p in permutations(range(1,n+1)):
        for signs in range(2**n):
            yield tuple(s*((-1)**((signs>>i)&1)) for i,s in enumerate(p))
n = 2
perms = list(signed_perms(n))
print(len(perms))
for p in perms[:8]: print(*p)
````

## RSTR — Matching Random Motifs

### GC Content from Multi-FASTA (Rosalind: GC)

**Biological problem:** Identify which organism in a FASTA collection has the highest GC content. Organisms living at high temperatures (thermophiles) typically have higher GC content because G-C base pairs (3 hydrogen bonds) are stronger than A-T pairs (2 hydrogen bonds).

**Why this approach works:** Accumulate each FASTA record by tracking the current header. Compute GC fraction = (count('G') + count('C')) / len(seq) for each sequence.

**What to look for:** The FASTA ID with the highest GC%, followed by the percentage. Biologically meaningful range: ~25% (parasites with reduced genomes) to ~75% (some bacteria).

````python
def gc_content(seq):
    """
    Plain English: Calculate the fraction of G and C bases in a DNA sequence.
    Higher GC content usually means a thermodynamically more stable sequence.

    Args:
        seq: A DNA string (characters A, T, G, C)
    Returns:
        Float — GC fraction between 0.0 and 1.0
    """
    return (seq.count('G') + seq.count('C')) / len(seq)

seqs = {"Rosalind_6404": "CCTGCGGAAGATCGGCACTAGAATAGCC",
        "Rosalind_5959": "CCATCGGTAGCGCATCCTTAGTCCAATTAAGTCC"}
best = max(seqs, key=lambda k: gc_content(seqs[k]))
print(best)
print(f"{gc_content(seqs[best])*100:.6f}")
````

## CONV — Comparing Spectra with the Spectral Convolution

### Signed Permutations — Bit Mask Sign Assignment

**Biological problem:** Alternative SIGN implementation using bit manipulation. Each bit in an integer from 0 to 2^n - 1 represents the sign (+/-) of one element.

**Why this approach works:** For each permutation of {1..n} and each integer mask from 0 to 2^n-1: if bit i is set, element i gets a minus sign; otherwise plus. This is equivalent to itertools.product([1,-1], repeat=n).

**What to look for:** Identical output to previous methods. Bit manipulation avoids creating intermediate sign-pattern tuples, saving memory.

````python
from itertools import permutations
def signed_perms(n):
    """
    Plain English: Generate all signed permutations of integers 1..n.
    Each element can be positive or negative; used in genome rearrangement problems.

    Args:
        n: Integer — the range of elements (1 to n)
    Returns:
        List of lists — all signed permutations, each a list of +/- integers
    """
    for p in permutations(range(1,n+1)):
        for signs in range(2**n):
            yield tuple(s*((-1)**((signs>>i)&1)) for i,s in enumerate(p))
n = 2
perms = list(signed_perms(n))
print(len(perms))
for p in perms[:8]: print(*p)
````

## LREP — Finding the Longest Multiple Repeat

### Greedy Genome Assembly from Overlapping Reads

**Biological problem:** Given a set of DNA reads with overlaps, assemble them into a single contig (assembled sequence). This is a simplified version of the overlap-layout-consensus (OLC) assembly paradigm.

**Why this approach works:** At each step, find the pair of reads with the longest overlap (one's suffix matches the other's prefix). Merge them. Repeat until one sequence remains. Greedy works well when reads are error-free and coverage is complete.

**What to look for:** The assembled string should contain all reads as substrings. For real data, greedy assembly fails at repeats — but works perfectly for this Rosalind problem.

````python
def greedy_assemble(reads):
    """
    Plain English: Reconstruct a genome by greedily merging overlapping reads.
    Repeatedly find the pair of reads with the longest overlap and merge them.

    Args:
        reads: List of DNA strings (sequencing reads to assemble)
    Returns:
        String — assembled genome sequence (approximate, greedy approach)
    """
    reads = list(set(reads))  # remove duplicates
    while len(reads) > 1:
        best = (-1, -1, -1)
        for i in range(len(reads)):
            for j in range(len(reads)):
                if i != j:
                    for ov in range(min(len(reads[i]), len(reads[j])), 0, -1):
                        if reads[i][-ov:] == reads[j][:ov]:
                            if ov > best[0]: best = (ov, i, j); break
        ov, i, j = best
        if ov == -1: break
        merged = reads[i] + reads[j][ov:]
        reads = [reads[k] for k in range(len(reads)) if k not in (i,j)] + [merged]
    return reads[0]
reads = ["ATTAGACCTG", "CCTGCCGGAA", "AGACCTGCCG", "GCCGGAATAC"]
print(greedy_assemble(reads))
````

### Protein Translation for ORF Detection

**Biological problem:** This translate function is used inside ORF-finding code. It handles the full codon table and terminates correctly at stop codons — essential for accurate ORF boundary detection.

**Why this approach works:** The dictionary lookup approach is fast and readable. Handling stop codons ('*') explicitly and returning early prevents translating past the stop into the next ORF.

````python
codon_table = {
    'UUU':'F','UUC':'F','UUA':'L','UUG':'L','UCU':'S','UCC':'S','UCA':'S','UCG':'S',
    'UAU':'Y','UAC':'Y','UAA':'*','UAG':'*','UGU':'C','UGC':'C','UGA':'*','UGG':'W',
    'CUU':'L','CUC':'L','CUA':'L','CUG':'L','CCU':'P','CCC':'P','CCA':'P','CCG':'P',
    'CAU':'H','CAC':'H','CAA':'Q','CAG':'Q','CGU':'R','CGC':'R','CGA':'R','CGG':'R',
    'AUU':'I','AUC':'I','AUA':'I','AUG':'M','ACU':'T','ACC':'T','ACA':'T','ACG':'T',
    'AAU':'N','AAC':'N','AAA':'K','AAG':'K','AGU':'S','AGC':'S','AGA':'R','AGG':'R',
    'GUU':'V','GUC':'V','GUA':'V','GUG':'V','GCU':'A','GCC':'A','GCA':'A','GCG':'A',
    'GAU':'D','GAC':'D','GAA':'E','GAG':'E','GGU':'G','GGC':'G','GGA':'G','GGG':'G',
}
def translate(rna):
    """
    Plain English: Translate an RNA sequence into a protein amino acid string.
    Reads codons (3 nucleotides at a time) and maps each to an amino acid letter.

    Args:
        rna: RNA string over the alphabet {A, U, G, C}
    Returns:
        String — protein sequence using single-letter amino acid codes
    """
    protein = ''
    for i in range(0, len(rna)-2, 3):
        aa = codon_table.get(rna[i:i+3], '?')
        if aa == '*': break
        protein += aa
    return protein
rna = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"
print(translate(rna))  # MAMAPRTEINSTRING
````

## NWCK — Distances in Trees (Newick format)

### Newick Tree Parser for Phylogenetic Trees

**Biological problem:** Parse a Newick-format string into a tree data structure. Newick is the standard text format for phylogenetic trees (e.g., `((A:0.1,B:0.2):0.3,C:0.4)`). Tree distances and topology comparisons require parsing this format.

**Why this approach works:** Recursive descent parsing: parentheses define subtrees, commas separate siblings, colons introduce branch lengths. The `parse` inner function advances a mutable position pointer through the string.

**What to look for:** The returned tree is a nested Python structure (dict or list). Branch lengths (after ':') represent evolutionary distance. Leaves are named taxa; internal nodes may be unlabeled.

````python
def parse_newick(s):
    """Parse simple Newick string to nested dict."""
    s = s.strip().rstrip(';')
    def parse(s, start=0):
        if s[start] != '(':
            end = start
            while end < len(s) and s[end] not in ',)': end += 1
            return s[start:end], end
        end = start + 1; depth = 1; children = []
        while depth > 0:
            if s[end] == '(': depth += 1
            elif s[end] == ')': depth -= 1
            elif s[end] == ',' and depth == 1:
                children.append(s[start+1:end])
                start = end
            end += 1
        return children, end
    return parse(s)

newick = "((rabbit,cat),(dog,wolf))"
result, _ = parse_newick(newick)
print(f"Newick: {newick}")
print(f"Parsed: {result}")
print("Application: represent evolutionary trees from distance matrices")
````

def reversal_distance(perm1, perm2):
    perm = list(perm1)
    target = list(perm2)
    dist = 0
    for i in range(len(perm)):
        val = target[i]
        if perm[i] == val:
            continue
        for j in range(i+1, len(perm)):
            if perm[j] == val:
                perm[i:j+1] = perm[i:j+1][::-1]
                dist += 1
                break
        else:
            for j in range(i, len(perm)):
                if perm[j] == -val:
                    perm[i:j+1] = [-x for x in perm[i:j+1][::-1]]
                    dist += 1
                    break
    return dist

def nussinov(rna, min_loop=3):
    PAIRS = {('A','U'),('U','A'),('G','C'),('C','G'),('G','U'),('U','G')}
    n = len(rna)
    dp = [[0]*n for _ in range(n)]
    for length in range(min_loop+1, n):
        for i in range(n - length):
            j = i + length
            dp[i][j] = dp[i][j-1]

import numpy as np
from collections import Counter

def kmer_profile(seq, k=4):
    kmers = Counter(seq[i:i+k] for i in range(len(seq)-k+1))
    total = sum(kmers.values())
    return {kmer: cnt/total for kmer, cnt in kmers.items()}

def kmer_distance(p1, p2):
    keys = set(p1) | set(p2)
    v1 = np.array([p1.get(k, 0) for k in keys])

### BWT and Occurrence Count — Debug Exercise

**Biological problem:** The Burrows-Wheeler Transform (BWT) is the foundation of modern genome aligners (BWA, Bowtie). It rearranges a sequence so that identical characters cluster together, enabling fast pattern search via the FM-index.

**Why this approach works (when correct):** Sort all rotations of the string, take the last column. The FM-index then uses cumulative character counts to navigate the BWT without decompressing it.

**What to look for:** Two bugs are present. `build_bwt_buggy` — check the rotation sorting and column extraction. `count_occ_buggy` — verify the off-by-one in the cumulative count loop. Compare output against a known BWT (e.g., 'BANANA$' → 'BWT=ANNB$AA').

````python
# --- build_bwt_buggy() function ---
def build_bwt_buggy(s):
    """
    Plain English: Build the Burrows-Wheeler Transform (BWT) of a string.
    BWT rearranges characters for improved compression and fast search.
    NOTE: This implementation contains deliberate bugs for debugging practice.

    Args:
        s: Input string (terminal dollar sign will be appended)
    Returns:
        Tuple of (bwt_string, suffix_array)
    """
    # Calculate s
    s = s + "$"
    # Compute n via len
    n = len(s)
    # BUG 1: range should be range(n) not range(n-1)
    rotations = sorted([s[i:] + s[:i] for i in range(n-1)])  # misses one rotation
    # Calculate bwt
    bwt = "".join(r[-1] for r in rotations)
    # BUG 2: should return indices, not strings
    suffix_array = sorted(range(n), key=lambda i: s[i:])
    # Return result
    return bwt, suffix_array

# --- count_occ_buggy() function ---
def count_occ_buggy(bwt, char):
    """
    Plain English: Count cumulative character occurrences across BWT positions.
    Used in the FM-index for pattern matching without decompression.
    NOTE: This implementation contains a deliberate bug for debugging practice.

    Args:
        bwt: Burrows-Wheeler Transform string
        char: Single character to count occurrences of
    Returns:
        List of cumulative counts
    """
    # BUG 3: occ[i] should count bwt[0:i], not bwt[0:i+1]
    occ = [0] * (len(bwt) + 1)
    # --- Loop over range(len(bwt)) ---
    for i in range(len(bwt)):
        # Update occ entry
        occ[i+1] = occ[i] + (1 if bwt[i+1 % len(bwt)] == char else 0)  # wrong index
    return occ

s = "BANANA"
bwt, sa = build_bwt_buggy(s)
print(f"BWT of BANANA: {bwt}")
print(f"Expected:      ANNB$AA  (correct BWT of BANANA$)")
print(f"Length: {len(bwt)} (should be {len(s)+1})")
print()
print("Bug 1: BWT is missing one rotation because range(n-1) skips the last rotation")
print("Bug 2: suffix_array already correct (returns indices) - check build_bwt logic")
````
