## DP Table Template (Memorize This Pattern)
```python
def dp_align(s, t, match=1, mismatch=-1, gap=-2):
    m, n = len(s), len(t)
    dp = [[0]*(n+1) for _ in range(m+1)]
    # Initialize borders
    for i in range(m+1): dp[i][0] = i * gap
    for j in range(n+1): dp[0][j] = j * gap
    # Fill
    for i in range(1, m+1):
        for j in range(1, n+1):
            score = match if s[i-1] == t[j-1] else mismatch
            dp[i][j] = max(
                dp[i-1][j-1] + score,   # diagonal
                dp[i-1][j]   + gap,     # up (gap in t)
                dp[i][j-1]   + gap      # left (gap in s)
            )
    return dp
```

## 1. Edit Distance (Levenshtein) — Dynamic Programming Foundation

````python
# Edit Distance (Levenshtein) — Dynamic Programming
import numpy as np

def edit_distance(s1, s2):
    """Minimum edits (insert, delete, substitute) to convert s1 to s2."""
    m, n = len(s1), len(s2)
    dp = np.zeros((m+1, n+1), dtype=int)
    dp[:, 0] = np.arange(m+1)
    dp[0, :] = np.arange(n+1)
    for i in range(1, m+1):
        for j in range(1, n+1):
            if s1[i-1] == s2[j-1]:
                dp[i, j] = dp[i-1, j-1]
            else:
                dp[i, j] = 1 + min(dp[i-1, j],   # deletion
                                   dp[i, j-1],    # insertion
                                   dp[i-1, j-1])  # substitution
    return dp[m, n]

# Test cases
pairs = [
    ("KITTEN", "SITTING"),
    ("ATGCGATCG", "ATGCCATCG"),  # single SNP
    ("ACGT", "ACGT"),
    ("", "ATGC"),
]
print("Edit Distance:")
for s1, s2 in pairs:
    d = edit_distance(s1, s2)
    print(f"  d('{s1}', '{s2}') = {d}")

# Biological interpretation: edit distance ≈ evolutionary distance
# between DNA sequences (simplified model, ignoring gaps)
````

## 2. Needleman-Wunsch (Global Alignment)

### 📖 Reading Guide — Needleman-Wunsch Dynamic Programming Table

`dp = np.zeros((m+1, n+1))`
→ *Create a 2D grid of zeros. Rows = characters of seq1 (+1 for the empty prefix). Columns = seq2 characters.*

`dp[0, :] = np.arange(n+1) * gap`
→ *First row: aligning nothing with seq2[0..j] costs j × gap penalty. Fill left to right with 0, -2, -4, -6...*

`dp[:, 0] = np.arange(m+1) * gap`
→ *First column: aligning seq1[0..i] with nothing costs i × gap penalty. Fill top to bottom.*

`for i in range(1, m+1):`
→ *Fill the table row by row, left to right — each cell depends only on cells already filled.*

`match_score = match if seq1[i-1]==seq2[j-1] else mismatch`
→ *If the characters match, reward. If they differ, penalise. Note: i-1 because dp is 1-indexed but seq is 0-indexed.*

`dp[i,j] = max(dp[i-1,j-1]+match_score, dp[i-1,j]+gap, dp[i,j-1]+gap)`
→ *Three choices: diagonal (match/mismatch), up (gap in seq2), left (gap in seq1). Take the best.*

`# Traceback from dp[m][n] back to dp[0][0]`
→ *Follow the path of best choices backwards to reconstruct the optimal alignment.*



````python
# Needleman-Wunsch — Global Sequence Alignment
import numpy as np

def needleman_wunsch(seq1, seq2, match=1, mismatch=-1, gap=-2):
    """Global alignment with linear gap penalty."""
    m, n = len(seq1), len(seq2)
    dp = np.zeros((m+1, n+1))
    dp[0, :] = np.arange(n+1) * gap
    dp[:, 0] = np.arange(m+1) * gap

    for i in range(1, m+1):
        for j in range(1, n+1):
            s = match if seq1[i-1] == seq2[j-1] else mismatch
            dp[i,j] = max(dp[i-1,j-1]+s, dp[i-1,j]+gap, dp[i,j-1]+gap)

    # Traceback
    i, j = m, n
    a1, a2 = [], []
    while i > 0 or j > 0:
        s = match if i>0 and j>0 and seq1[i-1]==seq2[j-1] else mismatch
        if i>0 and j>0 and dp[i,j]==dp[i-1,j-1]+s:
            a1.append(seq1[i-1]); a2.append(seq2[j-1]); i-=1; j-=1
        elif i>0 and dp[i,j]==dp[i-1,j]+gap:
            a1.append(seq1[i-1]); a2.append('-'); i-=1
        else:
            a1.append('-'); a2.append(seq2[j-1]); j-=1

    a1 = ''.join(reversed(a1)); a2 = ''.join(reversed(a2))
    identity = sum(c1==c2 for c1,c2 in zip(a1,a2) if c1!='-' and c2!='-')
    return a1, a2, dp[m,n], identity

# Test
for s1, s2 in [("ACGT", "ACGT"), ("AGCT", "AACT"), ("ATGCGATCG", "ATCGATCG")]:
    a1, a2, score, ident = needleman_wunsch(s1, s2)
    print(f"NW('{s1}', '{s2}'):")
    print(f"  {a1}")
    print(f"  {a2}")
    print(f"  Score={score:.0f}, Identity={ident}/{len(s1)}")
    print()
````

## 3. Smith-Waterman (Local Alignment)

````python
# Smith-Waterman — Local Sequence Alignment
import numpy as np

def smith_waterman(seq1, seq2, match=2, mismatch=-1, gap=-2):
    """Find highest-scoring local alignment between seq1 and seq2."""
    m, n = len(seq1), len(seq2)
    dp = np.zeros((m+1, n+1))

    for i in range(1, m+1):
        for j in range(1, n+1):
            s = match if seq1[i-1] == seq2[j-1] else mismatch
            dp[i,j] = max(0, dp[i-1,j-1]+s, dp[i-1,j]+gap, dp[i,j-1]+gap)

    # Traceback from maximum
    max_i, max_j = np.unravel_index(dp.argmax(), dp.shape)
    i, j = max_i, max_j
    a1, a2 = [], []
    while i > 0 and j > 0 and dp[i,j] > 0:
        s = match if seq1[i-1]==seq2[j-1] else mismatch
        if dp[i,j] == dp[i-1,j-1]+s:
            a1.append(seq1[i-1]); a2.append(seq2[j-1]); i-=1; j-=1
        elif dp[i,j] == dp[i-1,j]+gap:
            a1.append(seq1[i-1]); a2.append('-'); i-=1
        else:
            a1.append('-'); a2.append(seq2[j-1]); j-=1

    a1 = ''.join(reversed(a1)); a2 = ''.join(reversed(a2))
    return a1, a2, dp[max_i, max_j]

seq1 = "ATGCGATCGATCGAATTCGATCG"
seq2 = "GAATTC"  # EcoRI restriction site
a1, a2, score = smith_waterman(seq1, seq2)
print(f"SW local alignment:")
print(f"  seq1 (local): {a1}")
print(f"  seq2:         {a2}")
print(f"  Score: {score:.0f}")
print(f"  Biological use: find conserved motifs within longer sequences")
````

## 🎯 Predict Before Running — Alignment Intuition

Before running alignment code, answer:

1. Given sequences `"AGCT"` and `"AGT"` with match=+1, mismatch=-1, gap=-2: what is the optimal NW alignment score? Work out the DP table by hand (4×3 matrix).
2. If you align two identical 100-residue sequences, what is the maximum possible NW score (match=+1)?
3. Smith-Waterman always reports a non-negative score. Why? What does a score of 0 mean biologically?

````python
# PREDICTION VERIFICATION — check your hand-computed answers

def nw_score(seq1, seq2, match=1, mismatch=-1, gap=-2):
    """Return only the optimal NW score (not the alignment)."""
    m, n = len(seq1), len(seq2)
    dp = [[0]*(n+1) for _ in range(m+1)]
    for i in range(m+1): dp[i][0] = i * gap
    for j in range(n+1): dp[0][j] = j * gap
    for i in range(1, m+1):
        for j in range(1, n+1):
            s = match if seq1[i-1] == seq2[j-1] else mismatch
            dp[i][j] = max(dp[i-1][j-1]+s, dp[i-1][j]+gap, dp[i][j-1]+gap)
    return dp[m][n]

# Question 1
s1, s2 = "AGCT", "AGT"
score = nw_score(s1, s2)
print(f"1. NW score for '{s1}' vs '{s2}': {score}")
print("   Hand-trace: AGC-T | AG-T → 3 matches - 1 gap penalty = 3(+1) + 1(-2) = +1")

# Question 2
identical = "A" * 100
print(f"\n2. NW score for 100 identical residues: {nw_score(identical, identical)}")
print("   = 100 × match_score = 100 × 1 = 100 (no gaps needed)")

# Question 3 — SW always ≥ 0
def sw_score(seq1, seq2, match=1, mismatch=-1, gap=-2):
    m, n = len(seq1), len(seq2)
    dp = [[0]*(n+1) for _ in range(m+1)]
    best = 0
    for i in range(1, m+1):
        for j in range(1, n+1):
            s = match if seq1[i-1] == seq2[j-1] else mismatch
            dp[i][j] = max(0, dp[i-1][j-1]+s, dp[i-1][j]+gap, dp[i][j-1]+gap)
            best = max(best, dp[i][j])
    return best

print(f"\n3. SW score for completely random unrelated sequences:")
import random; random.seed(42)
random_seqs = [''.join(random.choices('ACGT', k=50)) for _ in range(5)]
scores = [sw_score(random_seqs[i], random_seqs[i+1]) for i in range(4)]
print(f"   Scores: {scores}")
print("   Score=0 means NO local alignment is better than random chance (no conserved region).")
````

## 🔧 Connect to Production Tools — BLAST vs Your Implementation

Your NW/SW implementation is correct for educational purposes. In production, BLAST is used for database searches because it trades a small accuracy cost for 1000× speed improvement using a heuristic seed-and-extend approach.

````python
# COMPARISON: Your SW implementation vs BLAST strategy

# BLAST core idea:
# 1. Find "seeds" — short exact matches (k-mers) between query and database
# 2. Extend seeds if they pass a score threshold
# 3. Only run full alignment on promising regions

# Let's implement the seed-finding step (the BLAST speedup) and measure the difference

import time
from collections import defaultdict

def sw_full(seq1, seq2, match=2, mismatch=-3, gap=-2):
    """Full Smith-Waterman — O(m*n)"""
    m, n = len(seq1), len(seq2)
    dp = [[0]*(n+1) for _ in range(m+1)]
    best = 0
    for i in range(1, m+1):
        for j in range(1, n+1):
            s = match if seq1[i-1] == seq2[j-1] else mismatch
            dp[i][j] = max(0, dp[i-1][j-1]+s, dp[i-1][j]+gap, dp[i][j-1]+gap)
            best = max(best, dp[i][j])
    return best

def blast_like_seeds(query, database, k=3, threshold=3):
    """Find all exact k-mer seeds between query and database — O((|query|+|db|)*k)"""
    # Index database k-mers
    db_index = defaultdict(list)
    for j in range(len(database)-k+1):
        db_index[database[j:j+k]].append(j)

    # Find seeds in query
    hits = []
    for i in range(len(query)-k+1):
        kmer = query[i:i+k]
        for j in db_index[kmer]:
            hits.append((i, j))
    return hits

import random; random.seed(0)
# Simulate a 500bp query against a 5000bp database
query = ''.join(random.choices('ACGT', k=500))
database = ''.join(random.choices('ACGT', k=4500)) + query  # database contains the query

# Time full SW
t0 = time.perf_counter()
score = sw_full(query, database[:500])  # only 500bp for speed demo
sw_time = time.perf_counter() - t0

# Time BLAST-like seed finding
t0 = time.perf_counter()
seeds = blast_like_seeds(query, database, k=11)  # BLAST uses k=11 for nucleotides
seed_time = time.perf_counter() - t0

print(f"Full SW on 500×500: {sw_time*1000:.2f} ms")
print(f"BLAST seed-finding (500 query vs 5000 db): {seed_time*1000:.2f} ms")
print(f"Seeds found: {len(seeds)} (these are candidate regions to extend)")
print()
print("BLAST strategy: only run SW on regions with seeds → 1000-10000× speedup")
print("Cost: misses alignments with no exact k-mer matches (rare for similar sequences)")
print()
print("KEY INSIGHT: Your NW/SW implementations are algorithmically optimal.")
print("BLAST approximates SW with heuristics to scale to genome-sized databases.")
print("For protein databases: BLAST uses k=3 and BLOSUM62 neighborhood seeds.")
````
