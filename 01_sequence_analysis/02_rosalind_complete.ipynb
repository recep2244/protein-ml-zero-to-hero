### Smith-Waterman Local Alignment (Rosalind: LOCA)

**Biological problem:** Given two DNA or protein sequences, find the best-matching *substrings* — not the full sequences. This is useful when one sequence is a fragment of another.

**Why this approach works:** Dynamic programming fills a scoring matrix where each cell represents the best alignment ending at that position. A score of 0 acts as a "reset", so the algorithm finds local (partial) matches rather than forcing an end-to-end alignment.

**What to look for:** The highest value in the `dp` matrix is the best local alignment score. Traceback reveals which characters were aligned, mismatched, or gapped.

````python
import numpy as np
def local_align(s1, s2, match=2, mismatch=-1, gap=-2):
    """
    Plain English: Find the highest-scoring local alignment between two sequences (Smith-Waterman).

    Args:
        s1: First DNA/protein sequence string
        s2: Second DNA/protein sequence string
        match: Score bonus when characters are identical (default 2)
        mismatch: Score penalty when characters differ (default -1)
        gap: Score penalty for inserting a gap (default -2)
    Returns:
        Tuple of (score, aligned_s1, aligned_s2) — best local alignment
    """
    m, n = len(s1), len(s2)
    dp = np.zeros((m+1, n+1))
    for i in range(1,m+1):
        for j in range(1,n+1):
            sc = match if s1[i-1]==s2[j-1] else mismatch
            dp[i,j] = max(0, dp[i-1,j-1]+sc, dp[i-1,j]+gap, dp[i,j-1]+gap)
    return dp.max()
print(local_align("ACBDCA","ACBD"))
````

> **Expected output:** A numeric local alignment score, e.g., `8.0` for the test pair `ACBDCA` vs `ACBD`

### Fibonacci Rabbit Population (Rosalind: FIB)

**Biological problem:** Model an idealized rabbit population where every pair produces k offspring per month. This Fibonacci-like recurrence approximates real exponential population growth.

**Why this approach works:** `fib(n, k)` counts pairs at month n: newborns (k * pairs from 2 months ago) + surviving adults (pairs from last month). Memoization via dictionary avoids redundant recomputation.

**What to look for:** The output is a single large integer representing total pairs at month n. For large n and k, this number grows very fast.

````python
def fib(n, k=1):
    """
    Plain English: Model a rabbit population using Fibonacci-style recursion.
    Each pair produces k offspring pairs each month; models exponential growth.

    Args:
        n: Number of months to simulate
        k: Number of offspring pairs per mature pair per month (default 1)
    Returns:
        Integer — total number of rabbit pairs after n months
    """
    if n <= 2: return 1
    a, b = 1, 1
    for _ in range(n-2):
        a, b = b, b + k*a
    return b
print(fib(5, 3))  # 19
````

### Highest GC-Content Sequence (Rosalind: GC)

**Biological problem:** Given multiple DNA sequences (in FASTA format), find the one with the highest fraction of G and C bases. Higher GC content indicates thermodynamic stability — important for primer design and organism classification.

**Why this approach works:** GC content = (count of G + count of C) / total length. We parse FASTA by detecting lines starting with '>' as headers and accumulate sequence lines beneath each header.

**What to look for:** Output is the FASTA ID of the sequence with the highest GC%, followed by the percentage value.

````python
def gc_content(seq):
    """
    Plain English: Calculate the fraction of G and C bases in a DNA sequence.
    Higher GC content usually means a thermodynamically more stable sequence.

    Args:
        seq: A DNA string (characters A, T, G, C)
    Returns:
        Float — GC fraction between 0.0 and 1.0
    """
    return (seq.count('G') + seq.count('C')) / len(seq)

seqs = {"Rosalind_6404": "CCTGCGGAAGATCGGCACTAGAATAGCC",
        "Rosalind_5959": "CCATCGGTAGCGCATCCTTAGTCCAATTAAGTCC"}
best = max(seqs, key=lambda k: gc_content(seqs[k]))
print(best)
print(f"{gc_content(seqs[best])*100:.6f}")
````

### Signed Permutations (Rosalind: SIGN)

**Biological problem:** In genome rearrangement, chromosomes can be modeled as signed permutations — sequences where each gene has an orientation (+/-). Counting all possible rearrangements requires enumerating all signed permutations of n elements.

**Why this approach works:** For each permutation of {1..n}, we generate all 2^n sign assignments (each element can be + or -). The total count is n! × 2^n.

**What to look for:** First line is the total count. Each subsequent line is one signed permutation (space-separated signed integers).

````python
from itertools import permutations
def signed_perms(n):
    """
    Plain English: Generate all signed permutations of integers 1..n.
    Each element can be positive or negative; used in genome rearrangement problems.

    Args:
        n: Integer — the range of elements (1 to n)
    Returns:
        List of lists — all signed permutations, each a list of +/- integers
    """
    for p in permutations(range(1,n+1)):
        for signs in range(2**n):
            yield tuple(s*((-1)**((signs>>i)&1)) for i,s in enumerate(p))
n = 2
perms = list(signed_perms(n))
print(len(perms))
for p in perms[:8]: print(*p)
````

### Mortal Fibonacci Rabbits (Rosalind: FIBD)

**Biological problem:** Extend the Fibonacci model: rabbits die after m months. This more realistic model is used in population dynamics where generations have finite lifespans.

**Why this approach works:** We track a list where each index represents an age cohort. Each month, all cohorts age by one step, the oldest cohort dies, and newborns = sum of all reproductive-age adults.

**What to look for:** The count decreases or stabilizes after many months (unlike pure Fibonacci which grows without bound).

````python
def fib(n, k=1):
    """
    Plain English: Model a rabbit population using Fibonacci-style recursion.
    Each pair produces k offspring pairs each month; models exponential growth.

    Args:
        n: Number of months to simulate
        k: Number of offspring pairs per mature pair per month (default 1)
    Returns:
        Integer — total number of rabbit pairs after n months
    """
    if n <= 2: return 1
    a, b = 1, 1
    for _ in range(n-2):
        a, b = b, b + k*a
    return b
print(fib(5, 3))  # 19
````

### Longest Common Substring Motif (Rosalind: LCSM)

**Biological problem:** Find the longest DNA motif shared across all given sequences. A motif conserved in all samples is likely functionally important (e.g., a transcription factor binding site).

**Why this approach works:** Binary search on the length k: "does any k-length substring appear in all sequences?" We use rolling hash or direct set intersection. Shorter motifs are always easier to find than longer ones — binary search finds the maximum length efficiently.

**What to look for:** A single DNA string that is a substring of every input sequence.

````python
def lcsm(seqs):
    """
    Plain English: Find the Longest Common Substring present in all given sequences.
    A motif shared across all samples is a biologically significant signal.

    Args:
        seqs: List of DNA/protein strings
    Returns:
        String — the longest substring found in every input sequence
    """
    shortest = min(seqs, key=len)
    for L in range(len(shortest), 0, -1):
        for i in range(len(shortest)-L+1):
            motif = shortest[i:i+L]
            if all(motif in s for s in seqs):
                return motif
    return ""
seqs = ["GATTACA", "TAGACCA", "ATACA"]
print(lcsm(seqs))
````

### Enumerating Signed Permutations — Recursive Approach

**Biological problem:** Same SIGN problem with recursive backtracking. Backtracking is intuitive and easily extended to constrained enumeration (e.g., inversions between adjacent elements only).

**Why this approach works:** At each step, pick an unused element, assign it a +/- sign, recurse, then undo. This depth-first search visits every branch of the permutation tree exactly once.

**What to look for:** Output is identical to iterative approach; compare runtime for large n.

**Biological problem:** Find all positions where a short motif (pattern) appears in a longer DNA sequence. Motif positions reveal functional elements like promoters, restriction sites, or splice signals.

### Needleman-Wunsch Global Alignment (Rosalind: GLOB)

**Biological problem:** Align two sequences end-to-end optimally. Unlike Smith-Waterman, every character must be placed — no "free" boundaries. Used when sequences are similar in length and you want to compare them fully.

**Why this approach works:** DP table initialized with cumulative gap penalties at boundaries (unlike local alignment which starts at 0). Traceback follows max-scoring arrows from bottom-right to top-left.

**What to look for:** Alignment score (can be negative if sequences are very different) and the aligned sequences with '-' for gaps.

````python
import numpy as np
def global_align(s1, s2, match=1, mismatch=-1, gap=-2):
    """
    Plain English: Align two sequences end-to-end, penalizing gaps (Needleman-Wunsch).

    Args:
        s1: First sequence string
        s2: Second sequence string
        match: Score bonus for identical characters (default 1)
        mismatch: Score penalty for mismatched characters (default -1)
        gap: Score penalty for gaps (default -2)
    Returns:
        Tuple of (score, aligned_s1, aligned_s2)
    """
    m, n = len(s1), len(s2)
    dp = np.zeros((m+1, n+1))
    dp[0,:] = np.arange(n+1)*gap; dp[:,0] = np.arange(m+1)*gap
    for i in range(1,m+1):
        for j in range(1,n+1):
            sc = match if s1[i-1]==s2[j-1] else mismatch
            dp[i,j] = max(dp[i-1,j-1]+sc, dp[i-1,j]+gap, dp[i,j-1]+gap)
    return dp[m,n]
print(global_align("AACCTTGG","ACACTGTGA"))
````

import urllib.request, re

def fetch_uniprot(protein_id: str) -> str:
    """Fetch protein FASTA from UniProt by accession ID."""
    url = f"https://www.uniprot.org/uniprot/{protein_id}.fasta"
    with urllib.request.urlopen(url) as r:
        fasta = r.read().decode()
    lines = fasta.strip().split('\n')
    return ''.join(lines[1:])  # skip header

def find_glycosylation_sites(protein_id: str) -> list[int]:
    """
    Find all N-glycosylation motif positions (1-indexed) in a UniProt protein.
    Motif: N[^P][ST] (Asn followed by any-except-Pro followed by Ser or Thr)
    """
    seq = fetch_uniprot(protein_id)
    motif = re.compile(r'(?=(N[^P][ST]))')  # overlapping matches
    return [m.start() + 1 for m in motif.finditer(seq)]

## 🎯 Interview Questions — Sequence Algorithms

**Q1 (Warm-up):** What is the time complexity of Needleman-Wunsch? Why?
> **A:** O(mn) time and space, where m, n are sequence lengths. We fill an m×n DP table, each cell in O(1).

**Q2 (Medium):** How does Smith-Waterman differ from Needleman-Wunsch for protein homology search?
> **A:** SW uses local alignment (zeros out negatives, allows alignment to start/end anywhere); NW global alignment forces alignment of entire sequences. SW finds the best-matching subsequence, which matters when proteins share a conserved domain but differ elsewhere.

**Q3 (Hard):** BLAST uses a heuristic instead of exact Smith-Waterman. What is the heuristic and why does it work?
> **A:** BLAST first finds exact k-mer "seeds" (words) between query and database. Only regions with high-scoring seeds are extended with gap-free alignment, then gapped alignment. This works because truly homologous sequences almost always share some exact short subsequences. It reduces runtime from O(query × database) to near-linear.

**Q4 (Research-level):** Why does profile HMM alignment (HHblits) outperform pairwise alignment (BLAST) for remote homologs?
> **A:** Profile HMMs represent the evolutionary distribution at each position (via emission probabilities), capturing conservation patterns across thousands of homologs. This makes them sensitive to evolutionary signal even when pairwise identity drops below 20%. AF3 uses HHblits MSA as input precisely because of this sensitivity.

### Complete Utility Function Suite

**Biological problem:** This cell collects all essential utility functions into one place — reverse complement, GC content, and translation. Useful as a template for new Rosalind problems.

**Why this approach works:** Centralizing these functions eliminates copy-paste errors across cells. The translate function handles the full codon table including stop codons.

**What to look for:** Each function is self-contained. The codon table maps all 64 RNA codons; '*' indicates stop codons.

````python
# Build complement mapping
complement = {"A":"T", "C":"G", "G":"C"}  # BUG 1: missing "T":"A"

# --- reverse_complement() function ---
def reverse_complement(seq):
    """
    Plain English: Return the reverse complement of a DNA string.
    Used to check both strands of DNA for ORFs, motifs, or alignments.
    NOTE: This cell contains deliberate bugs for debugging practice.

    Args:
        seq: DNA string using characters A, T, G, C
    Returns:
        String — reverse complement sequence
    """
    # Return result
    return "".join(complement[b] for b in reversed(seq))

# --- gc_content() function ---
def gc_content(seq):
    """
    Plain English: Calculate the fraction of G and C bases in a DNA sequence.
    Higher GC content = more thermodynamically stable sequence.
    NOTE: This cell contains deliberate bugs for debugging practice.

    Args:
        seq: A DNA string (characters A, T, G, C)
    Returns:
        Float — GC fraction between 0.0 and 1.0
    """
    # BUG 2: gaps counted in denominator
    return (seq.count("G") + seq.count("C")) / len(seq)

# Define codon_table dict/set
codon_table = {
    "ATG":"M","GCT":"A","GAA":"E","TGG":"W","AAA":"K",
    "TAA":"*","TAG":"*","TGA":"*"
}

# --- translate() function ---
def translate(seq):
    """
    Plain English: Translate a DNA/RNA sequence into a protein string.
    Reads codons (3 nucleotides at a time) and maps each to an amino acid.
    NOTE: This cell contains deliberate bugs for debugging practice.

    Args:
        seq: DNA or RNA string — codons read in triplets
    Returns:
        String — protein sequence using single-letter amino acid codes
    """
    # Define protein list
    protein = []
    # --- Loop over range(0, len(seq), 3) ---
    for i in range(0, len(seq), 3):  # BUG 3: doesnt stop at stop codon
        # Calculate codon
        codon = seq[i:i+3]
        # Conditional check
        if len(codon) == 3 and codon in codon_table:
            # Append to protein
            protein.append(codon_table[codon])
    return "".join(protein)

# Test
test_seq = "ATGCTTAAGCCG"
# Error handling block
try:
    rc = reverse_complement(test_seq)
    print(f"Reverse complement: {rc}")
except KeyError as e:
    print(f"Bug 1 triggered: KeyError {e} - T is missing from complement dict")

gapped = "ATGC--ATGC"
print(f"GC content with gaps (buggy):  {gc_content(gapped):.3f}  (gaps inflate denominator)")
print(f"GC content correct:           {(gapped.count('G')+gapped.count('C'))/(len(gapped)-gapped.count('-')):.3f}")

orf = "ATGAAATAATGCTTGG"  # ATG-AAA-TAA-GCT-TGG
print(f"Translation (buggy, no stop): {translate(orf)}")
print(f"Expected (stop at TAA):       MK")
````
