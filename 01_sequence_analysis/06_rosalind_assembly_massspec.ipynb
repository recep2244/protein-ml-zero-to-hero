## Interview Tip Bank
> **Eulerian circuit exists iff**: all vertices have equal in-degree and out-degree AND the graph is connected. For circular genome, these should both hold.

> **PDPL backtracking pruning**: Before placing a candidate position x, check that all distances from x to already-placed positions are in the remaining multiset. If any distance is missing, prune immediately.

> **PRSM gotcha**: You're comparing a theoretical spectrum (computed from candidate protein) to an experimental spectrum. Score = number of peaks in common. Normalize by dividing by total spectrum length.

> **Progressive alignment**: "Center star" picks the sequence with max total similarity to all others as the center. Aligns all others to the center. Fast but not optimal.

## What This Notebook Teaches (Plain English)

**Two completely different but equally important problems**:

### Genome Assembly
Sequencing machines read short DNA fragments (150-300 letters). Assembly = puzzle: put millions of fragments back in order to reconstruct the original genome.

**De Bruijn graph approach**: Instead of comparing every pair of fragments (too slow), split each fragment into overlapping k-mers and build a directed graph. Finding an Eulerian path through this graph = the assembled sequence.

### Mass Spectrometry Peptide Sequencing
A mass spectrometer breaks a protein into fragments and weighs each fragment. Given only the mass list, reconstruct the amino acid sequence.

**The algorithm**: Mass differences between sorted fragment masses → amino acids → sequence.

**Prerequisites**: Graph algorithms (directed graphs, DFS/BFS) + combinatorics.

### Sequencing Error Correction (Rosalind: CORR)

**Biological problem:** Next-generation sequencers produce millions of short reads with ~0.1-1% error rates. Before assembly, we must correct these errors. A "bad" read differs from its true sequence by exactly one base (Hamming distance 1). The majority of reads are correct; we use them as a reference to fix the minority of erroneous reads.

**Why this approach works:** First, identify which reads are "good" (appear at least twice or have a complement that appears). Then, for each "bad" read, find the closest good read at Hamming distance 1. Return the correction pair.

**What to look for:** Output is a list of (error_read → corrected_read) pairs. Each line shows a substitution: the original erroneous read followed by an arrow and the corrected version.

````python
from collections import Counter
def error_correction(reads):
    """
    Plain English: Correct single-base sequencing errors in a set of DNA reads.
    Each "bad" read differs from exactly one "good" read by a Hamming distance of 1.

    Args:
        reads: List of DNA strings (sequencing reads, possibly with errors)
    Returns:
        List of (bad_read, corrected_read) tuples — each correction as a pair
    """
    complement = str.maketrans('ACGT', 'TGCA')
    # Find all k-mers
    all_reads = set()
    for r in reads:
        all_reads.add(r)
        all_reads.add(r.translate(complement)[::-1])
    correct = set()
    for r in all_reads:
        if reads.count(r) > 1 or reads.count(r.translate(complement)[::-1]) > 1:
            correct.add(r)
    corrections = []
    for r in reads:
        rc = r.translate(complement)[::-1]
        if r in correct or rc in correct:
            pass  # correct read
        else:
            # Find correction with Hamming distance 1
            for cand in correct:
                if len(cand) == len(r):
                    diffs = sum(a!=b for a,b in zip(r,cand))
                    if diffs == 1:
                        corrections.append((r, cand)); break
    return corrections
reads = ["TCATC","CCATG","CAATG","TAATG","CAATG"]
for orig, corr in error_correction(reads): print(f"{orig}->{corr}")
````

## PCOV — Genome Assembly with Perfect Coverage

### GC Content for Assembly Quality Control

**Biological problem:** During genome assembly, GC content is used as a quality filter. Regions with extreme GC (>70% or <30%) are prone to sequencing errors and assembly gaps because GC-rich regions are harder to sequence reliably.

**Why this approach works:** Compute GC fraction per read or contig. Flag outliers for manual review or additional coverage. Uniform GC distribution across reads suggests good library prep.

**What to look for:** A float between 0 and 1. Typical bacterial genomes: 40-65%. Extreme values (>75% or <25%) may indicate sequencing artifacts or contamination.

````python
def gc_content(seq):
    """
    Plain English: Calculate the fraction of G and C bases in a DNA sequence.
    Higher GC content usually means a thermodynamically more stable sequence.

    Args:
        seq: A DNA string (characters A, T, G, C)
    Returns:
        Float — GC fraction between 0.0 and 1.0
    """
    return (seq.count('G') + seq.count('C')) / len(seq)

seqs = {"Rosalind_6404": "CCTGCGGAAGATCGGCACTAGAATAGCC",
        "Rosalind_5959": "CCATCGGTAGCGCATCCTTAGTCCAATTAAGTCC"}
best = max(seqs, key=lambda k: gc_content(seqs[k]))
print(best)
print(f"{gc_content(seqs[best])*100:.6f}")
````

## GREP — Genome Assembly with Perfect Coverage and Repeats

### Signed Permutations for Comparative Genomics

**Biological problem:** Genome assembly produces contigs, not chromosomes. To arrange contigs into chromosomes, we need to determine both order and orientation — exactly the signed permutation problem.

**Why this approach works:** Enumerate all n! × 2^n arrangements of n contigs with n orientations. Scaffolding algorithms then use paired-end read evidence to score which arrangements are consistent with the data.

**What to look for:** This cell generates all candidate contig arrangements. Subsequent scoring (using read pair distances) selects the most likely chromosome-level assembly.

````python
from itertools import permutations
def signed_perms(n):
    """
    Plain English: Generate all signed permutations of integers 1..n.
    Each element can be positive or negative; used in genome rearrangement problems.

    Args:
        n: Integer — the range of elements (1 to n)
    Returns:
        List of lists — all signed permutations, each a list of +/- integers
    """
    for p in permutations(range(1,n+1)):
        for signs in range(2**n):
            yield tuple(s*((-1)**((signs>>i)&1)) for i,s in enumerate(p))
n = 2
perms = list(signed_perms(n))
print(len(perms))
for p in perms[:8]: print(*p)
````

## SPEC — Inferring Protein from Spectrum

### Mass Spectrometry Protein Sequencing — Translation Setup

**Biological problem:** In mass spectrometry proteomics, we know a protein's mass spectrum (list of fragment masses) and want to reconstruct its amino acid sequence. This requires the amino acid mass table: each amino acid has a unique monoisotopic mass.

**Why this approach works:** Every amino acid has a distinct mass (to 4 decimal places). A protein's mass spectrum shows masses of all prefix fragments (b-ions) or suffix fragments (y-ions). By taking mass differences, we read off the amino acid sequence.

**What to look for:** The codon-to-amino acid mapping and the amino acid mass table. The mass of each amino acid is in Daltons (unified atomic mass units).

````python
codon_table = {
    'UUU':'F','UUC':'F','UUA':'L','UUG':'L','UCU':'S','UCC':'S','UCA':'S','UCG':'S',
    'UAU':'Y','UAC':'Y','UAA':'*','UAG':'*','UGU':'C','UGC':'C','UGA':'*','UGG':'W',
    'CUU':'L','CUC':'L','CUA':'L','CUG':'L','CCU':'P','CCC':'P','CCA':'P','CCG':'P',
    'CAU':'H','CAC':'H','CAA':'Q','CAG':'Q','CGU':'R','CGC':'R','CGA':'R','CGG':'R',
    'AUU':'I','AUC':'I','AUA':'I','AUG':'M','ACU':'T','ACC':'T','ACA':'T','ACG':'T',
    'AAU':'N','AAC':'N','AAA':'K','AAG':'K','AGU':'S','AGC':'S','AGA':'R','AGG':'R',
    'GUU':'V','GUC':'V','GUA':'V','GUG':'V','GCU':'A','GCC':'A','GCA':'A','GCG':'A',
    'GAU':'D','GAC':'D','GAA':'E','GAG':'E','GGU':'G','GGC':'G','GGA':'G','GGG':'G',
}
def translate(rna):
    """
    Plain English: Translate an RNA sequence into a protein amino acid string.
    Reads codons (3 nucleotides at a time) and maps each to an amino acid letter.

    Args:
        rna: RNA string over the alphabet {A, U, G, C}
    Returns:
        String — protein sequence using single-letter amino acid codes
    """
    protein = ''
    for i in range(0, len(rna)-2, 3):
        aa = codon_table.get(rna[i:i+3], '?')
        if aa == '*': break
        protein += aa
    return protein
rna = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"
print(translate(rna))  # MAMAPRTEINSTRING
````

## SGRA — Using Spectrum Graph to Infer Peptides

### Signed Permutations for Contig Scaffolding

**Biological problem:** During genome finishing, assembled contigs must be ordered and oriented. This is the "scaffolding" step, mathematically equivalent to finding the optimal signed permutation of contigs.

**Why this approach works:** Enumerate all possible orderings and orientations. Paired-end reads that span contig boundaries provide constraints (distance and orientation) to score each arrangement.

**What to look for:** The count is the scaffolding search space size. Real assemblers (SPAdes, Velvet) use graph algorithms to avoid this exhaustive search, but the combinatorics here explain why scaffolding is computationally hard.

````python
from itertools import permutations
def signed_perms(n):
    """
    Plain English: Generate all signed permutations of integers 1..n.
    Each element can be positive or negative; used in genome rearrangement problems.

    Args:
        n: Integer — the range of elements (1 to n)
    Returns:
        List of lists — all signed permutations, each a list of +/- integers
    """
    for p in permutations(range(1,n+1)):
        for signs in range(2**n):
            yield tuple(s*((-1)**((signs>>i)&1)) for i,s in enumerate(p))
n = 2
perms = list(signed_perms(n))
print(len(perms))
for p in perms[:8]: print(*p)
````

## PDPL — Creating a Restriction Map

### Smith-Waterman Local Alignment in Assembly Context (Rosalind: LOCA)

**Biological problem:** Local alignment is used in assembly to detect and merge overlapping reads, and in variant calling to align reads to a reference genome. Finding the best local match identifies where a read belongs.

**Why this approach works:** The Smith-Waterman DP matrix computes the best alignment between any substring of s1 and any substring of s2. The match score is maximized at the boundary of the true overlap.

**What to look for:** The score and aligned substrings. In assembly, we care about the overlap suffix of one read aligning to the prefix of another. Here, the full local alignment is returned for general use.

````python
import numpy as np
def local_align(s1, s2, match=2, mismatch=-1, gap=-2):
    """
    Plain English: Find the highest-scoring local alignment between two sequences (Smith-Waterman).

    Args:
        s1: First DNA/protein sequence string
        s2: Second DNA/protein sequence string
        match: Score bonus when characters are identical (default 2)
        mismatch: Score penalty when characters differ (default -1)
        gap: Score penalty for inserting a gap (default -2)
    Returns:
        Tuple of (score, aligned_s1, aligned_s2) — best local alignment
    """
    m, n = len(s1), len(s2)
    dp = np.zeros((m+1, n+1))
    for i in range(1,m+1):
        for j in range(1,n+1):
            sc = match if s1[i-1]==s2[j-1] else mismatch
            dp[i,j] = max(0, dp[i-1,j-1]+sc, dp[i-1,j]+gap, dp[i,j-1]+gap)
    return dp.max()
print(local_align("ACBDCA","ACBD"))
````

## MULT — Multiple Alignment

### GC Content for Contig Quality Scoring

**Biological problem:** In genome assembly quality assessment, we compute GC content per contig to detect misassemblies. A contig with GC content very different from the genomic average likely results from a misassembly or contamination.

**Why this approach works:** Compare each contig's GC content to the genome-wide average (computed from all contigs). Flag contigs with |GC_contig - GC_genome| > 0.1 for manual inspection.

**What to look for:** Contigs with extreme GC content (e.g., <20% or >80%) are suspicious. Real assembler quality reports (e.g., QUAST) use this same metric.

````python
def gc_content(seq):
    """
    Plain English: Calculate the fraction of G and C bases in a DNA sequence.
    Higher GC content usually means a thermodynamically more stable sequence.

    Args:
        seq: A DNA string (characters A, T, G, C)
    Returns:
        Float — GC fraction between 0.0 and 1.0
    """
    return (seq.count('G') + seq.count('C')) / len(seq)

seqs = {"Rosalind_6404": "CCTGCGGAAGATCGGCACTAGAATAGCC",
        "Rosalind_5959": "CCATCGGTAGCGCATCCTTAGTCCAATTAAGTCC"}
best = max(seqs, key=lambda k: gc_content(seqs[k]))
print(best)
print(f"{gc_content(seqs[best])*100:.6f}")
````

**Dataset/Source**: E. coli K-12 MG1655 (SRA: SRR2584863 — 1.1M reads, 150 bp Illumina)

```python
from collections import defaultdict
import random

def build_debruijn(reads, k):
    graph = defaultdict(list)
    in_deg, out_deg = defaultdict(int), defaultdict(int)
    for read in reads:
        for i in range(len(read) - k):
            u, v = read[i:i+k], read[i+1:i+k+1]
            graph[u].append(v); out_deg[u] += 1; in_deg[v] += 1
    return graph, in_deg, out_deg

def check_eulerian(graph, in_deg, out_deg):
    nodes = set(graph) | set(in_deg)
    imbalanced = [(n, out_deg[n]-in_deg.get(n,0)) for n in nodes
                  if out_deg.get(n,0) != in_deg.get(n,0)]
    return len(imbalanced)==0, imbalanced

def assemble(reads, k=15):
    import copy
    graph, in_deg, out_deg = build_debruijn(reads, k)
    ok, bad = check_eulerian(graph, in_deg, out_deg)
    if not ok:
        print(f'  k={k}: {len(bad)} imbalanced nodes — need more coverage')
        return None
    g = copy.deepcopy(graph)
    start = next(iter(g))
    stack, path = [start], []
    while stack:
        v = stack[-1]
        if g.get(v): stack.append(g[v].pop())
        else: path.append(stack.pop())
    path = path[::-1]
    return path[0] + ''.join(n[-1] for n in path[1:-1])

# Simulate E. coli-like reads
ecoli = ('ATGCGTAACGTCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGA'
         'CGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATC')

random.seed(42)
doubled = ecoli + ecoli
reads = [doubled[i:i+25] for i in range(0, len(ecoli), 3)]
print(f'Reads: {len(reads)}, genome size: {len(ecoli)} bp')

for k in [8, 12, 15]:
    result = assemble(reads, k=k)
    if result:
        ident = sum(a==b for a,b in zip(result,ecoli))/max(len(result),len(ecoli))
        print(f'  k={k}: {len(result)} bp assembled, {ident:.1%} identity')

def compute_n50(contigs):
    lengths = sorted([len(c) for c in contigs], reverse=True)
    cumsum, total = 0, sum(lengths)
    for L in lengths:
        cumsum += L
        if cumsum >= total/2: return L
    return 0

print(f'N50 example: {compute_n50(["A"*300,"A"*200,"A"*150,"A"*100,"A"*80])} bp')
```

```python
import copy

def circular_assembly(reads, k):
    graph = defaultdict(list)
    in_deg, out_deg = defaultdict(int), defaultdict(int)
    for read in reads:
        for i in range(len(read) - k):
            u, v = read[i:i+k], read[i+1:i+k+1]
            graph[u].append(v); out_deg[u]+=1; in_deg[v]+=1
    
    # Circular: every node must be balanced
    imbalanced = [(n, out_deg.get(n,0)-in_deg.get(n,0))
                  for n in set(graph)|set(in_deg)
                  if out_deg.get(n,0)!=in_deg.get(n,0)]
    if imbalanced:
        print(f'  {len(imbalanced)} imbalanced nodes — try different k')
        return None
    
    g = copy.deepcopy(graph)
    start = next(iter(g))
    stack, path = [start], []
    while stack:
        v = stack[-1]
        if g.get(v): stack.append(g[v].pop())
        else: path.append(stack.pop())
    path = path[::-1]
    
    # Verify circular: circuit starts and ends at same node
    if path[0] == path[-1]:
        genome = path[0] + ''.join(n[-1] for n in path[1:-1])
```

**Bug 1:** k-mer extraction: `range(len(seq) - k)` should be `range(len(seq) - k + 1)` — off-by-one misses the last k-mer.

### De Bruijn Graph Assembly — Debug Exercise

**Biological problem:** De Bruijn graph assembly is the dominant approach for short-read (Illumina) genome assembly. k-mers become nodes, and overlaps between consecutive k-mers become directed edges. An Eulerian path through this graph spells out the assembled genome.

**Why this approach works (when correct):** 1) Extract all k-mers from reads. 2) Build a directed graph where each edge is a k-mer, connecting its (k-1)-prefix to its (k-1)-suffix. 3) Find an Eulerian path (if one exists — requires specific in-degree/out-degree conditions).

**What to look for — debugging guide:** `extract_kmers_buggy`: check the range boundary (should iterate to len(seq)-k+1, not len(seq)-k). `build_debruijn_buggy`: verify edge direction (prefix → suffix, not suffix → prefix). `has_eulerian_path_buggy`: check the degree condition — exactly 0 or 2 nodes with in-degree ≠ out-degree, not the wrong inequality.

````python
from collections import defaultdict

def extract_kmers_buggy(seq, k):
    """
    Plain English: Extract all k-mers of length k from a DNA sequence.
    k-mers are the building blocks of De Bruijn graph assembly.
    NOTE: This implementation has a deliberate off-by-one bug.

    Args:
        seq: DNA string to extract k-mers from
        k: Integer length of each k-mer
    Returns:
        List of k-mer strings (missing the last one due to the bug)
    """
    # BUG 1: off-by-one — misses last k-mer
    return [seq[i:i+k] for i in range(len(seq) - k)]  # should be range(len(seq)-k+1)

def build_debruijn_buggy(kmers):
    """
    Plain English: Build a De Bruijn graph from k-mers.
    Each k-mer becomes a directed edge: (k-1)-prefix -> (k-1)-suffix.
    NOTE: This implementation has a deliberate edge-direction bug.

    Args:
        kmers: List of k-mer strings
    Returns:
        Dict mapping each (k-1)-mer to a list of neighbor (k-1)-mers
    """
    graph = defaultdict(list)
    for kmer in kmers:
        # BUG 2: edge direction reversed
        suffix, prefix = kmer[:-1], kmer[1:]  # should be prefix, suffix = kmer[:-1], kmer[1:]
        graph[suffix].append(prefix)           # edge goes wrong way
    return dict(graph)

def has_eulerian_path_buggy(graph):
    """
    Plain English: Check if a directed graph has an Eulerian path (visits every edge once).
    Eulerian path = genome assembly is possible in one contig.
    NOTE: This implementation contains a deliberate logical bug.

    Args:
        graph: Dict mapping each node to a list of neighbor nodes
    Returns:
        Boolean — True if an Eulerian path exists
    """
    in_deg = defaultdict(int)
    out_deg = defaultdict(int)
    for node, neighbors in graph.items():
        out_deg[node] += len(neighbors)
        for n in neighbors:
            in_deg[n] += 1
    all_nodes = set(in_deg) | set(out_deg)
    # BUG 3: checks circuit condition (all balanced), not path condition
    return all(out_deg[n] == in_deg[n] for n in all_nodes)  # wrong for path

seq = "AGTACGT"
k = 3
kmers = extract_kmers_buggy(seq, k)
print(f"k-mers (buggy, k={k}): {kmers}")
print(f"Expected:             {[seq[i:i+k] for i in range(len(seq)-k+1)]}")
print(f"Missing:              {seq[-k:]} (last k-mer)")

g = build_debruijn_buggy(kmers)
print(f"\nDe Bruijn graph (buggy): {dict(g)}")
print("Check: AGT should have edge TO GTA (forward direction in sequence)")
````
