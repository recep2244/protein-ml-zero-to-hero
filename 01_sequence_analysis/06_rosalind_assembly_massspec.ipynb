## Interview Tip Bank
> **Eulerian circuit exists iff**: all vertices have equal in-degree and out-degree AND the graph is connected. For circular genome, these should both hold.

> **PDPL backtracking pruning**: Before placing a candidate position x, check that all distances from x to already-placed positions are in the remaining multiset. If any distance is missing, prune immediately.

> **PRSM gotcha**: You're comparing a theoretical spectrum (computed from candidate protein) to an experimental spectrum. Score = number of peaks in common. Normalize by dividing by total spectrum length.

> **Progressive alignment**: "Center star" picks the sequence with max total similarity to all others as the center. Aligns all others to the center. Fast but not optimal.

## What This Notebook Teaches (Plain English)

**Two completely different but equally important problems**:

### Genome Assembly
Sequencing machines read short DNA fragments (150-300 letters). Assembly = puzzle: put millions of fragments back in order to reconstruct the original genome.

**De Bruijn graph approach**: Instead of comparing every pair of fragments (too slow), split each fragment into overlapping k-mers and build a directed graph. Finding an Eulerian path through this graph = the assembled sequence.

### Mass Spectrometry Peptide Sequencing
A mass spectrometer breaks a protein into fragments and weighs each fragment. Given only the mass list, reconstruct the amino acid sequence.

**The algorithm**: Mass differences between sorted fragment masses → amino acids → sequence.

**Prerequisites**: Graph algorithms (directed graphs, DFS/BFS) + combinatorics.

````python
from collections import Counter
def error_correction(reads):
    complement = str.maketrans('ACGT', 'TGCA')
    # Find all k-mers
    all_reads = set()
    for r in reads:
        all_reads.add(r)
        all_reads.add(r.translate(complement)[::-1])
    correct = set()
    for r in all_reads:
        if reads.count(r) > 1 or reads.count(r.translate(complement)[::-1]) > 1:
            correct.add(r)
    corrections = []
    for r in reads:
        rc = r.translate(complement)[::-1]
        if r in correct or rc in correct:
            pass  # correct read
        else:
            # Find correction with Hamming distance 1
            for cand in correct:
                if len(cand) == len(r):
                    diffs = sum(a!=b for a,b in zip(r,cand))
                    if diffs == 1:
                        corrections.append((r, cand)); break
    return corrections
reads = ["TCATC","CCATG","CAATG","TAATG","CAATG"]
for orig, corr in error_correction(reads): print(f"{orig}->{corr}")
````

## PDPL — Creating a Restriction Map

````python
import numpy as np
def local_align(s1, s2, match=2, mismatch=-1, gap=-2):
    m, n = len(s1), len(s2)
    dp = np.zeros((m+1, n+1))
    for i in range(1,m+1):
        for j in range(1,n+1):
            sc = match if s1[i-1]==s2[j-1] else mismatch
            dp[i,j] = max(0, dp[i-1,j-1]+sc, dp[i-1,j]+gap, dp[i,j-1]+gap)
    return dp.max()
print(local_align("ACBDCA","ACBD"))
````

## Problem 1 — E. coli de Novo Assembly with de Bruijn Graph
**Dataset/Source**: E. coli K-12 MG1655 (SRA: SRR2584863 — 1.1M reads, 150 bp Illumina)
**GitHub**: [github.com/rrwick/Unicycler](https://github.com/rrwick/Unicycler) — hybrid assembler
**Skills**: PCOV (de Bruijn graph), Eulerian circuit, N50 quality metric

```python
from collections import defaultdict
import random

def build_debruijn(reads, k):
    graph = defaultdict(list)
    in_deg, out_deg = defaultdict(int), defaultdict(int)
    for read in reads:
        for i in range(len(read) - k):
            u, v = read[i:i+k], read[i+1:i+k+1]
            graph[u].append(v); out_deg[u] += 1; in_deg[v] += 1
    return graph, in_deg, out_deg

def check_eulerian(graph, in_deg, out_deg):
    nodes = set(graph) | set(in_deg)
    imbalanced = [(n, out_deg[n]-in_deg.get(n,0)) for n in nodes
                  if out_deg.get(n,0) != in_deg.get(n,0)]
    return len(imbalanced)==0, imbalanced

def assemble(reads, k=15):
    import copy
    graph, in_deg, out_deg = build_debruijn(reads, k)
    ok, bad = check_eulerian(graph, in_deg, out_deg)
    if not ok:
        print(f'  k={k}: {len(bad)} imbalanced nodes — need more coverage')
        return None
    g = copy.deepcopy(graph)
    start = next(iter(g))
    stack, path = [start], []
    while stack:
        v = stack[-1]
        if g.get(v): stack.append(g[v].pop())
        else: path.append(stack.pop())
    path = path[::-1]
    return path[0] + ''.join(n[-1] for n in path[1:-1])

# Simulate E. coli-like reads
ecoli = ('ATGCGTAACGTCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGA'
         'CGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATC')

random.seed(42)
doubled = ecoli + ecoli
reads = [doubled[i:i+25] for i in range(0, len(ecoli), 3)]
print(f'Reads: {len(reads)}, genome size: {len(ecoli)} bp')

for k in [8, 12, 15]:
    result = assemble(reads, k=k)
    if result:
        ident = sum(a==b for a,b in zip(result,ecoli))/max(len(result),len(ecoli))
        print(f'  k={k}: {len(result)} bp assembled, {ident:.1%} identity')

def compute_n50(contigs):
    lengths = sorted([len(c) for c in contigs], reverse=True)
    cumsum, total = 0, sum(lengths)
    for L in lengths:
        cumsum += L
        if cumsum >= total/2: return L
    return 0

print(f'N50 example: {compute_n50(["A"*300,"A"*200,"A"*150,"A"*100,"A"*80])} bp')
```

## Problem 3 — Circular Plasmid Assembly for Antibiotic Resistance Tracking
**Dataset/Source**: pOXA-48 carbapenem resistance plasmid (NCBI: JN626286.1, 62,009 bp)
**GitHub**: [github.com/rrwick/Bandage](https://github.com/rrwick/Bandage) — assembly graph visualization
**Kaggle**: [Antibiotic Resistance Challenge](https://www.kaggle.com/competitions)
**Skills**: GREP (circular assembly with repeats), Eulerian circuit verification

```python
import copy

def circular_assembly(reads, k):
    graph = defaultdict(list)
    in_deg, out_deg = defaultdict(int), defaultdict(int)
    for read in reads:
        for i in range(len(read) - k):
            u, v = read[i:i+k], read[i+1:i+k+1]
            graph[u].append(v); out_deg[u]+=1; in_deg[v]+=1
    
    # Circular: every node must be balanced
    imbalanced = [(n, out_deg.get(n,0)-in_deg.get(n,0))
                  for n in set(graph)|set(in_deg)
                  if out_deg.get(n,0)!=in_deg.get(n,0)]
    if imbalanced:
        print(f'  {len(imbalanced)} imbalanced nodes — try different k')
        return None
    
    g = copy.deepcopy(graph)
    start = next(iter(g))
    stack, path = [start], []
    while stack:
        v = stack[-1]
        if g.get(v): stack.append(g[v].pop())
        else: path.append(stack.pop())
    path = path[::-1]
    
    # Verify circular: circuit starts and ends at same node
    if path[0] == path[-1]:
        genome = path[0] + ''.join(n[-1] for n in path[1:-1])
        return genome
    return path[0] + ''.join(n[-1] for n in path[1:])

def verify_circular(genome, reads):
    doubled = genome + genome
    return all(r in doubled for r in reads), sum(1 for r in reads if r not in doubled)

# Simulate a resistance plasmid (circular)
plasmid = ('ATGCGATCGATCGATCGATCGATCGATCGATCGATCGATCGCGATCGATCGATCGATCGATCGATCGATCG'
           'ATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGAT')
k = 10
doubled = plasmid + plasmid
reads = [doubled[i:i+18] for i in range(0, len(plasmid), 4)]

print(f'Plasmid: {len(plasmid)} bp (circular), {len(reads)} reads of 18 bp')
assembled = circular_assembly(reads, k=k)
if assembled:
    is_valid, n_bad = verify_circular(assembled, reads)
    print(f'Assembly: {len(assembled)} bp, valid: {is_valid}, unmatched: {n_bad}')
    print(f'Circular match: {plasmid in (assembled+assembled)}')

gc = lambda s: (s.count('G')+s.count('C'))/len(s)
print(f'Plasmid GC: {gc(plasmid):.1%}')
```

## 🐛 Debug Exercise — De Bruijn Graph Bugs

Find and fix **3 bugs** in this genome assembly pipeline.

<details><summary>Show bugs</summary>

**Bug 1:** k-mer extraction: `range(len(seq) - k)` should be `range(len(seq) - k + 1)` — off-by-one misses the last k-mer.

**Bug 2:** De Bruijn graph edge direction: for k-mer `ACGT`, the edge goes from prefix `ACG` to suffix `CGT`. The buggy code reverses this (suffix → prefix).

**Bug 3:** Eulerian path check: a directed graph has an Eulerian path when exactly one node has `out_degree - in_degree = 1` (start) and one has `in_degree - out_degree = 1` (end). The buggy code checks `out == in` for ALL nodes (Eulerian circuit condition, not path).

</details>

````python
from collections import defaultdict

def extract_kmers_buggy(seq, k):
    # BUG 1: off-by-one — misses last k-mer
    return [seq[i:i+k] for i in range(len(seq) - k)]  # should be range(len(seq)-k+1)

def build_debruijn_buggy(kmers):
    graph = defaultdict(list)
    for kmer in kmers:
        # BUG 2: edge direction reversed
        suffix, prefix = kmer[:-1], kmer[1:]  # should be prefix, suffix = kmer[:-1], kmer[1:]
        graph[suffix].append(prefix)           # edge goes wrong way
    return dict(graph)

def has_eulerian_path_buggy(graph):
    in_deg = defaultdict(int)
    out_deg = defaultdict(int)
    for node, neighbors in graph.items():
        out_deg[node] += len(neighbors)
        for n in neighbors:
            in_deg[n] += 1
    all_nodes = set(in_deg) | set(out_deg)
    # BUG 3: checks circuit condition (all balanced), not path condition
    return all(out_deg[n] == in_deg[n] for n in all_nodes)  # wrong for path

seq = "AGTACGT"
k = 3
kmers = extract_kmers_buggy(seq, k)
print(f"k-mers (buggy, k={k}): {kmers}")
print(f"Expected:             {[seq[i:i+k] for i in range(len(seq)-k+1)]}")
print(f"Missing:              {seq[-k:]} (last k-mer)")

g = build_debruijn_buggy(kmers)
print(f"\nDe Bruijn graph (buggy): {dict(g)}")
print("Check: AGT should have edge TO GTA (forward direction in sequence)")
````
