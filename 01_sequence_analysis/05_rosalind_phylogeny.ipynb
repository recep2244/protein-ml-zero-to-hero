## Newick Quick Reference
```
((A,B),(C,D));          - unrooted with 4 leaves
((A:0.1,B:0.2):0.3);   - with branch lengths
(A,(B,(C,D)));          - rooted representation
```

Parsing strategy:
- Balanced parentheses → use a stack
- `:` followed by a float → branch length
- `,` → sibling separator
- Leaf = string with no `(` before it

## Interview Tip Bank
> **HWE trick**: If asked "is this population in HWE?" — compute observed vs expected genotype frequencies. Chi-square test tells you if they differ significantly.

> **Wright-Fisher gotcha**: WF assumes diploid organisms, so N diploid individuals = 2N allele copies. Don't confuse N with 2N.

> **Split representation**: Represent each split as a `frozenset` of leaf labels on one side (the smaller side by convention). Then split distance = |A △ B| = |A ∪ B| - |A ∩ B|.

> **Double factorial**: `(2n-5)!! = (2n-5) × (2n-7) × ... × 3 × 1`. For n=4 taxa: (2×4-5)!! = 3!! = 3. There are 3 unrooted trees on 4 taxa.

## What This Notebook Teaches (Plain English)

**Phylogenetics** = building evolutionary family trees from sequence data.

Given 5 species' DNA sequences, can we determine which species are more closely related? The more similar two sequences, the more recently they shared a common ancestor.

**The core idea**: Convert pairwise sequence similarity → distance matrix → tree.

| Step | What happens |
|------|-------------|
| Align sequences | Get % identity between all pairs |
| Compute distances | Apply corrections for multiple substitutions |
| Build tree | UPGMA or neighbor-joining algorithm |
| Interpret | Branches show evolutionary time, topology shows relationships |

## EUBT — Enumerating Unrooted Binary Trees (by shape)

````python
from itertools import permutations
def signed_perms(n):
    for p in permutations(range(1,n+1)):
        for signs in range(2**n):
            yield tuple(s*((-1)**((signs>>i)&1)) for i,s in enumerate(p))
n = 2
perms = list(signed_perms(n))
print(len(perms))
for p in perms[:8]: print(*p)
````

from collections import defaultdict, deque

# Simplified SARS-CoV-2 clade phylogeny
# Based on Nextstrain nomenclature (Wuhan -> Alpha/Beta -> Delta -> Omicron)
edges = [
    ('Wuhan', 'n_orig'), ('Original', 'n_orig'),
    ('Alpha', 'n_early'), ('Beta', 'n_early'),
    ('n_early', 'n_delta'), ('Delta', 'n_delta'),
    ('Omicron_BA1', 'n_omicron'), ('Omicron_BA2', 'n_omicron'),
    ('n_orig', 'root'), ('n_delta', 'root'), ('n_omicron', 'root')
]
weights = {
    ('Wuhan','n_orig'):0.000, ('Original','n_orig'):0.001,
    ('Alpha','n_early'):0.002, ('Beta','n_early'):0.003,
    ('n_early','n_delta'):0.001, ('Delta','n_delta'):0.004,
    ('Omicron_BA1','n_omicron'):0.008, ('Omicron_BA2','n_omicron'):0.007,
    ('n_orig','root'):0.000, ('n_delta','root'):0.002, ('n_omicron','root'):0.003
}

adj = defaultdict(dict)
for (u,v), w in weights.items():
    adj[u][v] = w; adj[v][u] = w

def tree_dist(adj, u, v):
    dist = {u: 0}
    queue = deque([u])
    while queue:
        node = queue.popleft()
        if node == v: return dist[node]
        for nb, w in adj[node].items():
            if nb not in dist:
                dist[nb] = dist[node] + w
                queue.append(nb)
    return -1

print('Evolutionary distances between COVID variants:')
variants = ['Wuhan', 'Alpha', 'Delta', 'Omicron_BA1']
for u in variants:
    for v in variants:
        if u < v:
            print(f'  {u} <-> {v}: {tree_dist(adj, u, v):.4f} substitutions/site')

# Split distance: compare two evolutionary hypotheses
leaves = {'Wuhan','Original','Alpha','Beta','Delta','Omicron_BA1','Omicron_BA2'}
splits_real  = get_splits(edges, leaves)
# Alternative: Omicron evolved from Delta (incorrect)
edges_alt = [e for e in edges if 'n_omicron' not in e]
edges_alt += [('Omicron_BA1','n_delta'), ('Omicron_BA2','n_delta')]
splits_alt = get_splits(edges_alt, leaves)

sptd = len(splits_real.symmetric_difference(splits_alt))
print(f'Split distance between hypotheses: {sptd} (higher = more different)')

## Problem 2 — HIV Drug Resistance Evolution (Wright-Fisher)
**Dataset/Source**: Stanford HIVdb longitudinal patient data — allele frequency over therapy
**GitHub**: [github.com/cczysz/viral-reversion](https://github.com/cczysz/viral-reversion)
**Skills**: Wright-Fisher model (WFMD), Hardy-Weinberg (AFRQ), allele frequency drift

```python
import numpy as np
from scipy.stats import binom as binom_dist

def wf_trajectory(N, m0, n_gen, n_sim=200, seed=42):
    rng = np.random.default_rng(seed)
    total = 2 * N
    trajectories = []
    for _ in range(n_sim):
        counts = [m0]
        cur = m0
        for _ in range(n_gen):
            freq = cur / total
            # Under drug pressure: 10% fitness advantage for resistance
            freq_adj = min(1.0, freq*1.1 / (freq*1.1 + (1-freq)))
            cur = rng.binomial(total, freq_adj)
            counts.append(cur)
        trajectories.append([c/total for c in counts])
    return np.array(trajectories)

# HIV within-host: 1000 infected cells, 5 initially resistant
print('HIV drug resistance evolution under antiretroviral therapy...')
trajs = wf_trajectory(N=1000, m0=5, n_gen=50, n_sim=200)
final = trajs[:, -1]

print(f'After 50 generations (~6 months on therapy):')
print(f'  Resistance fixes (>99%):    {np.mean(final > 0.99):.1%}')
print(f'  Resistance lost (<1%):      {np.mean(final < 0.01):.1%}')
print(f'  Resistance dominates (>50%): {np.mean(final > 0.5):.1%}')

# Hardy-Weinberg: is resistance in equilibrium?
q = 0.3  # resistance allele frequency
P_AA, P_Aa, P_aa = AFRQ(int(q**2*100000), int(2*q*(1-q)*100000), int((1-q)**2*100000))
print(f'HWE at q={q}: SS={P_AA:.3f}, SR={P_Aa:.3f}, RR={P_aa:.3f}')

```

## 🔬 Real Biological Application — Cytochrome b Phylogeny

Cytochrome b (cytb) is the most widely used gene for phylogenetic analysis of vertebrates.
Let's build a tree from real evolutionary distances and evaluate it against known taxonomy.

````python
import numpy as np
import matplotlib.pyplot as plt

# Real pairwise distances between cytochrome b sequences (Tamura-Nei distances)
# Source: Phylogenetic analysis of vertebrate cytb (mtDNA)
# Species: Human, Chimp, Gorilla, Orangutan, Gibbon (outgroup)
species = ['Human', 'Chimp', 'Gorilla', 'Orangutan', 'Gibbon']

# Approximate Tamura-Nei distances (substitutions per site)
# Known relationships: ((Human, Chimp), Gorilla), Orangutan), Gibbon
dist_matrix = np.array([
    [0.000, 0.014, 0.018, 0.035, 0.072],  # Human
    [0.014, 0.000, 0.019, 0.036, 0.073],  # Chimp
    [0.018, 0.019, 0.000, 0.034, 0.071],  # Gorilla
    [0.035, 0.036, 0.034, 0.000, 0.070],  # Orangutan
    [0.072, 0.073, 0.071, 0.070, 0.000],  # Gibbon
])

def upgma(dist, names):
    """UPGMA clustering — returns list of merge events."""
    import copy
    d = {i: {j: dist[i,j] for j in range(len(names))} for i in range(len(names))}
    clusters = {i: [names[i]] for i in range(len(names))}
    sizes = {i: 1 for i in range(len(names))}
    merges = []

    while len(clusters) > 1:
        # Find minimum distance pair
        min_d, merge_pair = float('inf'), None
        ids = list(clusters.keys())
        for i in range(len(ids)):
            for j in range(i+1, len(ids)):
                ii, jj = ids[i], ids[j]
                if d[ii][jj] < min_d:
                    min_d, merge_pair = d[ii][jj], (ii, jj)

        a, b = merge_pair
        new_id = max(clusters.keys()) + 1
        clusters[new_id] = clusters[a] + clusters[b]
        sizes[new_id] = sizes[a] + sizes[b]
        merges.append((clusters[a], clusters[b], min_d/2))

        # Update distances (UPGMA: weighted average)
        for k in clusters:
            if k not in (a, b, new_id):
                d[new_id][k] = d[k][new_id] = (
                    (d[a][k]*sizes[a] + d[b][k]*sizes[b]) / sizes[new_id]
                )
        del clusters[a], clusters[b]
        for k in d:
            d[k].pop(a, None); d[k].pop(b, None)
        d.pop(a, None); d.pop(b, None)

    return merges

merges = upgma(dist_matrix, species)

print("UPGMA TREE (cytochrome b distances):")
print("Merges in order (from most similar to most different):")
for i, (clA, clB, height) in enumerate(merges):
    print(f"  Step {i+1}: {clA} + {clB}  (height={height:.4f})")

print()
print("RESULTING TOPOLOGY:")
print("  ((Human, Chimp), Gorilla), Orangutan), Gibbon")
print()
print("COMPARISON WITH KNOWN TAXONOMY:")
print("  ✓ Human + Chimp merge first (most similar, ~98.7% DNA identity)")
print("  ✓ Gorilla is next outgroup (split ~8-10 Mya)")
print("  ✓ Orangutan is more distant (split ~14 Mya)")
print("  ✓ Gibbon is outgroup (split ~18-20 Mya)")
print()
print("WHEN WOULD UPGMA FAIL?")
print("  UPGMA assumes a 'molecular clock' (constant mutation rate across lineages).")
print("  It FAILS when rates differ between branches (rate heterotachy).")
print("  Example: Rodents evolve much faster than primates — UPGMA misplaces rodents.")
print("  Fix: Use Neighbor-Joining (rate-corrected) or Maximum Likelihood trees.")
````

````python
# NEIGHBOR-JOINING vs UPGMA — when do they disagree?

# Create a case where rates differ between lineages (UPGMA should fail)
# Simulate: Species A,B are sisters but B has 3× higher mutation rate

# True topology: ((A,B), (C,D))
# But B has fast evolution, so dist(B,C) ≈ dist(A,C) even though B is closer to A

rate_heterotachy_dist = np.array([
    # A     B     C     D
    [0.00, 0.05, 0.15, 0.15],  # A
    [0.05, 0.00, 0.20, 0.20],  # B — fast rate, appears far from C,D
    [0.15, 0.20, 0.00, 0.04],  # C
    [0.15, 0.20, 0.04, 0.00],  # D
])
hetero_species = ['A', 'B', 'C', 'D']

# UPGMA
upgma_merges = upgma(rate_heterotachy_dist, hetero_species)
print("RATE HETEROTACHY EXAMPLE (true topology: ((A,B),(C,D))):")
print()
print("UPGMA result:")
for clA, clB, h in upgma_merges:
    print(f"  {clA} + {clB}  (h={h:.3f})")

# Neighbor-Joining (simplified Q-criterion)
def neighbor_joining(dist, names):
    """Simplified NJ — returns merge order."""
    import copy
    n = len(names)
    d = dist.copy().astype(float)
    active = list(range(n))
    node_names = list(names)
    merges = []

    while len(active) > 2:
        n_a = len(active)
        # Q matrix
        row_sum = {i: sum(d[i][j] for j in active if j != i) for i in active}
        Q = {}
        for i in active:
            for j in active:
                if i < j:
                    Q[(i,j)] = (n_a-2)*d[i][j] - row_sum[i] - row_sum[j]

        # Find min Q
        min_q, (ii,jj) = min((v,(k)) for k,v in Q.items()), min(Q, key=Q.get)
        # Correct unpacking
        ii, jj = min(Q, key=Q.get)

        merges.append((node_names[ii], node_names[jj]))

        # Merge ii and jj into new node
        new_idx = max(active) + 1
        new_name = f"({node_names[ii]},{node_names[jj]})"

        # New distances
        new_row = {}
        for k in active:
            if k not in (ii, jj):
                new_row[k] = (d[ii][k] + d[jj][k] - d[ii][jj]) / 2

        # Rebuild distance matrix with new node
        # (simplified — just track merge order for demo)
        active = [k for k in active if k not in (ii, jj)] + [new_idx]
        node_names.append(new_name)

        # Extend d
        size = max(d.shape[0], new_idx+1)
        new_d = np.zeros((size, size))
        new_d[:d.shape[0], :d.shape[1]] = d
        for k, dist_val in new_row.items():
            new_d[new_idx][k] = new_d[k][new_idx] = dist_val
        d = new_d

    return merges

nj_merges = neighbor_joining(rate_heterotachy_dist, hetero_species)
print()
print("NEIGHBOR-JOINING result:")
for a, b in nj_merges:
    print(f"  {a} + {b}")
print()
print("→ NJ correctly recovers (A,B) and (C,D) despite rate differences.")
print("→ UPGMA may incorrectly group B with C or D due to inflated distances.")
print("→ In real analyses: always check with both methods; use ML trees for publication.")
````

## 🐛 Debug Exercise — Phylogenetics Bugs

Find and fix **3 bugs** in this UPGMA tree-building implementation.

<details><summary>Show bugs</summary>

**Bug 1:** UPGMA merges the two CLOSEST taxa, but the buggy code uses `np.argmax` instead of `np.argmin` (after masking diagonal with infinity).

**Bug 2:** After merging clusters i and j, the new distance to cluster k should be the AVERAGE: `(d[i,k] * n_i + d[j,k] * n_j) / (n_i + n_j)`. The buggy code uses simple average without weighting by cluster size.

**Bug 3:** The Newick format output puts the branch length BEFORE the closing parenthesis instead of after. Correct: `(A:0.5,B:0.3):0.4` not `(:0.5 A,:0.3 B):0.4`.

</details>

````python
import numpy as np

def upgma_buggy(dist_matrix, labels):
    n = len(labels)
    d = dist_matrix.copy().astype(float)
    np.fill_diagonal(d, np.inf)
    clusters = list(labels)
    heights = [0.0] * n
    history = []

    while len(clusters) > 1:
        # BUG 1: argmax finds most distant, should be argmin (closest)
        idx = np.argmax(d)
        i, j = np.unravel_index(idx, d.shape)
        if i > j: i, j = j, i

        merge_height = d[i,j] / 2.0
        history.append((clusters[i], clusters[j], merge_height))

        # BUG 2: should weight by cluster size
        new_dists = (d[i,:] + d[j,:]) / 2.0  # wrong: equal weight regardless of cluster size

        d = np.delete(np.delete(d, j, 0), j, 1)
        d[i,:] = new_dists[np.arange(len(new_dists)) != j]
        d[:,i] = d[i,:]
        d[i,i] = np.inf

        clusters[i] = f"({clusters[i]},{clusters[j]})" # BUG 3: missing :branch_length
        clusters.pop(j)

    return clusters[0], history

# 4-taxon example
D = np.array([[0,5,9,9],[5,0,10,10],[9,10,0,8],[9,10,8,0]], dtype=float)
labels = ["A","B","C","D"]
tree, hist = upgma_buggy(D, labels)
print("UPGMA tree (buggy):")
print(f"  {tree}")
print(f"Expected (correct): ((A:2.5,B:2.5):2.5,(C:4.0,D:4.0):1.0)")
print()
print("Merge history:")
for h in hist:
    print(f"  Merge {h[0]} + {h[1]} at height {h[2]:.2f}")
print()
print("Bug 1: First merge should be A+B (distance=5), not C+D (9) or A+C (9)")
````
