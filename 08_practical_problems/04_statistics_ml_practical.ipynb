```python
# Statistics Section 1 — Descriptive stats + Rosalind PROB
import numpy as np
from scipy import stats

# --- HackerRank 10 Days: Day 0-2 ---
data = [64630, 11735, 14216, 99233, 14470, 4978, 73429, 38120, 51135, 67060]
print(f"Mean:   {np.mean(data):.1f}")
print(f"Median: {np.median(data):.1f}")
print(f"Mode:   {stats.mode(data, keepdims=True).mode[0]}")
print(f"Std:    {np.std(data):.2f}")
print(f"Var:    {np.var(data):.2f}")
print()
# Quartiles and IQR
q1, q2, q3 = np.percentile(data, [25, 50, 75])
print(f"Q1={q1}, Q2={q2}, Q3={q3}, IQR={q3-q1}")

# --- Rosalind PROB ---
def prob_substring(s, gc):
    """Log prob of string s under random model with given GC content."""
    import math
    log_prob = 0
    for c in s:
        if c in 'GC':
            log_prob += math.log10(gc/2)
        else:
            log_prob += math.log10((1-gc)/2)
    return log_prob

s = "ACGATACAA"
gc_values = [0.129, 0.287, 0.423, 0.476, 0.641, 0.762, 0.951]
probs = [prob_substring(s, gc) for gc in gc_values]
print("\nRosalind PROB:")
print("GC:", gc_values)
print("Log prob:", [f"{p:.3f}" for p in probs])
```

```python
# Statistics Section 2 — Distributions + Hardy-Weinberg
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# --- HackerRank Day 3-4: Distributions ---
# Binomial
n, p = 10, 0.3
print(f"Binomial B(10,0.3):")
print(f"  P(X=3) = {stats.binom.pmf(3,n,p):.4f}")
print(f"  P(X<=3) = {stats.binom.cdf(3,n,p):.4f}")

# Poisson
lam = 2.5
print(f"\nPoisson(λ=2.5):")
print(f"  P(X=3) = {stats.poisson.pmf(3,lam):.4f}")

# Normal
mu, sigma = 70, 10
print(f"\nNormal(μ=70, σ=10):")
print(f"  P(X>80) = {1-stats.norm.cdf(80,mu,sigma):.4f}")
print(f"  P(60<X<80) = {stats.norm.cdf(80,mu,sigma)-stats.norm.cdf(60,mu,sigma):.4f}")

# --- Rosalind IEV (independent alleles / Mendel) ---
def mendel_prob(k, m, n):
    """Prob that dominant allele appears in offspring."""
    total = k+m+n
    pairs = total*(total-1)
    # AA, Aa, aa: dominant prob = 1, 3/4, 1/4, 0 respectively
    dominant_pairs = (k*k + k*m + k*n +      # AA × anything: always dominant
                      m*k +                   # Aa × AA
                      m*m*0.75 +              # Aa × Aa: 3/4 dominant
                      m*n*0.5 +               # Aa × aa: 1/2 dominant
                      n*k + n*m*0.5)          # aa × AA or Aa
    return 2*dominant_pairs / pairs

print(f"\nRosalind MENDL prob(k=2,m=2,n=2) = {mendel_prob(2,2,2):.5f}")
```

```python
# Statistics Section 3 — Regression + Correlation
import numpy as np
from scipy import stats

# --- HackerRank Day 6-7: Regression ---
X = np.array([95,85,80,70,60,50,40,30,20,10], dtype=float)
Y = np.array([85,75,70,55,50,40,30,25,20,10], dtype=float)

# Pearson correlation
r, p_corr = stats.pearsonr(X, Y)
print(f"Pearson r = {r:.4f}, p = {p_corr:.4e}")

# Linear regression
slope, intercept, r_val, p_val, stderr = stats.linregress(X, Y)
print(f"y = {slope:.4f}x + {intercept:.4f}")
print(f"R² = {r_val**2:.4f}")

# Predict
for x_new in [57, 66, 70]:
    y_pred = slope * x_new + intercept
    print(f"  x={x_new} → predicted y={y_pred:.2f}")

# Multiple regression with numpy
X_mat = np.column_stack([np.ones(len(X)), X])
beta = np.linalg.lstsq(X_mat, Y, rcond=None)[0]
print(f"\nLeast squares: intercept={beta[0]:.4f}, slope={beta[1]:.4f}")
```

```python
# Statistics Section 4 — Hypothesis testing + WFMD
import numpy as np
from scipy import stats

# --- HackerRank Day 8-9: Hypothesis testing ---
# One-sample t-test
sample = [30.02,29.99,30.11,29.97,30.01,29.99]
t_stat, p_val = stats.ttest_1samp(sample, popmean=30)
print(f"One-sample t-test: t={t_stat:.4f}, p={p_val:.4f}")
print(f"Reject H0 (μ=30) at α=0.05? {'Yes' if p_val < 0.05 else 'No'}")

# Two-sample t-test
group_a = [21, 22, 19, 20, 21, 22, 20, 19, 21, 22]
group_b = [24, 25, 23, 24, 25, 24, 23, 25, 24, 23]
t2, p2 = stats.ttest_ind(group_a, group_b)
print(f"\nTwo-sample t-test: t={t2:.4f}, p={p2:.4f}")

# Chi-squared test
observed = [90, 60, 104, 95]
expected = [87.5, 62.5, 106.25, 93.75]
chi2, p_chi = stats.chisquare(observed, expected)
print(f"\nChi² test: χ²={chi2:.4f}, p={p_chi:.4f}")

# --- Rosalind WFMD (Wright-Fisher) ---
def wright_fisher_drift(N, p0, generations=10):
    """Simulate allele frequency drift in finite population."""
    p = p0
    trajectory = [p]
    for _ in range(generations):
        # Sample 2N alleles from binomial
        count = np.random.binomial(2*N, p)
        p = count / (2*N)
        trajectory.append(p)
    return trajectory

np.random.seed(42)
traj = wright_fisher_drift(100, 0.5, 20)
print(f"\nWright-Fisher (N=100, p0=0.5): final p = {traj[-1]:.3f}")
```

```python
# Statistics Section 5 — ML pipeline + feature selection
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.datasets import make_classification
import numpy as np

# --- HackerRank ML certification pipeline ---
X, y = make_classification(n_samples=500, n_features=20, n_informative=10,
                            random_state=42)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(max_iter=1000))
])

# Cross-validation
cv_scores = cross_val_score(pipeline, X, y, cv=5, scoring='accuracy')
print(f"CV accuracy: {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")

# Grid search
param_grid = {'clf__C': [0.01, 0.1, 1.0, 10.0]}
gs = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')
gs.fit(X, y)
print(f"Best C: {gs.best_params_['clf__C']}")
print(f"Best CV accuracy: {gs.best_score_:.3f}")

# Feature importance
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(f_classif, k=10)
X_sel = selector.fit_transform(X, y)
selected = np.where(selector.get_support())[0]
print(f"Top features selected: {selected[:5]}")
```

```python
# Statistics Section 6 — Population genetics + real-world
import numpy as np

# --- Rosalind LIA (independent alleles) ---
def prob_gen_dominant(k, N):
    """Prob that at least N individuals in gen k are AaBb.
    In generation k (starting from Aa Bb): P(AaBb) = (1/2)^(2k) ... wait:
    Actually, selfing AaBb: P(AaBb offspring) = 1/2.
    Gen k: prob individual is AaBb = (1/2)^(2*(k-1)) ... use the actual formula.
    Simpler: each gen, P(AaBb) = 1/4 (from Aa×Aa and Bb×Bb).
    """
    from math import comb
    # After k generations of self-crossing, population size = 2^k
    # P(offspring = AaBb) = (1/2)^2 = 1/4 for each individual
    pop_size = 2**k
    p = 0.25  # probability each individual is AaBb
    # P(at least N are AaBb) = sum P(X=j) for j>=N
    from scipy import stats
    prob = 1 - stats.binom.cdf(N-1, pop_size, p)
    return prob

print(f"P(≥1 AaBb in gen 2) = {prob_gen_dominant(2,1):.3f}")
print(f"P(≥1 AaBb in gen 3) = {prob_gen_dominant(3,1):.3f}")

# Genome-wide association study power
import numpy as np
np.random.seed(42)
n_variants = 1000000
n_cases, n_controls = 5000, 5000
# Effect size: OR=1.3 for 10 causal variants
p_values = np.random.uniform(0, 1, n_variants)
p_values[:10] = np.random.beta(0.1, 10, 10)  # significant variants
bonferroni = 0.05 / n_variants
n_significant = (p_values < bonferroni).sum()
print(f"\nGWAS: {n_significant} variants pass Bonferroni correction")
print(f"Bonferroni threshold: p < {bonferroni:.2e}")
```

## Interview Cheat Sheet — Statistics & ML

| Concept | Formula | When to use |
|---------|---------|------------|
| Pearson r | `sum((x-xbar)(y-ybar)) / (n*sx*sy)` | Linear relationships |
| Spearman r | `1 - 6*sum(d²)/(n*(n²-1))` | Monotonic, outlier-robust |
| OLS slope | `sum((xi-xbar)(yi-ybar)) / sum((xi-xbar)²)` | Simple linear regression |
| t-test | `t = (x1bar - x2bar) / sqrt(s1²/n1 + s2²/n2)` | Compare two group means |
| Chi-squared | `sum((O-E)²/E)` | Categorical independence |
| Binomial P(k) | `C(n,k) * p^k * (1-p)^(n-k)` | k successes in n trials |
| Normal P(X<x) | `0.5*(1 + erf((x-mu)/(sigma*sqrt(2))))` | Continuous probability |

## Interview Q&A — Statistics & ML

**Q: What is the difference between Type I and Type II error?**
A: Type I (false positive): reject H₀ when it's true (α = significance level, typically 0.05). Type II (false negative): fail to reject H₀ when it's false (β). Power = 1-β. In drug discovery: Type I = approve ineffective drug; Type II = reject effective drug. Trade-off: lowering α increases β.

**Q: What does AUC-ROC measure and when is it misleading?**
A: AUC-ROC measures a classifier's ability to discriminate at ALL thresholds (area under the receiver operating characteristic curve). Value 0.5 = random, 1.0 = perfect. Misleading when: class imbalance (99% negative) — a model predicting all negative gets AUC=0.5 but accuracy=99%. Use AUC-PR (precision-recall) for imbalanced data.

**Q: Explain Pearson vs Spearman correlation.**
A: Pearson measures linear correlation between two variables, sensitive to outliers. Spearman measures rank correlation — monotonic relationships, robust to outliers. For protein stability data (heavy-tailed distributions): Spearman is standard (ProteinGym uses Spearman ρ).

**Q: What is the central limit theorem and why does it matter for ML?**
A: CLT: the mean of n independent samples approaches a normal distribution as n→∞, regardless of the original distribution. Why it matters: (1) justifies using t-tests even for non-normal data (large n), (2) explains why batch statistics in BatchNorm are well-behaved, (3) underpins confidence intervals on model performance metrics.
