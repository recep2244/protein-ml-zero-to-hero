## DNA Strings & Algorithms — Concepts for Beginners

| Term | Plain English |
|------|---------------|
| **string algorithm** | Algorithm that operates on sequences of characters — DNA is just a very long string |
| **substring** | A contiguous part of a string — finding motifs is a substring search problem |
| **complement** | Replace each base: A<->T, G<->C — the other strand of double-stranded DNA |
| **reverse complement** | Reverse the complement — how the anti-parallel strand reads 5'->3' |
| **Hamming distance** | Number of positions where two strings of equal length differ |
| **edit distance** | Minimum insertions + deletions + substitutions to transform string A into B |
| **k-mer** | A substring of length k — counting k-mers is a key technique in sequence analysis |
| **frequency table** | Count of how often each substring/k-mer appears in a sequence |
| **string hashing** | Convert a string to a number for fast lookup — Rabin-Karp uses this for pattern matching |
| **trie** | Tree data structure for fast string prefix search — used in genome assembly |
| **suffix array** | Sorted list of all suffixes of a string — enables O(log n) pattern search |
| **BWT (Burrows-Wheeler Transform)** | String permutation that groups similar characters — base of most genome aligners |

````python
# Section 2 — GC content + regex + itertools
import re
from itertools import combinations, permutations
from collections import Counter

# --- Rosalind GC ---
def gc_content(seq):
    return (seq.count('G') + seq.count('C')) / len(seq)

seqs = {
    'Rosalind_6404': 'CCTGCGGAAGATCGGCACTAGAATAGCCAGAACCGTTTCTCTGAGGCTTCCGGCCTTCCCTCCCACTAATAATTCTGAGG',
    'Rosalind_5959': 'CCATCGGTAGCGCATCCTTAGTCCAATTAAGTCCCTATCCAGGCGCTCCGCCGAAGGTCTATATCCATTTGTCAGCAGACACGC',
}
best = max(seqs, key=lambda k: gc_content(seqs[k]))
print(f"Highest GC: {best} ({gc_content(seqs[best]):.4%})")

# --- Rosalind SUBS (motif finding) ---
def find_motif(dna, motif):
    return [i+1 for i in range(len(dna)-len(motif)+1) if dna[i:i+len(motif)]==motif]

positions = find_motif("GATATATGCATATACTT", "ATAT")
print("Motif positions:", positions)

# --- itertools ---
print("2-combinations of ACGT:", list(combinations('ACGT', 2))[:4], "...")
print("Permutations of 123:", list(permutations([1,2,3])))
````

````python
# Section 3 — Protein translation + ORF finding
from Bio.Seq import Seq

# --- Rosalind PROT ---
mrna = "AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA"
protein = Seq(mrna).translate(to_stop=True)
print("Protein:", protein)

# --- Rosalind ORF ---
def find_orfs(dna):
    seq = Seq(dna)
    orfs = set()
    for strand, s in [(1, seq), (-1, seq.reverse_complement())]:
        for frame in range(3):
            trans = str(s[frame:].translate())
            # Find all start-stop pairs
            aa_seq = trans
            i = 0
            while i < len(aa_seq):
                if aa_seq[i] == 'M':
                    j = aa_seq.find('*', i)
                    if j != -1:
                        orfs.add(aa_seq[i:j])
                i += 1
    return orfs

dna = "AGCCATGTAGCTAACTCAGGTTACATGGGGATGACCCCGCGACTTGGATTAGAGTCTCTTTTGGAATAAGCCTGAATGGCGAATGGACGCGCCCTGTAGCGGCGCATTAAGCGCGGCGGGTGTGGTGGTTACGCGCAGCGTGACCGCTACACTTGCCAGCGCCCTAGCGCCCGCTCCTTTCGCTTTCTTCCCTTCCTTTCTCGCCACGTTCGCCGGCTTTCCCCGTCAAGCTCTAAATCGGGGGCTCCCTTTAGGGTTCCGATTTAGTGCTTTACGGCACCTCGACCCCAAAAAACTTGATTAGGGTGATGGTTCACGTAGTGGGCCATCGCCCTGATAGACGGTTTTTCGCCCTTTGACGTTGGAGTCCACGTTCTTTAATAGTGGACTCTTGTTCCAAACTGGAACAACACTCAACCCTATCTCGGTCTATTCTTTTGATTTATAAGGGATTTTGCCGATTTCGGCCTATTGGTTAAAAAATGAGCTGATTTAACAAAAATTTAACGCGAATTTTAACAAAATATTAACGTTTACAATTTCAGGTGGCACTTTTCGGGGAAATGTGCGCGGAACCCCTATTTGTTTATTTTTCTAAATACATTCAAATATGTATCCGCTCATGAGACAATAACCCTGATAAATGCTTCAATAATATATGTCCACTCTTACAAGATGTCATAAATTTTTGTTCCCGGCCATCCACACAAAATGAAATCAATTTTTTTCACTGAAACAATGATCCTTTAAACATTTAAATAAAATTTTATTTTTAATTTCAAGAGACTTGCTTTTTTGATTTACCCTAAGTTAGACAAATTATTTTTAAATGTTGCTCCAGACCTAAGATCTTTTAATTTATTCCCAGTCTGAAACAGCAGCCTTTATTTTTTCACCTTCCTAAAGCAGCCCTGCCAAGTCCAATGCCAGGTTTACATGCAACACGTATGTTTTCATCATCGGCCCCGTCGTTGCAAAC")
orfs = find_orfs(dna)
print(f"ORFs found: {len(orfs)}")
for orf in sorted(orfs)[:5]:
    print(f"  {orf}")
````

````python
# Section 4 — k-mer frequency + consensus
from collections import Counter
import numpy as np

# --- k-mer counting ---
def count_kmers(seq, k):
    return Counter(seq[i:i+k] for i in range(len(seq)-k+1))

dna = "ACGTTGCATGTCGCATGATGCATGAGAGCT"
kmers = count_kmers(dna, 4)
print("Most common 4-mers:", kmers.most_common(5))

# --- Rosalind CONS (consensus matrix + string) ---
def consensus(seqs):
    L = len(seqs[0])
    n = len(seqs)
    matrix = {b: np.zeros(L, dtype=int) for b in 'ACGT'}
    for seq in seqs:
        for i, b in enumerate(seq):
            if b in matrix:
                matrix[b][i] += 1
    consensus_str = ''.join(max('ACGT', key=lambda b: matrix[b][i]) for i in range(L))
    return consensus_str, matrix

seqs = ["ATCCAGCT","GGGCAACT","ATGGATCT","AAGCAACC","TTGGAACT","ATGCCATT","ATGGCACT"]
cons, mat = consensus(seqs)
print("Consensus:", cons)
for b in 'ACGT':
    print(f"{b}: {' '.join(map(str, mat[b]))}")
````

````python
# Section 5 — RNA splicing + HackerRank string tasks
from Bio.Seq import Seq

# --- Rosalind SPLC (RNA splicing) ---
def rna_splice(dna, introns):
    """Remove introns from DNA, translate to protein."""
    mrna = dna
    for intron in introns:
        mrna = mrna.replace(intron, '')
    return str(Seq(mrna).translate(to_stop=True))

dna_str  = "ATGGTCTACATAGCTGACAAACAGCACGTAGCAATCGGTCGAATCTCGAGAGGCATATGGTCACATGATCGGTCGAGCTTCACTTTTAGTCAGATAA"
introns  = ["ATCGGTCGAA","ATCGGTCGAG"]
protein  = rna_splice(dna_str, introns)
print("Spliced protein:", protein)

# --- HackerRank: Anagram detection ---
from collections import Counter
def are_anagrams(s1, s2):
    return Counter(s1.lower()) == Counter(s2.lower())

pairs = [("listen","silent"),("hello","world"),("race","care")]
for a, b in pairs:
    print(f"'{a}' and '{b}': {'anagram' if are_anagrams(a,b) else 'not anagram'}")

# --- HackerRank: String formatting ---
def formatted_table(n):
    for i in range(1, n+1):
        print(f"{i:>{len(str(n))}} {bin(i):>{len(bin(n))}} {oct(i):>{len(oct(n))}} {hex(i):>{len(hex(n))}}")
formatted_table(5)
````

````python
# Section 6 — Trie + KMP pattern matching
from collections import defaultdict

# --- Trie for restriction site detection ---
class Trie:
    def __init__(self):
        self.root = {}
        self.end = '#'

    def insert(self, word):
        node = self.root
        for ch in word:
            node = node.setdefault(ch, {})
        node[self.end] = True

    def search(self, text):
        matches = []
        for start in range(len(text)):
            node = self.root
            for i, ch in enumerate(text[start:]):
                if ch not in node:
                    break
                node = node[ch]
                if self.end in node:
                    matches.append((start+1, text[start:start+i+1]))
        return matches

trie = Trie()
patterns = ["EcoRI:GAATTC", "BamHI:GGATCC", "HindIII:AAGCTT"]
restriction_sites = [p.split(':')[1] for p in patterns]
for site in restriction_sites:
    trie.insert(site)

genome = "ATGAATTCGATGGATCCATGAAGCTTAATG"
matches = trie.search(genome)
print("Restriction sites found:")
for pos, seq in matches:
    enzyme = next(p.split(':')[0] for p in patterns if p.split(':')[1]==seq)
    print(f"  {enzyme} ({seq}) at position {pos}")

# --- KMP ---
def kmp_search(text, pattern):
    def build_lps(p):
        lps = [0] * len(p)
        length = 0; i = 1
        while i < len(p):
            if p[i] == p[length]:
                length += 1; lps[i] = length; i += 1
            elif length:
                length = lps[length-1]
            else:
                lps[i] = 0; i += 1
        return lps
    lps = build_lps(pattern)
    matches = []; i = j = 0
    while i < len(text):
        if text[i] == pattern[j]:
            i += 1; j += 1
        if j == len(pattern):
            matches.append(i-j); j = lps[j-1]
        elif i < len(text) and text[i] != pattern[j]:
            j = lps[j-1] if j else (i := i+1) or 0
    return matches

print("KMP:", kmp_search("ATATATATATA","ATAT"))
````

````python
# Section 7 — Advanced string problems
import re
import numpy as np

# --- Longest palindromic substring ---
def longest_palindrome(s):
    def expand(s, l, r):
        while l>=0 and r<len(s) and s[l]==s[r]:
            l -= 1; r += 1
        return s[l+1:r]
    best = ''
    for i in range(len(s)):
        for p in [expand(s,i,i), expand(s,i,i+1)]:
            if len(p) > len(best):
                best = p
    return best

print("Longest palindrome in 'ACGTGCA':", longest_palindrome('ACGTGCA'))
print("Longest palindrome in 'GCATGCAATGCAGCTAGCATGCATGCA':", longest_palindrome('GCATGCAATGCAGCTAGCATGCATGCA'))

# --- Z-algorithm for string matching ---
def z_algorithm(s):
    n = len(s)
    z = [0] * n
    z[0] = n
    l = r = 0
    for i in range(1, n):
        if i < r:
            z[i] = min(r-i, z[i-l])
        while i+z[i] < n and s[z[i]] == s[i+z[i]]:
            z[i] += 1
        if i + z[i] > r:
            l, r = i, i+z[i]
    return z

def z_search(text, pattern):
    s = pattern + '$' + text
    z = z_algorithm(s)
    p = len(pattern)
    return [i - p - 1 for i in range(p+1, len(s)) if z[i] == p]

print("Z-search 'ATG' in genome:", z_search("ATGATGATG", "ATG"))
````

## Interview Q&A — Strings & DNA Algorithms

**Q: What is the time complexity of naive string search and how does KMP improve it?**
A: Naive: O(n·m) — for each position in text (n), try matching pattern (m). KMP (Knuth-Morris-Pratt): O(n+m) — precompute a "failure function" (partial match table) that tells you how far to skip when a mismatch occurs. For bioinformatics: searching a 3GB genome for a 20bp motif, KMP is essential.

**Q: Given a DNA string, write a one-liner for reverse complement.**
A: `rc = seq.translate(str.maketrans("ACGT", "TGCA"))[::-1]`
Or: `''.join({'A':'T','T':'A','C':'G','G':'C'}[b] for b in reversed(seq))`

**Q: What data structure would you use to count k-mer frequencies efficiently?**
A: `collections.Counter({seq[i:i+k]: for i in range(len(seq)-k+1)})` — O(n) time, O(4^k) space. For large genomes, use a hash map (dict) or a space-efficient structure like a Bloom filter for membership queries.

**Q: Explain the difference between find() and index() in Python strings.**
A: `find()` returns -1 if not found; `index()` raises `ValueError`. For bioinformatics: use `find()` in loops (safer), `index()` when you're certain the pattern exists.

**Q: What is the purpose of `re.findall()` vs `re.finditer()` for motif finding?**
A: `findall()` returns a list of all match strings — loads all into memory. `finditer()` returns a lazy iterator of Match objects (includes position info). For genomics: always use `finditer()` — get positions and don't load all matches at once. `m.start()` gives match position.

````python
# TIMED PRACTICE INFRASTRUCTURE
# Use this pattern for every problem in this module

import time
import sys

class TimedProblem:
    """
    A timed problem container.
    Call .start() to begin timing, .reveal() to see the solution.
    """
    def __init__(self, title, problem, solution, hints=None, time_limit=10):
        self.title = title
        self.problem = problem
        self.solution = solution
        self.hints = hints or []
        self.time_limit = time_limit  # minutes
        self._start_time = None

    def start(self):
        self._start_time = time.time()
        print(f"{'='*60}")
        print(f"PROBLEM: {self.title}")
        print(f"TIME LIMIT: {self.time_limit} minutes")
        print(f"{'='*60}")
        print(self.problem)
        print(f"\n[Write your solution in the cell below, then call .check_time() and .reveal()]")

    def check_time(self):
        if self._start_time is None:
            print("Call .start() first")
            return
        elapsed = (time.time() - self._start_time) / 60
        remaining = self.time_limit - elapsed
        if remaining > 0:
            print(f"Time elapsed: {elapsed:.1f}m | Remaining: {remaining:.1f}m")
        else:
            print(f"⏰ TIME UP! Elapsed: {elapsed:.1f}m (limit: {self.time_limit}m)")

    def hint(self, n=1):
        if n <= len(self.hints):
            print(f"HINT {n}: {self.hints[n-1]}")
        else:
            print("No more hints available.")

    def reveal(self):
        if self._start_time:
            elapsed = (time.time() - self._start_time) / 60
            print(f"\n[Revealed after {elapsed:.1f} minutes]")
        print("\n--- SOLUTION ---")
        exec(self.solution, {'__builtins__': __builtins__})
        print("--- END SOLUTION ---")

# ─── PROBLEM SET: 3 difficulty levels for String/DNA problems ───

# LEVEL 1 — WARM-UP (2 min)
p1 = TimedProblem(
    title="[EASY] DNA Complement",
    problem="""
Write a function complement(dna) that returns the complement strand
(not reversed) of a DNA string.
A↔T, G↔C

Example:
  complement("ATGC") → "TACG"
  complement("GATTACA") → "CTAATGT"
""",
    solution="""
def complement(dna):
    table = str.maketrans('ATGC', 'TACG')
    return dna.translate(table)

print(complement("ATGC"))     # TACG
print(complement("GATTACA"))  # CTAATGT
""",
    hints=["Use str.maketrans for a one-liner.", "maketrans maps each character to its complement."],
    time_limit=2
)

# LEVEL 2 — CORE (8 min)
p2 = TimedProblem(
    title="[MEDIUM] Finding All ORFs",
    problem="""
Write a function find_all_orfs(dna) that returns all Open Reading Frames
in all 6 reading frames (3 forward, 3 reverse complement).

An ORF: starts with ATG, ends with TAA/TAG/TGA, length >= 30 nt (10 codons).
Return a list of (frame, start, end, protein_sequence) tuples, sorted by length descending.

Example input: "ATGAAATTTGGG..." (at least 60 bp for meaningful results)
""",
    solution="""
CODON_TABLE = {
    'TTT':'F','TTC':'F','TTA':'L','TTG':'L','CTT':'L','CTC':'L','CTA':'L','CTG':'L',
    'ATT':'I','ATC':'I','ATA':'I','ATG':'M','GTT':'V','GTC':'V','GTA':'V','GTG':'V',
    'TCT':'S','TCC':'S','TCA':'S','TCG':'S','CCT':'P','CCC':'P','CCA':'P','CCG':'P',
    'ACT':'T','ACC':'T','ACA':'T','ACG':'T','GCT':'A','GCC':'A','GCA':'A','GCG':'A',
    'TAT':'Y','TAC':'Y','TAA':'*','TAG':'*','TGT':'C','TGC':'C','TGA':'*','TGG':'W',
    'CAT':'H','CAC':'H','CAA':'Q','CAG':'Q','AAT':'N','AAC':'N','AAA':'K','AAG':'K',
    'GAT':'D','GAC':'D','GAA':'E','GAG':'E','GGT':'G','GGC':'G','GGA':'G','GGG':'G',
}

def reverse_complement(seq):
    return seq.translate(str.maketrans('ATGC','TACG'))[::-1]

def find_orfs_in_frame(seq, frame_offset, frame_name):
    orfs = []
    for start in range(frame_offset, len(seq)-2, 3):
        if seq[start:start+3] == 'ATG':
            protein = ['M']
            end = start + 3
            while end + 2 < len(seq):
                codon = seq[end:end+3]
                aa = CODON_TABLE.get(codon, 'X')
                if aa == '*':
                    if (end+3 - start) >= 30:
                        orfs.append((frame_name, start, end+3, ''.join(protein)))
                    break
                protein.append(aa)
                end += 3
    return orfs

def find_all_orfs(dna):
    dna = dna.upper()
    rc = reverse_complement(dna)
    orfs = []
    for offset in range(3):
        orfs.extend(find_orfs_in_frame(dna, offset, f'+{offset+1}'))
        orfs.extend(find_orfs_in_frame(rc, offset, f'-{offset+1}'))
    return sorted(orfs, key=lambda x: len(x[3]), reverse=True)

import random; random.seed(42)
test_dna = ''.join(random.choices('ATGC', k=300))
orfs = find_all_orfs(test_dna)
print(f"Found {len(orfs)} ORFs in 300bp sequence")
for frame, start, end, prot in orfs[:3]:
    print(f"  Frame {frame}: pos {start}-{end}, {len(prot)} aa: {prot[:10]}...")
""",
    hints=[
        "Process 3 forward frames (offset 0,1,2) and 3 reverse frames on reverse_complement.",
        "In each frame, scan with step 3. When you hit ATG, start collecting. When you hit stop codon, save.",
        "Remember to track frame offset so codons stay in-frame after finding ATG."
    ],
    time_limit=8
)

# LEVEL 3 — ADVANCED (Rosalind hard variant, 15 min)
p3 = TimedProblem(
    title="[HARD] Rosalind PDST — Creating a Distance Matrix",
    problem="""
Given a collection of DNA strings in FASTA format, return the p-distance matrix
where p(s,t) = (number of positions where s and t differ) / length.

Input: dict of sequences {'Seq1': 'TTTCCATTTA', 'Seq2': 'GATTCATTTC', ...}
Output: n×n matrix as a 2D list, rounded to 5 decimal places.

Note: All sequences have the same length. Handle up to n=100 sequences of length 1000.

ROSALIND PROBLEM: http://rosalind.info/problems/pdst/
""",
    solution="""
def pdst(sequences):
    names = list(sequences.keys())
    seqs = [sequences[n] for n in names]
    n = len(seqs)
    L = len(seqs[0])
    matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            if i == j:
                row.append(0.0)
            else:
                diff = sum(a != b for a, b in zip(seqs[i], seqs[j]))
                row.append(round(diff / L, 5))
        matrix.append(row)
    return matrix, names

seqs = {
    'Seq1': 'TTTCCATTTA',
    'Seq2': 'GATTCATTTC',
    'Seq3': 'TTTCCATTTT',
    'Seq4': 'GTTCCATTTT',
}
matrix, names = pdst(seqs)
print("Distance matrix:")
for i, row in enumerate(matrix):
    print(f"  {names[i]}: {[round(v,5) for v in row]}")

# Performance: O(n^2 * L) — for n=100, L=1000: 10M operations, fast in Python
import time
import random
big_seqs = {f'S{i}': ''.join(random.choices('ACGT', k=1000)) for i in range(50)}
t0 = time.time()
m, _ = pdst(big_seqs)
print(f"\n50 sequences × 1000 bp: {(time.time()-t0)*1000:.1f}ms")
""",
    hints=[
        "p-distance = Hamming distance / length. Use zip() to compare positions.",
        "The matrix is symmetric: p(s,t) = p(t,s). You can compute upper triangle only and mirror.",
        "Performance tip: for very large inputs, use numpy: np.mean(seq1_array != seq2_array, axis=1)"
    ],
    time_limit=15
)

print("TIMED PRACTICE PROBLEMS LOADED")
print()
print("Usage:")
print("  p1.start()        # Begin timing, see the problem")
print("  # ... write your solution ...")
print("  p1.check_time()   # See how much time you've used")
print("  p1.hint(1)        # Get a hint if stuck")
print("  p1.reveal()       # See the solution")
print()
print("Problems available:")
print("  p1: [EASY]   DNA Complement (2 min)")
print("  p2: [MEDIUM] Finding All ORFs (8 min)")
print("  p3: [HARD]   Rosalind PDST — Distance Matrix (15 min)")
````

## 🔗 Cross-Module Integration Exercise

This problem requires combining algorithms from Modules 01, 02, and 08.
It mirrors the type of problem seen in bioinformatics take-home assessments.


````python
# CROSS-MODULE INTEGRATION: Assembly → ORF Finding → Translation
# Module 01 (De Bruijn assembly) → Module 02 (ORF finding) → Module 08 (optimization)

print("INTEGRATION CHALLENGE:")
print("=" * 60)
print("""
Given 15 simulated 30-mer reads from an unknown organism's gene:
  1. Assemble them into a contig using De Bruijn graph (Module 01/06)
  2. Find all ORFs in the contig (Module 08)
  3. Translate the longest ORF to protein (Module 02)
  4. Compute GC content and codon usage bias (Module 00)

This is a mini-pipeline mirroring real metagenomic assembly.
""")

import random
random.seed(99)

# The "true" underlying gene (simulated)
true_seq = ("ATGAAAGCAATTTTCGTACTGAAAGGTTTTGTTGGTTTTCTTAAATCGGATGGAAAG"
            "GCAGGTGGCCTTAAAGCAGGTGTTAAAGGCAAGGTGCAAATCGGTGGCGGTACTTGG"
            "GCTAAAGCAGGTGTGATTATTGGCAAAGCAGGTGTTGGCTTCTAAATCGTTTAAAGCC")

print(f"True gene length: {len(true_seq)} bp")

# Simulate 30-mer reads with 2× coverage (some reads are noisy)
read_length = 30
reads = []
for start in range(0, len(true_seq) - read_length + 1, 15):  # 2x coverage
    read = list(true_seq[start:start+read_length])
    # 5% chance of sequencing error at each position
    for i in range(len(read)):
        if random.random() < 0.02:
            read[i] = random.choice('ACGT')
    reads.append(''.join(read))

print(f"Generated {len(reads)} reads of length {read_length}")
print(f"First 3 reads: {reads[:3]}")

# ── STEP 1: De Bruijn Assembly ──
k = 15  # k-mer size

def debruijn_assemble(reads, k):
    """Simple De Bruijn graph assembly."""
    edges = {}
    for read in reads:
        for i in range(len(read) - k + 1):
            kmer = read[i:i+k]
            left, right = kmer[:-1], kmer[1:]
            if left not in edges:
                edges[left] = {}
            edges[left][right] = edges[left].get(right, 0) + 1

    # Find start node (in-degree 0 or dominant start)
    all_nodes = set(edges.keys())
    start = next(iter(all_nodes))

    # Greedy Eulerian path (simplified)
    contig = start
    current = start
    visited_edges = {}
    max_steps = 1000

    for _ in range(max_steps):
        if current not in edges or not edges[current]:
            break
        # Choose most frequent next k-mer
        next_node = max(edges[current], key=edges[current].get)
        if visited_edges.get((current, next_node), 0) >= edges[current][next_node]:
            break
        visited_edges[(current, next_node)] = visited_edges.get((current, next_node), 0) + 1
        contig += next_node[-1]
        current = next_node

    return contig

assembled = debruijn_assemble(reads, k)
print(f"\n── STEP 1: Assembly ──")
print(f"Assembled contig: {len(assembled)} bp")

# Check assembly quality
matches = sum(assembled[i:i+20] in true_seq for i in range(0, len(assembled)-19, 20))
print(f"Assembly accuracy (20-mer matches in true): {matches}/{len(assembled)//20}")

# ── STEP 2: ORF Finding ──
STOPS = {'TAA', 'TAG', 'TGA'}
def find_best_orf(seq):
    best = ('', 0)
    for offset in range(3):
        for i in range(offset, len(seq)-2, 3):
            if seq[i:i+3] == 'ATG':
                j = i + 3
                while j+2 < len(seq) and seq[j:j+3] not in STOPS:
                    j += 3
                if j - i > len(best[0]):
                    best = (seq[i:j], i)
    return best

orf, pos = find_best_orf(assembled)
print(f"\n── STEP 2: ORF Finding ──")
print(f"Longest ORF: {len(orf)} bp at position {pos}")
print(f"  ({len(orf)//3} codons = {len(orf)//3} amino acids)")

# ── STEP 3: Translation ──
CODON_TABLE = {
    'ATG':'M','TTT':'F','TTC':'F','TTA':'L','TTG':'L','CTT':'L','CTC':'L','CTA':'L',
    'CTG':'L','ATT':'I','ATC':'I','ATA':'I','GTT':'V','GTC':'V','GTA':'V','GTG':'V',
    'TCT':'S','TCC':'S','TCA':'S','TCG':'S','CCT':'P','CCC':'P','CCA':'P','CCG':'P',
    'ACT':'T','ACC':'T','ACA':'T','ACG':'T','GCT':'A','GCC':'A','GCA':'A','GCG':'A',
    'GGT':'G','GGC':'G','GGA':'G','GGG':'G','AAT':'N','AAC':'N','AAA':'K','AAG':'K',
    'GAT':'D','GAC':'D','GAA':'E','GAG':'E','CAT':'H','CAC':'H','CAA':'Q','CAG':'Q',
    'TGT':'C','TGC':'C','TGG':'W','TAT':'Y','TAC':'Y','CGT':'R','CGC':'R','CGA':'R',
    'CGG':'R','AGA':'R','AGG':'R',
}
protein = ''.join(CODON_TABLE.get(orf[i:i+3], 'X') for i in range(0, len(orf)-2, 3))
print(f"\n── STEP 3: Translation ──")
print(f"Protein: {protein}")
print(f"  Length: {len(protein)} aa")

# ── STEP 4: Sequence Statistics ──
gc = sum(1 for b in orf if b in 'GC') / len(orf) * 100
codon_usage = {}
for i in range(0, len(orf)-2, 3):
    codon_usage[orf[i:i+3]] = codon_usage.get(orf[i:i+3], 0) + 1

top_codons = sorted(codon_usage.items(), key=lambda x: -x[1])[:5]
print(f"\n── STEP 4: Statistics ──")
print(f"GC content: {gc:.1f}%")
print(f"Top 5 codons: {top_codons}")
print()
print("PIPELINE COMPLETE. Compare to true gene:")
true_protein = ''.join(CODON_TABLE.get(true_seq[i:i+3], 'X')
                       for i in range(0, len(true_seq)-2, 3)
                       if CODON_TABLE.get(true_seq[i:i+3]) != 'X'
                       and true_seq[i:i+3] not in STOPS)
overlap = sum(p == t for p, t in zip(protein, true_protein)) / max(len(protein), 1)
print(f"Protein identity vs true: {overlap*100:.0f}%")
````

Find and fix the **3 bugs** in the DNA analysis code below.

<details>
<summary>Show answer</summary>

**Bug 1 — Incomplete complement dict:** Missing the 'N' (any nucleotide) complement.
More critically, 'T' maps to 'A' but is missing from the dict — the mapping given
only has A→T, C→G, G→C, missing T→A. Fix: add `'T': 'A'` and `'N': 'N'`.

**Bug 2 — GC content formula:** `gc = (seq.count('G') + seq.count('C')) / len(seq)` is
correct. The buggy code uses `gc = seq.count('G') + seq.count('C') / len(seq)` —
missing parentheses causes operator precedence error: only C is divided by length.

**Bug 3 — Only 3 reading frames checked:** ORFs can be in the reverse complement too.
A full search needs all 6 frames (3 forward + 3 reverse complement).
Fix: also search `reverse_complement(seq)` with frames 0, 1, 2.
</details>


````python
# DEBUG EXERCISE — DNA string operations, find and fix 3 bugs

# BUG 1: complement dict is missing T->A and N->N
COMPLEMENT = {'A': 'T', 'C': 'G', 'G': 'C'}   # missing 'T': 'A' and 'N': 'N'

def reverse_complement(seq):
    return ''.join(COMPLEMENT.get(b, '?') for b in reversed(seq.upper()))

test_seq = 'ATCGN'
rc = reverse_complement(test_seq)
print(f"Reverse complement of '{test_seq}': '{rc}'")
print(f"Expected: 'NCGAT', got: '{rc}' (T and N give '?')")

# BUG 2: operator precedence — missing parentheses around (G+C) count
def gc_content(seq):
    # should be: (seq.count('G') + seq.count('C')) / len(seq)
    gc = seq.count('G') + seq.count('C') / len(seq)   # BUG: C/len added to G count
    return gc

for s in ['ATCG', 'GGCC', 'ATAT']:
    print(f"GC content of '{s}': {gc_content(s):.4f}  (expected {(s.count('G')+s.count('C'))/len(s):.4f})")

# BUG 3: ORF search only in 3 forward frames, missing 3 reverse-complement frames
def find_orfs(seq, min_len=10):
    orfs = []
    # Only searches forward strand — misses ORFs on reverse complement
    for frame in range(3):
        i = frame
        while i < len(seq) - 2:
            codon = seq[i:i+3]
            if codon == 'ATG':
                j = i + 3
                while j < len(seq) - 2:
                    stop = seq[j:j+3]
                    if stop in ('TAA', 'TAG', 'TGA'):
                        if j + 3 - i >= min_len:
                            orfs.append(seq[i:j+3])
                        break
                    j += 3
            i += 3
    return orfs

dna = 'ATGATCGATCGTAACATGCTAGCTAGTAA'
orfs = find_orfs(dna)
print(f"\nORFs found (3 frames only): {len(orfs)} — {orfs}")
print("BUG: reverse complement ORFs not checked — 3 more frames missing")
````
