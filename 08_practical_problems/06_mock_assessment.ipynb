```python
def find_all_orfs(s):
    STOP_CODONS = {'TAA', 'TAG', 'TGA'}
    orfs = []
    for i in range(len(s) - 2):
        if s[i:i+3] == 'ATG':
            # Walk in-frame from this ATG
            for j in range(i + 3, len(s) - 2, 3):
                codon = s[j:j+3]
                if codon in STOP_CODONS:
                    orfs.append(s[i:j+3])
                    break  # stop at first in-frame stop codon
    return orfs

def longest_orf(s):
    orfs = find_all_orfs(s)
    if not orfs:
        return ""
    return max(orfs, key=len)

def codon_frequencies(s):
    orf = longest_orf(s)
    if not orf:
        return {}
    # Codons: skip first codon (ATG at position 0)
    freq = {}
    for i in range(3, len(orf), 3):
        codon = orf[i:i+3]
        freq[codon] = freq.get(codon, 0) + 1
    return freq
```

```python
def lcs_length(a, b):
    m, n = len(a), len(b)
    # Two-row rolling array: O(n) space
    prev = [0] * (n + 1)
    curr = [0] * (n + 1)
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if a[i-1] == b[j-1]:
                curr[j] = prev[j-1] + 1
            else:
                curr[j] = max(prev[j], curr[j-1])
        prev, curr = curr, [0] * (n + 1)
    return prev[n]

def lcs_sequence(a, b):
    m, n = len(a), len(b)
    # Full O(mn) DP table for traceback
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if a[i-1] == b[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    # Traceback
    result = []
    i, j = m, n
    while i > 0 and j > 0:
        if a[i-1] == b[j-1]:
            result.append(a[i-1])
            i -= 1
            j -= 1
        elif dp[i-1][j] >= dp[i][j-1]:
            i -= 1
        else:
            j -= 1
    return ''.join(reversed(result))
```

**Complexity:** O(mn) time and O(mn) space for `lcs_sequence`; O(n) space for `lcs_length`.  
**Key insight:** When `a[i-1] == b[j-1]`, the LCS includes this character and shortens both sequences by 1.  
**Common mistake 1 (off-by-one):** Indexing `a[i]` instead of `a[i-1]` in the DP fill — the DP table is 1-indexed but the strings are 0-indexed.  
**Common mistake 2:** Forgetting `reversed()` on traceback — you build the result backwards.
</details>

```python
import math

def connected_components(adj):
    if not adj:
        return []
    visited = set()
    components = []

    def bfs(start):
        component = set()
        queue = [start]
        while queue:
            node = queue.pop()
            if node in visited:
                continue
            visited.add(node)
            component.add(node)
            queue.extend(n for n in adj.get(node, []) if n not in visited)
        return component

    for node in adj:
        if node not in visited:
            components.append(bfs(node))
    return sorted(components, key=len, reverse=True)

def degree_distribution(adj):
    freq = {}
    for node, neighbours in adj.items():
        d = len(neighbours)
        freq[d] = freq.get(d, 0) + 1
    return freq

def is_power_law(adj, min_degree=2):
    dd = degree_distribution(adj)
    # Filter to degrees >= min_degree with at least 1 node
    points = [(d, c) for d, c in dd.items() if d >= min_degree and c >= 1]
    if len(points) < 3:
        return False
    # Log-log transform
    xs = [math.log(d) for d, c in points]
    ys = [math.log(c) for d, c in points]
    n = len(xs)
    # Linear regression: y = a + b*x
    mean_x = sum(xs) / n
    mean_y = sum(ys) / n
    ss_xx = sum((x - mean_x)**2 for x in xs)
    ss_yy = sum((y - mean_y)**2 for y in ys)
    ss_xy = sum((xs[i] - mean_x) * (ys[i] - mean_y) for i in range(n))
    if ss_xx == 0 or ss_yy == 0:
        return False
    r2 = (ss_xy ** 2) / (ss_xx * ss_yy)
    return r2 >= 0.80
```

**Complexity:** `connected_components` O(V+E) with BFS/DFS; `degree_distribution` O(V); `is_power_law` O(V + k log k) where k = distinct degrees.  
**Common mistake 1 (power law):** Using `math.log(0)` when a degree count is 0 — filter to count >= 1 before taking log.  
**Common mistake 2 (BFS):** Using a list as a queue with `.pop(0)` (O(n) per pop) instead of a deque or using `.pop()` from the end (DFS, still correct, O(1)).  
**Common mistake 3 (connected components):** Forgetting isolated nodes — nodes in `adj` with no edges still exist (degree 0). The BFS starts from every unvisited key in `adj`.
</details>

## COMMON MISTAKES — Top 3 per Problem

Review these even if you passed all test cases.

### Problem 1 (DNA ORF)

| Mistake | Why it fails | Fix |
|---|---|---|
| Inner loop steps by 1 instead of 3 | Reads stop codons out of frame | `range(i+3, len(s)-2, 3)` — always step 3 |
| Missing `break` after finding stop | Returns the same ORF multiple times (once per stop codon after the first) | `break` after appending to `orfs` |
| Forgetting last codon boundary | `len(s)-2` is not enough; you need at least 3 chars for a codon | Add guard: `if j+3 <= len(s)` or use `s[j:j+3]` and check `len(codon)==3` |

### Problem 2 (LCS)

| Mistake | Why it fails | Fix |
|---|---|---|
| Off-by-one: `a[i]` instead of `a[i-1]` | DP table is 1-indexed (extra row/col for empty prefix); strings are 0-indexed | Always use `a[i-1]` and `b[j-1]` inside the loop |
| Forgetting `reversed()` in traceback | Traceback builds the LCS backwards (from end to start) | `''.join(reversed(result))` at the end |
| Using rolling array for `lcs_sequence` | Rolling array loses the full table needed for traceback | Use full O(mn) table for `lcs_sequence`; rolling array only for `lcs_length` |

### Problem 3 (Graph + Statistics)

| Mistake | Why it fails | Fix |
|---|---|---|
| `math.log(0)` in power-law check | log(0) is undefined; raises ValueError | Filter to counts >= 1 before computing log |
| Using `list.pop(0)` for BFS queue | O(n) per pop → O(n²) total; TLE on large graphs | Use `collections.deque` with `popleft()`, or use DFS with `pop()` from right |
| Forgetting to sort components by size | Problem says largest first; unsorted output fails ordered equality checks | `sorted(components, key=len, reverse=True)` |

## 🎤 Interview Questions

**Q1 (Easy):** What is the time complexity of a nested loop iterating over two arrays of length n and m?
<details><summary>Answer</summary>O(n × m). If n ≈ m, this simplifies to O(n²). For coding interviews, always state the complexity of your solution.</details>

**Q2 (Easy):** When should you use dynamic programming instead of recursion?
<details><summary>Answer</summary>When the problem has overlapping subproblems (same subproblem solved multiple times). DP stores results in a table to avoid redundant computation, reducing exponential time to polynomial.</details>

**Q3 (Medium):** You have 30 minutes left and 2 problems unsolved. What's your strategy?
<details><summary>Answer</summary>Quickly assess which problem has a simpler brute-force solution. Submit the brute force for partial credit (it may pass small test cases). Then attempt the harder problem with the remaining time. Partial credit > no credit.</details>

**Q4 (Medium):** Your solution passes sample cases but fails hidden tests. How do you debug?
<details><summary>Answer</summary>Test edge cases: empty input, single element, maximum constraints, duplicates, negative numbers. Check for off-by-one errors in indexing. Verify integer overflow (use Python's arbitrary precision or explicit casting). Print intermediate values for the failing case.</details>

**Q5 (Hard):** Design an O(n log n) solution for finding the longest increasing subsequence.
<details><summary>Answer</summary>Maintain a list `tails` where `tails[i]` is the smallest tail element for an increasing subsequence of length i+1. For each element, binary search for its position in `tails`. This gives O(n log n) vs the O(n²) DP approach.</details>
